# 11. Brainstorm — Nghĩ Trước, Làm Sau

**Part 3: Superpowers — Quy trình làm việc** | Bước 1 trong 5 bước Superpowers

---

> *"Không ai xây nhà mà không có bản vẽ."*

Brainstorm là bước đầu tiên — và quan trọng nhất — trong quy trình Superpowers.  
Đây là lúc bạn **nói chuyện với AI** để hai bên cùng hiểu rõ trước khi bắt tay vào làm.

## Brainstorm là gì? Một cuộc "họp kế hoạch" với AI

Hãy tưởng tượng bạn muốn **xây một căn nhà**:

- Bạn có thuê thợ xây rồi nói "xây đi" mà không có bản vẽ không? **Chắc chắn không.**
- Bạn sẽ ngồi lại với kiến trúc sư, bàn bạc: bao nhiêu phòng, cửa sổ hướng nào, ngân sách bao nhiêu...

Brainstorm với AI **y hệt vậy**:

| Xây nhà | Làm phần mềm với AI |
|---------|---------------------|
| Họp với kiến trúc sư | Brainstorm với AI |
| Nói rõ yêu cầu: 3 phòng ngủ, 2 WC | Nói rõ yêu cầu: trang đăng ký, xác nhận email |
| Kiến trúc sư hỏi lại: "Tầng mấy?" | AI hỏi lại: "Nếu email sai thì sao?" |
| Ra bản vẽ → bắt đầu xây | Ra kế hoạch → bắt đầu code |

**Thời gian:** Chỉ 5-10 phút. Không tốn nhiều thời gian nhưng tiết kiệm HÀNG GIỜ sửa sai sau này.

## Mục đích của Brainstorm

Trong 5-10 phút brainstorm, bạn và AI sẽ cùng:

1. **Phân tích phạm vi (Scope):** Tính năng này bao gồm những gì? Không bao gồm gì?
2. **Nhận diện rủi ro (Risks):** Điều gì có thể sai? Lỗi nào hay xảy ra?
3. **Xử lý trường hợp đặc biệt (Edge Cases):** Nếu người dùng nhập sai? Nếu mạng chập chờn?

### Ví dụ thực tế

Bạn nói: *"Tôi muốn trang đăng ký, cần xác nhận email, có trial 14 ngày"*

AI sẽ hỏi lại:
- *"Nếu email không hợp lệ thì hiện thông báo gì?"*
- *"Trial 14 ngày hết hạn thì chuyện gì xảy ra? Khóa tài khoản hay giới hạn tính năng?"*
- *"Cho phép đăng ký bằng Google/Facebook không?"*

Bạn trả lời → AI hiểu rõ hơn → Code tốt hơn ngay từ đầu.

## Khi nào BẮT BUỘC brainstorm?

### Phải brainstorm:
- Tính năng mới trung bình trở lên (trang đăng ký, hệ thống thanh toán...)
- Thay đổi kiến trúc (đổi database, thêm service mới...)
- Tích hợp nhiều module (đăng ký + email + thanh toán cùng lúc)

### Được phép bỏ qua:
- Sửa bug nhỏ (lỗi chính tả, sai màu nút)
- Chỉnh UI đơn giản (đổi font, thay ảnh)
- Thay đổi config (đổi tên app, đổi logo)

**Quy tắc ngón tay cái:** Nếu bạn có thể giải thích thay đổi trong 1 câu → bỏ qua brainstorm.  
Nếu cần 2 câu trở lên → nên brainstorm.

## Quy trình 3 bước

### Bước 1: Question Assumptions — Hỏi lại giả định (10 phút)

Đây là lúc bạn **đặt câu hỏi** cho chính mình và AI:

- "Mình có chắc người dùng cần tính năng này không?"
- "Giả định rằng mọi người có email — nhưng nếu không thì sao?"
- "Mình nghĩ chỉ cần 1 trang — nhưng có cần trang xác nhận không?"

**Ví dụ thực tế:**  
Bạn muốn tạo tính năng "đặt lịch hẹn".  
Giả định: *"Ai cũng biết múi giờ của mình."*  
AI hỏi: *"Nếu người dùng ở múi giờ khác thì hiển thị giờ nào?"*  
→ Phát hiện vấn đề TRƯỚC khi code, không phải SAU.

### Bước 2: Explore Trade-offs — Cân nhắc các phương án (15 phút)

Mỗi tính năng thường có **nhiều cách làm**. Bước này giúp chọn cách tốt nhất.

Ví dụ: Tính năng xác nhận email

| Phương án | Ưu điểm | Nhược điểm |
|-----------|---------|------------|
| **A: Gửi link xác nhận** | An toàn, phổ biến | Chậm hơn, cần server gửi email |
| **B: Gửi mã OTP 6 số** | Nhanh, dễ nhập trên điện thoại | Ít an toàn hơn |
| **C: Không xác nhận** | Nhanh nhất | Nhiều email rác, không an toàn |

Bạn không cần hiểu kỹ thuật — chỉ cần **chọn**: "Tôi muốn phương án A vì an toàn hơn."

AI sẽ implement đúng phương án bạn chọn.

### Bước 3: Document Decision — Ghi lại quyết định (10 phút)

Ghi lại quyết định quan trọng để **sau này nhớ tại sao mình chọn vậy**.

Không cần viết dài — chỉ cần ghi:

```
Quyết định: Xác nhận email bằng link (phương án A)
Lý do: An toàn hơn OTP, phù hợp với đối tượng doanh nghiệp
Bỏ qua: Phương án C (không xác nhận) vì lo ngại spam
Ngày: 2026-04-04
```

AI sẽ giúp bạn tạo **Decision Log** — một bản ghi các quyết định phi kỹ thuật quan trọng.  
6 tháng sau nhìn lại, bạn biết ngay tại sao mình chọn phương án này.

## Cho người không biết code: Brainstorm = Nói cho AI hiểu bạn muốn gì

Bạn không cần biết code để brainstorm. Bạn chỉ cần **nói rõ** bạn muốn gì.

### Cách nói hiệu quả:

**Không tốt:** *"Làm trang đăng ký đi"*  
(AI sẽ làm... nhưng thiếu rất nhiều chi tiết)

**Tốt hơn:** *"Tôi cần trang đăng ký cho khách hàng doanh nghiệp. Cần xác nhận email, có trial 14 ngày, sau trial tự chuyển sang gói trả phí. Không cho phép email cá nhân (gmail, yahoo)."*  
(AI hiểu rõ, sẽ hỏi thêm vài câu rồi bắt đầu)

### Mẫu câu bạn có thể dùng:

- *"Tôi muốn [tính năng], với [yêu cầu đặc biệt]"*
- *"Khi [tình huống] xảy ra, tôi muốn [hành vi]"*
- *"Người dùng của tôi là [ai], họ cần [gì]"*
- *"Tôi KHÔNG muốn [điều này] xảy ra"*

## Ví dụ cuộc brainstorm thực tế

Đây là ví dụ một cuộc brainstorm giữa bạn và AI:

---

**Bạn:** *Tôi muốn tạo trang quản lý học sinh. Giáo viên có thể thêm, sửa, xóa học sinh.*

**AI:** *Tốt. Để tôi hỏi thêm vài điều:*
- *Một giáo viên quản lý bao nhiêu học sinh? (10? 100? 1000?)*
- *"Xóa" là xóa hẳn hay chỉ ẩn đi (xóa mềm)?*
- *Học sinh có thuộc nhiều lớp không?*
- *Cần tìm kiếm/lọc học sinh không?*

**Bạn:** *Khoảng 50-100 học sinh. Xóa mềm thôi (ẩn đi, không xóa hẳn). Mỗi học sinh thuộc 1 lớp. Cần tìm kiếm theo tên.*

**AI:** *Rõ. Tôi đề xuất:*
- *Danh sách dạng bảng, có phân trang (50 học sinh/trang)*
- *Ô tìm kiếm ở trên cùng*
- *Nút "Xóa" sẽ ẩn học sinh, có thể khôi phục sau*
- *Cần xác nhận trước khi xóa*

**Bạn:** *OK, làm theo đó đi.*

---

Chỉ 5 phút, nhưng AI đã hiểu rõ **rất nhiều chi tiết** mà nếu không hỏi sẽ bị thiếu.

## Decision Log — Nhật ký quyết định

Sau brainstorm, AI sẽ tạo một bản ghi quyết định. Ví dụ:

```
# Decision Log — Trang Quản lý Học sinh

## Ngày: 2026-04-04

### Quyết định 1: Xóa mềm
- Chọn: Xóa mềm (soft delete) — ẩn học sinh thay vì xóa hẳn
- Lý do: Tránh mất dữ liệu, giáo viên có thể khôi phục
- Bỏ qua: Xóa cứng — rủi ro mất dữ liệu vĩnh viễn

### Quyết định 2: Phân trang
- Chọn: 50 học sinh/trang
- Lý do: Cân bằng giữa tốc độ tải và tiện dùng
- Bỏ qua: Tải tất cả 1 lần — chậm khi nhiều học sinh

### Quyết định 3: Tìm kiếm
- Chọn: Tìm theo tên (client-side filter)
- Lý do: Dưới 1000 học sinh, lọc phía client đủ nhanh
```

Bản ghi này **cực kỳ giá trị** khi 3 tháng sau bạn hoặc người khác muốn hiểu "tại sao làm vậy?"

## Mẹo brainstorm hiệu quả

### Nên:
- Nói rõ **ai** sẽ dùng tính năng (giáo viên? học sinh? quản trị viên?)
- Nói rõ **giới hạn** (tối đa bao nhiêu? cho phép gì? cấm gì?)
- Nói rõ **khi lỗi thì sao** (mạng chập chờn? nhập sai? hết hạn?)
- Đề cập **tính năng liên quan** (trang này liên kết với trang nào?)

### Không nên:
- Nói quá chung chung: *"Làm đẹp đi"* — AI không biết "đẹp" là gì
- Bỏ qua brainstorm rồi phàn nàn kết quả sai — bạn chưa nói rõ yêu cầu mà
- Nói quá chi tiết về kỹ thuật (bạn không cần) — hãy nói về **kết quả mong muốn**

## Sai lầm phổ biến: Nhảy thẳng vào code

Đây là sai lầm #1 khi làm việc với AI:

```
Bạn: "Code cho tôi trang đăng ký"
AI: *viết code ngay*
Bạn: "Không phải, tôi muốn có xác nhận email"
AI: *sửa lại*
Bạn: "Quên nữa, cần trial 14 ngày"
AI: *sửa lại LẦN NỮA*
Bạn: "Và không cho phép email gmail"
AI: *đập đi xây lại vì cấu trúc cũ không phù hợp*
```

**Kết quả:** Mất 2 giờ sửa đi sửa lại.

**Nếu brainstorm trước:** 10 phút nói rõ tất cả → AI làm đúng ngay lần đầu → Xong trong 30 phút.

**10 phút brainstorm tiết kiệm 90 phút sửa sai.**

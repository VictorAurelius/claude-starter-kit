## Templates là gì?

Hãy nghĩ về **form đơn** ở cơ quan.

Khi bạn cần xin phép nghỉ, bạn không cần tự viết từ đầu:
- Lấy **form có sẵn** (đã có tiêu đề, chỗ ký tên, mục lý do...)
- Điền thông tin vào chỗ trống
- Nộp

**Templates trong Kit cũng vậy:**
- File mẫu đã có **cấu trúc sẵn**
- AI (hoặc bạn) chỉ cần **điền nội dung** cụ thể của dự án
- Đảm bảo mọi file đều có **format nhất quán**

## 5 Templates trong Kit

|----------|------|----------|
| CLAUDE.md | `CLAUDE.md.template` | Hướng dẫn AI cho dự án |

| VS Code | `vscode-settings.json.template` | Cài đặt VS Code |

### Template 1: CLAUDE.md

Đây là template quan trọng nhất (xem chi tiết ở Notebook 06).

Cách dùng:
1. Copy: `cp .claude/templates/CLAUDE.md.template ./CLAUDE.md`
2. Mở file, thay `[placeholder]` bằng thông tin dự án
3. Xóa section không cần (dự án nhỏ có thể bỏ bớt)
4. Commit vào git

````bash
# Xem CLAUDE.md template (rút gọn)

cat << 'TEMPLATE'
# Claude Code Instructions

## CRITICAL: Communication Language
**ALWAYS communicate in [Vietnamese/English]**

## Project Overview
**[Project Name]** — [Mô tả ngắn]
- Backend: [công nghệ]
- Frontend: [công nghệ]
- Database: [công nghệ]

## Development Workflow
1. Quick Brainstorm
2. Task Breakdown
3. TDD - Test First
4. Implementation
5. Code Review

## Git Workflow
- Branch naming: [convention]
- NEVER commit directly to main

## Folder Structure
[Mô tả cấu trúc folder dự án]

## Skills Reference
Index: .claude/skills/_README-skills-index.md
TEMPLATE

echo ""
echo "--- Thay cac [placeholder] bang thong tin du an cua ban ---"
````

### Template 2: README.md

README = trang giới thiệu dự án, người mới đọc đầu tiên.

Template bao gồm:
- Tên dự án + mô tả
- Công nghệ sử dụng
- Cách cài đặt
- Cấu trúc folder
- Cách chạy dự án

Dùng cho cả **người** (developer mới) và **AI** (Claude Code đọc để hiểu dự án).

### Template 3: settings.local.json

File này quy định **quyền hạn** của Claude Code — được phép làm gì.

```json
{
  "permissions": {
    "allow": [
      "Read files",
      "Write files",
      "Run scripts"
    ],
    "deny": [
      "Delete directories",
      "Push to main branch"
    ]
  }
}
```

Giống như bảo vệ ở cổng: "Nhân viên được vào tầng 1-3, KHÔNG được vào tầng 4."

### Template 4: VS Code Settings

Cài đặt cho VS Code (trình soạn thảo code phổ biến).

Template cấu hình sẵn:
- Format code tự động khi save
- Extension gợi ý cài đặt
- Theme, font phù hợp

Giúp mọi người trong team dùng **cùng cài đặt**, code trông giống nhau.

### Template 5: Skill Folder

Mẫu để tạo skill mới:

```
skill-folder/
├── SKILL.md              <- Entry point (< 100 dòng)
└── reference/
    └── detail.md         <- Chi tiết (đọc khi cần)
```

Khi bạn nhờ AI "tạo skill kiểm tra SEO", AI sẽ dùng template này để tạo đúng cấu trúc.

````bash
# Xem skill template

cat << 'SKILLTEMPLATE'
---
name: [tên-skill]
description: >
  [Khi nào dùng skill này? Liệt kê các từ khóa trigger.]
  Examples: '[từ khóa 1]', '[từ khóa 2]', '[từ khóa 3]'
---

# [Tên Skill]

## Mục đích
[1-2 câu mô tả mục đích]

## Quy trình
1. [Bước 1]
2. [Bước 2]
3. [Bước 3]

## Gotchas
- [Bẫy 1 - điều dễ sai]
- [Bẫy 2 - điều cần chú ý]

## Skill Contents
- `reference/detail.md` — Chi tiết tiêu chí (đọc khi scoring)
SKILLTEMPLATE

echo ""
echo "--- Thay [placeholder] bang noi dung cu the ---"
````

## Rules là gì?

Nếu Templates là "form mẫu", thì Rules là **"nội quy"** hướng dẫn AI khi tạo nội dung mới.

Kit có 1 file rules chính: `rules/skill-conventions.md`

File này dạy AI cách viết skill đúng chuẩn **Anthropic best practices** — chuẩn chính thức từ công ty tạo ra Claude.

### Rule 1: Progressive Disclosure — Load theo nhu cầu

| Layer | Khi nào load | Token cost |
|-------|-------------|------------|
| Tên + mô tả | Luôn luôn | ~100 tokens |
| SKILL.md body | Khi kích hoạt | 200-500 tokens |
| Reference files | Khi cần tra cứu | 0 cho đến khi đọc |

**Ví von:** Giống đọc sách — đọc **mục lục** (nhanh) → mở đúng **chương cần đọc** (chi tiết).

Tại sao quan trọng? Claude Code có giới hạn "bộ nhớ làm việc" (context window). Load ít = suy nghĩ nhanh + chính xác hơn.

### Rule 2: Description = Trigger Condition

Phần `description` trong skill KHÔNG phải viết cho người đọc — mà viết cho **AI đọc** để biết khi nào kích hoạt skill.

```yaml
# SAI — mô tả chung chung, AI khó nhận biết khi nào dùng
description: "Run quality audit with 11 categories"

# ĐÚNG — liệt kê cụ thể khi nào dùng
description: >
  Use when user says 'audit', 'quality check', 'ready to merge?',
  'kiểm tra chất lượng'. Runs 11-category checks with 100-point scoring.
```

Mẹo:
- Liệt kê **từ khóa** người dùng hay nói
- Bao gồm **nhiều ngôn ngữ** (tiếng Việt + tiếng Anh)
- Mô tả **tình huống** khi nào dùng

### Rule 3: KHÔNG dạy AI điều nó đã biết

Claude Code đã biết lập trình, biết TDD, biết debug. **Đừng tốn token dạy lại.**

```markdown
# SAI — 100+ dòng giải thích TDD là gì (AI đã biết)
## TDD = Test Driven Development
Step 1: Write a failing test first
Step 2: Write minimum code to pass
Step 3: Refactor while keeping tests green
[... 80 dòng ví dụ generic ...]

# ĐÚNG — chỉ viết gotchas CỤ THỂ cho dự án
## TDD Gotchas cho dự án này
- Luôn mock speechSynthesis.cancel — test crash nếu thiếu
- Guard window access — SSR không có window
- Controller test PHẢI có @ActiveProfiles("test")
```

Nguyên tắc: Chỉ viết những gì **Claude không thể biết từ training** — tức là thông tin riêng của dự án.

````bash
# So sánh skill tốt vs skill xấu

echo "=== SKILL XẤU (50+ dòng, đa số generic) ==="
echo ""
echo "# Code Review Skill"
echo "## What is code review?"
echo "Code review is the process of examining code... (10 dong)"
echo "## Why code review?"
echo "Code review helps find bugs early... (10 dong)"
echo "## How to review?"
echo "1. Read the code carefully... (20 dong)"
echo "-> AI DA BIET het roi! Lang phi token."
echo ""
echo "========================================"
echo ""
echo "=== SKILL TỐT (20 dòng, toàn project-specific) ==="
echo ""
echo "# Code Review — Project Gotchas"
echo "## Checklist rieng du an nay:"
echo "- BaseEntity fields KHONG xuat hien trong @Builder"
echo "- @ActiveProfiles('test') BAT BUOC cho controller test"
echo "- Deprecated constructor: dung new Object[0] de force varargs"
echo "- TenantContext.setCurrentTenant() TRUOC moi DB operation trong test"
echo "-> Day la thong tin AI KHONG THE biet tu training!"
````

### Rule 4: SKILL.md < 100 dòng

File SKILL.md (entry point) phải ngắn gọn:
- Dưới **100 dòng**
- Chỉ chứa: mô tả, quy trình ngắn, gotchas, link đến reference
- Chi tiết (bảng điểm, ví dụ dài...) → để trong `reference/`

Tại sao? Vì SKILL.md được load vào "bộ nhớ làm việc" khi kích hoạt. File ngắn = AI còn chỗ suy nghĩ.

### Rule 5: Gotchas = Phần giá trị nhất

Mỗi skill nên có section **Gotchas** — đây là nơi ghi lại bẫy thực tế.

```markdown
## Gotchas
- Dev server PHẢI chạy port 5174 trước khi capture
- Screenshots dùng page.reload() — tăng timeout nếu theme sai
- Vocabulary page hiện trắng nếu không có lesson data — KHÔNG phải bug
```

Gotchas được cập nhật liên tục — mỗi lần gặp bẫy mới, thêm vào.

## Tại sao Templates + Rules quan trọng?

| Lợi ích | Giải thích |
|---------|------------|
| **Nhất quán** | Mọi file cùng format, dễ đọc, dễ maintain |
| **Tiết kiệm token** | Không tốn token dạy AI format, cấu trúc |
| **Chất lượng** | AI tạo skill mới theo đúng chuẩn Anthropic |
| **Tốc độ** | Copy template + điền = xong, nhanh hơn viết từ đầu |
| **Onboarding** | Người mới (hoặc AI phiên mới) hiểu ngay cấu trúc |

## Cách dùng trong thực tế

### Khi bắt đầu dự án mới:
1. Copy Kit vào dự án: `.claude/`
2. Copy template: `cp templates/CLAUDE.md.template ./CLAUDE.md`
3. Điền thông tin dự án vào CLAUDE.md
4. Copy settings: `cp templates/settings.local.json.template .claude/settings.local.json`
5. Bắt đầu làm việc!

### Khi cần tạo skill mới:
1. Bảo AI: "Tạo skill [mô tả]"
2. AI dùng template + rules để tạo đúng cấu trúc
3. Review + bổ sung gotchas theo thời gian

### Khi cần tạo README:
1. Copy template: `cp templates/README.md.template ./README.md`
2. Bảo AI: "Điền README dựa trên code hiện tại"
3. AI đọc code + điền vào template

## Bạn (non-dev) cần biết gì?

1. **Templates giúp bạn bắt đầu nhanh** — không cần biết format, chỉ cần điền nội dung

3. **Rules giúp AI tạo nội dung đúng chuẩn** — bạn không cần nhớ rules, AI tự tuân theo
4. **Gotchas tích lũy theo thời gian** — mỗi lần gặp vấn đề, thêm gotcha = dự án ngày càng tốt
5. **Bạn có thể yêu cầu AI tạo template mới** — "Tạo template cho báo cáo hàng tuần"

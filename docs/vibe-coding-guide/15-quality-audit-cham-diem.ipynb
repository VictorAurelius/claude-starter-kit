# 15. Quality Audit — Chấm Điểm Tổng

**Part 3: Superpowers — Quy trình làm việc** | Bước cuối: Đánh giá tổng thể

---

> *"Thanh tra trường học: kiểm tra 10 tiêu chí, chấm điểm, xếp hạng."*

Quality Audit là bước **chấm điểm tổng** cho toàn bộ dự án.  
Không chỉ 1 tính năng, mà **cả dự án** được đánh giá theo 10 tiêu chí, tổng 100 điểm.

## Quality Audit là gì? "Thanh tra trường học"

Hãy tưởng tượng thanh tra đến kiểm tra một trường học:

| Thanh tra trường học | Quality Audit phần mềm |
|---------------------|------------------------|
| Chất lượng giảng dạy | App chạy đúng không? |
| An ninh trường học | Bảo mật tốt không? |
| Cơ sở vật chất | Hạ tầng server ổn không? |
| Tài liệu giáo trình | Tài liệu đầy đủ không? |
| Quản lý hành chính | Quản lý dự án tốt không? |
| Chấm /100, xếp hạng A-D | Chấm /100, xếp hạng A+-D |

**Khi nào chạy Quality Audit?**
- Trước khi release phiên bản mới
- Trước khi merge (gộp) một loạt thay đổi lớn
- Bất kỳ khi nào muốn biết: "Dự án đang ở mức nào?"

## Khác nhau: Code Review vs Quality Audit

| | Code Review (bài 14) | Quality Audit (bài này) |
|---|---------------------|------------------------|
| **Phạm vi** | 1 PR (1 tính năng) | Toàn bộ dự án |
| **Khi nào** | Sau mỗi lần code | Trước release, khi cần đánh giá |
| **Mục đích** | Tính năng này OK chưa? | Toàn bộ dự án ở mức nào? |
| **Thời gian** | 20-30 phút | 30-60 phút |
| **Kết quả** | PASS/FAIL + issues | Điểm /100 + xếp hạng |

Code Review = kiểm tra **1 bài thi**.  
Quality Audit = kiểm tra **cả năm học**.

## 10 tiêu chí — Giải thích cho người không biết code

Mỗi tiêu chí tối đa 10 điểm. Tổng = 100 điểm.

Dưới đây, mỗi tiêu chí được giải thích bằng ngôn ngữ đời thường:

### 1. E2E Functionality — App chạy được đầu cuối không? (10 điểm)

**Ý nghĩa:** Người dùng có thể hoàn thành công việc từ đầu đến cuối không?

**Ví dụ đời thực:** Đặt hàng online
- Chọn sản phẩm → thêm giỏ hàng → thanh toán → nhận email xác nhận
- Nếu bất kỳ bước nào bị kẹt → FAIL

| Điểm | Mô tả |
|------|-------|
| 9-10 | Mọi chức năng hoạt động trơn tru |
| 7-8 | Hầu hết OK, vài lỗi nhỏ |
| 5-6 | Chức năng chính chạy, phụ bị lỗi |
| <5 | Nhiều chức năng không dùng được |

### 2. Security — Có an toàn không? (10 điểm)

**Ý nghĩa:** Dữ liệu người dùng có được bảo vệ không? Có ai hack được không?

**Ví dụ đời thực:** Khóa cửa nhà
- Mật khẩu có được mã hóa không? (không lưu dạng chữ thường)
- Đăng nhập sai 5 lần → khóa tạm?
- Người lạ có xem được dữ liệu người khác không?

**Tiêu chí:** Mã hóa mật khẩu, phân quyền đúng, chống tấn công phổ biến, HTTPS.

### 3. Backend Tests — Phần server có test không? (10 điểm)

**Ý nghĩa:** "Bộ máy bên trong" có được kiểm tra kỹ không?

**Ví dụ đời thực:** Kiểm tra động cơ xe
- Xe nhìn đẹp bên ngoài, nhưng bên trong — động cơ, hệ thống phanh, hộp số — có được kiểm tra không?
- Backend = "bộ máy" xử lý dữ liệu. Nếu không test → bug ẩn.

**Tiêu chí:** Tỷ lệ test coverage (bao nhiêu % code có test), test có chạy tự động không.

### 4. Frontend Tests — Phần giao diện có test không? (10 điểm)

**Ý nghĩa:** Giao diện người dùng thấy có được kiểm tra không?

**Ví dụ đời thực:** Kiểm tra nội thất xe
- Nút bấm có hoạt động không? Đồng hồ chỉ đúng không?
- Frontend = phần người dùng nhìn thấy và tương tác.

**Tiêu chí:** Test các component, test form, test navigation.

### 5. CI/CD — Tự động build/test/deploy tốt không? (10 điểm)

**Ý nghĩa:** Quy trình từ "viết code" đến "đưa lên internet" có tự động không?

**Ví dụ đời thực:** Dây chuyền sản xuất nhà máy
- Thủ công: Mỗi bước phải có người đứng làm → chậm, dễ sai
- Tự động: Dây chuyền chạy tự → nhanh, ít sai

**Tiêu chí:**
- CI (Continuous Integration): Mỗi lần push code → tự động chạy test
- CD (Continuous Deployment): Test pass → tự động đưa lên server

### 6. UI/UX — Giao diện đẹp, dễ dùng không? (10 điểm)

**Ý nghĩa:** Người dùng có thấy dễ hiểu, dễ dùng không?

**Ví dụ đời thực:** Bảng chỉ đường trong siêu thị
- Rõ ràng, dễ tìm, không bị lạc = UI/UX tốt
- Lung tung, phải hỏi nhân viên = UI/UX kém

**Tiêu chí:** Nhất quán (các trang giống style), responsive (xem được trên điện thoại), dễ hiểu (không cần hướng dẫn).

### 7. DevOps — Hạ tầng chạy ổn không? (10 điểm)

**Ý nghĩa:** Server, database, các dịch vụ phụ trợ có ổn định không?

**Ví dụ đời thực:** Hệ thống điện, nước trong tòa nhà
- Không ai thấy ống nước hay dây điện, nhưng nếu hỏng → cả tòa nhà ảnh hưởng
- DevOps = "hạ tầng ngầm" giữ cho app chạy ổn

**Tiêu chí:** Docker scripts chuẩn, monitoring (theo dõi sức khỏe server), backup dữ liệu.

### 8. Documentation — Tài liệu đầy đủ không? (10 điểm)

**Ý nghĩa:** Người mới vào có thể hiểu dự án không? Tài liệu có cập nhật không?

**Ví dụ đời thực:** Sổ tay nhân viên
- Có sổ tay → nhân viên mới tự đọc, tự làm
- Không sổ tay → phải hỏi từng chút, tốn thời gian

**Tiêu chí:** README rõ ràng, business docs đầy đủ (3 layer), cập nhật mỗi PR.

### 9. Code Quality — Code sạch không? (10 điểm)

**Ý nghĩa:** Code dễ đọc, dễ sửa, không có "mùi" (code smell) không?

**Ví dụ đời thực:** Bếp nhà hàng
- Bếp sạch, dụng cụ ngăn nắp → nấu nhanh, ít sai
- Bếp bẩn, đồ lung tung → nấu chậm, dễ nhầm

**Tiêu chí:** Không trùng lặp, đặt tên rõ ràng, cấu trúc hợp lý, không có TODO bị bỏ quên.

### 10. Project Management — Quản lý dự án tốt không? (10 điểm)

**Ý nghĩa:** Dự án có kế hoạch rõ ràng, tiến độ đúng hạn không?

**Ví dụ đời thực:** Tổ chức đám cưới
- Có timeline, checklist, phân công → suôn sẻ
- Không kế hoạch → thiếu hoa, quên mời bạn, nhạc sai

**Tiêu chí:** Git workflow chuẩn, PR descriptions rõ ràng, task tracking (theo dõi tiến độ), commit messages có ý nghĩa.

## Thang điểm xếp hạng

| Xếp hạng | Điểm | Ý nghĩa | Tương đương |
|----------|------|---------|-------------|
| **A+** | 95-100 | Xuất sắc | Nhà hàng 5 sao |
| **A** | 90-94 | Rất tốt | Nhà hàng 4 sao |
| **B+** | 85-89 | Tốt | Nhà hàng tốt, một vài điểm chưa hoàn hảo |
| **B** | 80-84 | Khá | Đáng tin cậy, còn cải thiện được |
| **C** | 70-79 | Trung bình | Chạy được, nhưng nhiều chỗ cần sửa |
| **D** | <70 | Cần cải thiện nhiều | Chưa nên release |

**Mục tiêu:** Đạt B+ (85+) trước khi release. A (90+) là lý tưởng.

## Đọc báo cáo audit — Bạn cần biết gì?

Báo cáo sẽ có:

### 1. Bảng điểm tổng (như trên)
Nhìn nhanh: dự án ở mức nào?

### 2. Chi tiết từng tiêu chí
Mỗi tiêu chí có giải thích: tại sao được điểm đó, cần cải thiện gì.

```
Frontend Tests: 7/10
  Reason: Test coverage 65% (target: 80%)
  Missing: No tests for error handling in product page
  Action: Add 5 more test cases for error scenarios
```

### 3. Danh sách cải thiện (ưu tiên)

```
TOP PRIORITIES:
1. [Frontend Tests] Add error handling tests (+8 points possible)
2. [Security] Add rate limiting for login API (+5 points possible)
3. [UI/UX] Fix responsive layout on mobile (+3 points possible)
```

**Bạn chỉ cần:** Nhìn bảng điểm → Đọc top priorities → Nói AI sửa theo thứ tự.

## UI Review riêng — Chấm giao diện /128 điểm

Ngoài Quality Audit tổng, Kit còn có **UI Review** — chuyên kiểm tra giao diện.

UI Review chấm điểm theo **5 chiều**, mỗi chiều tối đa khác nhau:

| Chiều | Tối đa | Ý nghĩa dễ hiểu |
|-------|--------|------------------|
| **Kỹ thuật** | 32 điểm | Responsive, tải nhanh, không lỗi console |
| **Thiết kế** | 32 điểm | Layout hợp lý, spacing đều, typography nhất quán |
| **Thẩm mỹ** | 24 điểm | Màu sắc hài hòa, nhìn chuyên nghiệp |
| **Thân thiện** | 24 điểm | Dễ dùng, người dùng biết bấm đâu |
| **Accessibility** | 16 điểm | Người khuyết tật dùng được (đọc màn hình, bàn phím) |
| **Tổng** | **128 điểm** | |

### Cách chạy:

AI sẽ chụp ảnh màn hình (screenshot) từng trang, so sánh trước/sau khi sửa, và chấm điểm.

**Ví dụ kết quả:**

```
UI REVIEW: Login Page
  Kỹ thuật:    28/32  (responsive OK, tải nhanh)
  Thiết kế:    26/32  (spacing hơi lệch ở mobile)
  Thẩm mỹ:    20/24  (màu sắc OK, logo hơi mờ)
  Thân thiện:  22/24  (rõ ràng, dễ dùng)
  Accessibility: 12/16 (thiếu aria-label cho icon)
  ─────────────────────
  Total: 108/128 (84%)
```

## Khi nào nên chạy audit?

| Thời điểm | Lý do |
|-----------|-------|
| Trước release | Đảm bảo chất lượng đủ tốt |
| Sau khi merge wave (đợt thay đổi lớn) | Kiểm tra không có regression |
| Hàng tuần (nếu dự án đang phát triển) | Theo dõi xu hướng chất lượng |
| Khi onboard người mới | Cho họ thấy dự án đang ở mức nào |
| Khi chuẩn bị demo | Đảm bảo không bị xấu hổ |

**Không cần chạy:**
- Sau mỗi commit nhỏ (Code Review đã đủ)
- Khi chỉ sửa tài liệu

## Vòng lặp cải thiện

Quality Audit không phải chạy 1 lần rồi thôi. Đây là **vòng lặp**:

```
Chạy audit → 84/100 (B)
     |
     v
Xem top priorities → sửa 3 vấn đề hàng đầu
     |
     v
Chạy audit lại → 89/100 (B+)
     |
     v
Sửa tiếp 2 vấn đề → 92/100 (A)
     |
     v
Release!
```

Mỗi lần chạy, điểm sẽ tăng. Bạn thấy **sự tiến bộ rõ ràng** bằng con số.

## Cho người không biết code: Tóm gọn

Bạn **không cần hiểu chi tiết** 10 tiêu chí. Bạn chỉ cần:

1. **Gõ `/quality-audit`** khi muốn kiểm tra
2. **Đọc điểm tổng:** 90+ = tốt, 80-89 = khá, <80 = cần cải thiện
3. **Đọc top priorities:** AI gợi ý sửa gì trước
4. **Nói AI sửa:** "Hãy sửa 3 vấn đề ưu tiên cao nhất"
5. **Chạy audit lại:** Xem điểm tăng chưa

**Cần nhớ:** Audit là "nhiệt kế" đo sức khỏe dự án. Chạy đều đặn, theo dõi xu hướng.

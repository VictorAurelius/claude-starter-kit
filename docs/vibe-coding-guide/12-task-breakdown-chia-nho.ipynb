Sau khi brainstorm xong, bước tiếp theo là **chia công việc lớn thành nhiều bước nhỏ**.  
AI làm từng bước nhỏ chính xác hơn rất nhiều so với làm "một cục" lớn.

## Task Breakdown là gì? Danh sách mua sắm cho AI

Hãy tưởng tượng bạn nhờ ai đó đi chợ:

**Cách 1 — Mơ hồ:**  
*"Đi chợ mua đồ nấu cơm nhé."*  
→ Người đó sẽ mua gì? Không biết. Có thể mua thiếu, mua thừa, mua sai.

**Cách 2 — Rõ ràng:**  
*"Mua: 500g thịt bò, 1 bó rau muống, 2 quả cà chua, nước mắm (chai nhỏ)."*  
→ Rõ ràng, chính xác, không sai.

Task Breakdown = **viết danh sách mua sắm cho AI**.

| Đi chợ | Làm phần mềm |
|--------|---------------|
| Danh sách mua sắm | Danh sách tasks |
| "500g thịt bò" | "Tạo form đăng nhập" |
| "1 bó rau muống" | "Validate email" |
| Mua xong từng món → check | Xong từng task → check |

## Tại sao quan trọng cho người không biết code?

### 1. Theo dõi được tiến độ

Thay vì: *"AI đang làm... không biết bao lâu nữa xong"*  
Bạn thấy: *"Task 3/10 xong. Còn 7 tasks nữa."*

### 2. Nếu sai, chỉ sai 1 bước nhỏ

Không chia nhỏ: AI làm cả trang web → sai → **phải làm lại TẤT CẢ**  
Có chia nhỏ: AI làm task 3 sai → **chỉ sửa task 3** → tiết kiệm thời gian

### 3. Không bị "choáng" với project lớn

Nhìn cả project: *"Trời ơi, nhiều quá, biết bắt đầu từ đâu?"*  
Nhìn task 1: *"OK, chỉ cần tạo form thôi. Được."*

**Ví von:** Ăn voi phải ăn từng miếng. Không ai nuốt cả con voi.

## Mức độ chi tiết — Mỗi task bao lớn?

Mỗi task nên mất **2-5 phút** để AI hoàn thành.

| Mức độ | Ví dụ | Đánh giá |
|--------|-------|----------|
| Quá nhỏ (<2 phút) | "Thêm 1 dòng import" | Gộp vào task khác |
| Vừa đủ (2-5 phút) | "Tạo form đăng nhập với 2 trường" | Hoàn hảo |
| Quá lớn (>10 phút) | "Làm cả hệ thống đăng nhập" | Chia nhỏ hơn |

### Quy tắc:
- Quá nhỏ → **gộp lại** thành 1 task
- Quá lớn → **chia tiếp** thành 2-3 task nhỏ hơn
- Nếu không biết task có quá lớn không → **hỏi AI**: *"Task này có nên chia nhỏ hơn không?"*

## 5 yếu tố của mỗi task

Mỗi task tốt cần có 5 thông tin:

| # | Yếu tố | Giải thích đơn giản | Ví dụ |
|---|--------|---------------------|-------|
| 1 | **File cụ thể** | Sửa ở đâu? | `login-page.tsx` |
| 2 | **Thay đổi cụ thể** | Làm gì? | Thêm form đăng nhập |
| 3 | **Code mẫu** | Trông như thế nào? | Form có 2 ô: email, mật khẩu |
| 4 | **Cách kiểm tra** | Biết xong chưa? | Mở trang, thấy form, nhập thử |
| 5 | **Thời gian** | Mất bao lâu? | ~5 phút |

**Bạn không cần tự viết 5 yếu tố này!** AI sẽ tạo cho bạn.  
Bạn chỉ cần **đọc qua** và nói "OK" hoặc "Sửa task 3 giúp tôi".

## Ví dụ thực tế: Tạo trang đăng nhập

Bạn nói: *"Tạo trang đăng nhập"*

AI chia thành 5 tasks:

### Task 1: Tạo form giao diện (5 phút)
- **Làm gì:** Tạo form với ô email, ô mật khẩu, nút "Đăng nhập"
- **Kiểm tra:** Mở trang → thấy form đẹp, có 2 ô nhập và 1 nút

### Task 2: Validate input — kiểm tra dữ liệu nhập (3 phút)
- **Làm gì:** Nếu email trống → hiện "Vui lòng nhập email". Nếu email sai định dạng → hiện "Email không hợp lệ"
- **Kiểm tra:** Bấm "Đăng nhập" khi chưa nhập gì → thấy lỗi màu đỏ

### Task 3: Gọi API đăng nhập (5 phút)
- **Làm gì:** Khi bấm "Đăng nhập", gửi email + mật khẩu lên server kiểm tra
- **Kiểm tra:** Đăng nhập đúng → vào trang chủ. Sai → hiện "Sai email hoặc mật khẩu"

### Task 4: Xử lý lỗi (3 phút)
- **Làm gì:** Nếu mạng lỗi → hiện "Không thể kết nối". Nếu server lỗi → hiện "Hệ thống đang bảo trì"
- **Kiểm tra:** Tắt mạng → bấm đăng nhập → thấy thông báo lỗi

### Task 5: Viết test tự động (5 phút)
- **Làm gì:** Viết bài kiểm tra tự động để đảm bảo mọi thứ hoạt động
- **Kiểm tra:** Chạy test → tất cả PASS (xanh)

### Tổng kết ví dụ:

```
Task 1: Tạo form UI ............ 5 phút  [ ]
Task 2: Validate input ......... 3 phút  [ ]
Task 3: Gọi API đăng nhập ..... 5 phút  [ ]
Task 4: Xử lý lỗi ............. 3 phút  [ ]
Task 5: Viết test .............. 5 phút  [ ]
─────────────────────────────────────────
Tổng: 5 tasks ................. ~21 phút
```

Thay vì 1 yêu cầu mơ hồ "tạo trang đăng nhập", giờ bạn có **5 bước rõ ràng** với thời gian cụ thể.

## Theo dõi tiến độ

Khi AI làm, bạn có thể theo dõi:

```
Task 1: Tạo form UI ............ 5 phút  [x] Xong!
Task 2: Validate input ......... 3 phút  [x] Xong!
Task 3: Gọi API đăng nhập ..... 5 phút  [ ] << Đang làm
Task 4: Xử lý lỗi ............. 3 phút  [ ]
Task 5: Viết test .............. 5 phút  [ ]
```

**Bạn biết chính xác:**
- Đang ở đâu (task 3/5)
- Còn bao lâu (~8 phút nữa)
- Nếu task 3 sai → chỉ sửa task 3, task 1 và 2 vẫn OK

## Điều chỉnh danh sách tasks

Bạn hoàn toàn có quyền **điều chỉnh** danh sách tasks:

| Tình huống | Bạn nói |
|------------|--------|
| Task thừa | *"Bỏ task 4 đi, không cần xử lý lỗi mạng"* |
| Task thiếu | *"Thêm task: trang quên mật khẩu"* |
| Task quá lớn | *"Task 3 chia nhỏ hơn được không?"* |
| Đổi thứ tự | *"Làm task 5 trước task 3 được không?"* |
| Thay đổi yêu cầu | *"Task 2 đổi thành validate số điện thoại thay vì email"* |

**Bạn là "giám đốc dự án", AI là "nhân viên".** Bạn quyết định danh sách, AI thực hiện.

## Sai lầm phổ biến

### Sai lầm 1: Không chia nhỏ, giao cả cục

```
Bạn: "Làm cả hệ thống quản lý trường học đi"
AI: *viết hàng trăm dòng code* → nhiều lỗi, khó sửa
```

**Sửa:** Chia thành: trang danh sách → trang thêm mới → trang sửa → ...

### Sai lầm 2: Chia quá nhỏ

```
Task 1: Thêm dòng 1 của form
Task 2: Thêm dòng 2 của form
Task 3: Thêm dòng 3 của form  ← Quá chi tiết!
```

**Sửa:** Gộp thành: "Tạo form hoàn chỉnh"

### Sai lầm 3: Không kiểm tra sau mỗi task

Làm xong task 1 → không kiểm tra → làm task 2 → task 1 sai → task 2 cũng sai theo

**Sửa:** Mỗi task xong → kiểm tra → OK → mới tiếp

## Ví dụ lớn hơn: Hệ thống đặt lịch hẹn

Sau brainstorm, AI chia thành:

```
Phase 1: Giao diện cơ bản (30 phút)
  Task 1.1: Trang danh sách lịch hẹn ......... 5 phút
  Task 1.2: Form tạo lịch hẹn mới ............ 5 phút
  Task 1.3: Chọn ngày/giờ (calendar picker) ... 5 phút
  Task 1.4: Hiển thị chi tiết lịch hẹn ........ 5 phút
  Task 1.5: Nút hủy lịch hẹn .................. 3 phút
  Task 1.6: Test cho phase 1 ................... 7 phút

Phase 2: Logic phía server (25 phút)
  Task 2.1: API tạo lịch hẹn .................. 5 phút
  Task 2.2: Kiểm tra trùng giờ ................ 5 phút
  Task 2.3: API hủy lịch hẹn .................. 3 phút
  Task 2.4: Gửi email xác nhận ................ 5 phút
  Task 2.5: Test cho phase 2 ................... 7 phút

Phase 3: Hoàn thiện (15 phút)
  Task 3.1: Tìm kiếm/lọc lịch hẹn ............ 5 phút
  Task 3.2: Phân trang ......................... 3 phút
  Task 3.3: Xử lý lỗi + thông báo ............. 4 phút
  Task 3.4: Test tổng hợp ...................... 3 phút
─────────────────────────────────────────────────────
Tổng: 14 tasks, 3 phases .................. ~70 phút
```

**Từ một yêu cầu "hệ thống đặt lịch" → 14 bước cụ thể, rõ ràng, có thời gian.**

## Tóm tắt

| Điều cần nhớ | Chi tiết |
|--------------|----------|
| **Task Breakdown là gì** | Chia công việc lớn → nhiều bước nhỏ (2-5 phút/task) |
| **Tại sao cần** | Theo dõi tiến độ, dễ sửa sai, không bị choáng |
| **Bạn cần làm gì** | Đọc danh sách tasks, điều chỉnh nếu cần, nói OK |
| **AI làm gì** | Tạo danh sách, ước lượng thời gian, thực hiện từng task |
| **Kích cỡ phù hợp** | 2-5 phút/task. Quá nhỏ → gộp. Quá lớn → chia tiếp |

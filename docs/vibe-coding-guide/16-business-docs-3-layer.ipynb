**Vibe Coding Guide cho nguoi KHONG biet lap trinh**

> AI code gioi, nhung AI KHONG biet san pham cua ban hoat dong the nao.
> **Ban** la nguoi hieu san pham nhat — va day la luc ban toa sang.

## Business Docs la gi?

Hay tuong tuong ban mo mot **nha hang**.

Truoc khi phuc vu khach, ban can:
- **So tay noi quy**: Gio mo cua, so luong ban, quy dinh dat truoc
- **Kich ban phuc vu**: Khach vao → tiep don → goi mon → bep nau → phuc vu → thanh toan
- **Menu**: Mon gi, gia bao nhieu, combo nao, mon het thi bao sao

Ba thu nay giup **bat ky ai** — du la nhan vien moi hay nguoi thay ca — deu biet cach van hanh nha hang.

**Business Docs cung y het vay**, nhung danh cho phan mem:
- Mo ta san pham hoat dong the nao
- Quy tac gi phai tuan theo
- Nguoi dung lam gi, he thong phan hoi the nao

Va quan trong nhat: **AI doc docs nay de code dung theo yeu cau cua ban**.

## Tai sao day la viec NON-DEV gioi nhat?

Nghe co ve nghich ly, nhung nguoi **khong biet code** lai viet Business Docs tot hon developer:

| Nguoi viet | Diem manh | Diem yeu |
|-----------|----------|----------|
| **Ban (non-dev)** | Hieu san pham, biet khach hang can gi | Khong biet code |
| **Developer** | Biet code | Hay bo qua nghiep vu, tap trung ky thuat |
| **AI** | Code nhanh | KHONG biet san pham cua ban |

**Quy trinh ly tuong:**
1. Ban viet Business Docs (mo ta san pham)
2. AI doc docs → hieu yeu cau
3. AI code dung theo nhung gi ban viet

**KHONG CO docs = AI tu bia logic → SAI**

Vi du: Ban muon "hoc sinh chi duoc dang ky toi da 5 lop". Neu khong viet ra, AI se khong biet gioi han nay va code se khong co kiem tra nao ca.

## 3 Layer — 3 goc nhin cho moi chuc nang

Moi phan chuc nang (goi la **domain**) can 3 files, moi file mo ta tu mot goc nhin khac nhau:

| Layer | File | Vi von nha hang | Tra loi cau hoi |
|-------|------|----------------|------------------|
| 1 | `rules.md` | Noi quy nha hang | "Quy tac la gi?" |
| 2 | `use-cases.md` | Kich ban phuc vu | "Ai lam gi, thu tu nao?" |
| 3 | `api-contract.md` | Menu nha hang | "Goi mon gi, nhan duoc gi?" |

Ba files nay tao thanh **bo tai lieu day du** de AI (hoac bat ky developer nao) hieu chinh xac chuc nang can lam gi.

## Layer 1: rules.md — Quy tac kinh doanh

### Vi von: Nhu "noi quy" cua nha hang

- Gio mo cua: 10h-22h
- Toi da 4 nguoi/ban
- Dat truoc phai xac nhan qua dien thoai
- Huy dat ban truoc 2 tieng

### Trong phan mem, rules.md chua:

- Cac **rang buoc** ma he thong phai tuan theo
- Cac **tham so** co the thay doi (goi la **config key**)
- Format bang de de theo doi

### Vi du: Domain "Dang ky hoc"

| ID | Quy tac | Chi tiet | Config Key |
|----|---------|---------|------------|
| BR-ENR-001 | Gioi han si so | Toi da 50 hoc sinh/lop | `enrollment.max-per-class` |
| BR-ENR-002 | Thanh toan truoc | Phai dong phi truoc khi xac nhan | `enrollment.require-payment` |
| BR-ENR-003 | Khong trung lich | Hoc sinh khong duoc dang ky 2 lop cung gio | — |
| BR-ENR-004 | Han dang ky | Dang ky truoc khai giang 3 ngay | `enrollment.deadline-days` |

### Config Key la gi?

La **tham so co the thay doi** ma khong can sua code. Vi du:
- Hom nay: toi da 50 hoc sinh/lop
- Thang sau: doi thanh 40
- Chi can thay doi config, khong can viet lai code

**Quy tac vang: KHONG BAO GIO hardcode** (viet cung so vao code). Luon dung config key.

## Layer 2: use-cases.md — Kich ban su dung

### Vi von: Nhu "kich ban phim"

Mot bo phim co:
- **Nhan vat**: ai xuat hien?
- **Canh quay**: lam gi, thu tu nao?
- **Tinh huong bat ngo**: neu loi thi xu ly sao?

### use-cases.md mo ta:

- **Actor** — Ai thuc hien? (phu huynh, hoc sinh, admin...)
- **Steps** — Cac buoc thuc hien theo thu tu
- **Errors** — Khi loi thi bao gi, xu ly sao
- **FE behavior** — Giao dien phan ung the nao (nut bam, thong bao...)

### Vi du: UC-ENR-001 — Phu huynh dang ky lop cho con

```
Actor: Phu huynh (da dang nhap)

Steps:
1. Vao trang "Danh sach lop"
2. Chon lop muon dang ky
3. He thong kiem tra: con slot khong? trung lich khong? da dong phi chua?
4. Neu OK → xac nhan dang ky
5. He thong gui email xac nhan

Errors:
- Lop day (het slot) → Hien "Lop da du si so" + nut "Dang ky lop khac"
- Trung lich → Hien "Trung lich voi lop X" + link den lop X
- Chua dong phi → Hien "Vui long thanh toan" + nut chuyen trang thanh toan

FE behavior:
- Nut "Dang ky" disable khi dang xu ly (tranh bam 2 lan)
- Loading spinner khi kiem tra slot
- Toast "Dang ky thanh cong!" khi xong
```

Ban thay khong? **Khong can biet code** van viet duoc hoan toan. Day la ngon ngu cua san pham, cua nguoi dung.

## Layer 3: api-contract.md — Hop dong API

### Vi von: Nhu "menu nha hang"

Menu cho ban biet:
- **Mon gi co**: Pho bo, bun cha, com tam...
- **Gia bao nhieu**: 45k, 55k, 65k...
- **Goi thi nhan duoc gi**: 1 to pho + rau + nuoc
- **Mon het thi sao**: "Xin loi, mon X da het"

### api-contract.md mo ta:

- **Endpoint** — "dia chi" de goi (nhu so ban de goi mon)
- **Request** — gui len cai gi (nhu "goi mon gi")
- **Response** — nhan lai cai gi (nhu "dia com")
- **Error codes** — loi thi bao gi (nhu "het mon")

### Vi du: Dang ky hoc

```
POST /api/enrollments

Request:
{
  "studentId": 123,
  "classId": 456
}

Response (thanh cong):
{
  "id": 789,
  "studentId": 123,
  "classId": 456,
  "status": "CONFIRMED",
  "enrolledAt": "2026-03-15T10:00:00"
}

Errors:
- 409 CLASS_FULL — Lop da du si so
- 409 SCHEDULE_CONFLICT — Trung lich
- 402 PAYMENT_REQUIRED — Chua thanh toan
- 400 DEADLINE_PASSED — Qua han dang ky
```

**Luu y:** Day la file ky thuat nhat trong 3 files. Neu ban khong quen, ban co the nho AI tao skeleton (khung mau) tu rules.md va use-cases.md — AI rat gioi viec nay.

## Verification Chain — Day chuyen kiem tra

3 files khong chi de doc — chung **lien ket voi nhau** tao thanh mot chuoi kiem tra:

```
rules.md → use-cases.md → api-contract.md → Code → Test
```

### Vi von: Nhu day chuyen san xuat

Trong nha may, moi khau kiem tra khau truoc:
- Nguyen lieu (rules) → che bien dung quy trinh (use-cases) → thanh pham dung mau (api) → dong goi (code) → kiem tra chat luong (test)

### Cu the:

| Buoc | File/Code | Kiem tra gi |
|------|----------|-------------|
| 1 | `rules.md` | BR-ENR-001 noi "toi da 50 hoc sinh" |
| 2 | `use-cases.md` | UC-ENR-001 step 3 co kiem tra slot |
| 3 | `api-contract.md` | Error 409 CLASS_FULL co trong danh sach |
| 4 | Code | Controller tra ve 409 khi lop day |
| 5 | Test | Test case kiem tra lop du 50 nguoi → 409 |

Neu **bat ky khau nao thieu**, nghia la co lo hong. Vi du:
- rules.md co "toi da 50" nhung use-cases.md khong co buoc kiem tra → **gap**
- api-contract.md co error 409 nhung code khong tra ve → **bug**
- Code dung nhung khong co test → **risk**

## Tai sao 3 files, khong gop 1?

Cau hoi hay! Ly do:

### 1. De review rieng
- Product owner review rules.md ("quy tac nay dung chua?")
- UX designer review use-cases.md ("flow nay hop ly chua?")
- Developer review api-contract.md ("API nay implement duoc chua?")

### 2. De cap nhat
- Doi gioi han tu 50 → 40? Chi sua rules.md
- Them buoc xac nhan email? Chi sua use-cases.md
- Them error code moi? Chi sua api-contract.md

### 3. De trace (truy vet)
- Co bug? Tim tu error code trong api-contract.md → tim use-case tuong ung → tim rule goc
- Nhanh hon nhieu so voi doc 1 file dai hang tram dong

### 4. AI xu ly tot hon
- AI doc 3 files nho tot hon 1 file khong lo
- Moi file co muc dich ro rang → AI hieu chinh xac hon

````bash
# Xem noi dung mau cua rules.md
echo "=== Noi dung mau: rules.md (Domain: Dang ky hoc) ==="
echo ""
cat << 'SAMPLE'
# Enrollment — Business Rules

## Rules

| ID          | Rule              | Detail                                    | Config Key                    |
|-------------|-------------------|-------------------------------------------|-------------------------------|
| BR-ENR-001  | Max students      | Max 50 students per class                 | enrollment.max-per-class      |
| BR-ENR-002  | Payment required  | Must pay before enrollment confirmed      | enrollment.require-payment    |
| BR-ENR-003  | No schedule clash | Student cannot enroll in overlapping class| —                             |
| BR-ENR-004  | Deadline          | Must enroll 3 days before class starts    | enrollment.deadline-days      |
| BR-ENR-005  | Max classes       | Student max 5 classes per semester        | enrollment.max-classes        |

## Status Lifecycle

PENDING → CONFIRMED → COMPLETED
PENDING → CANCELLED
CONFIRMED → DROPPED
SAMPLE
````

````bash
# Xem noi dung mau cua use-cases.md
echo "=== Noi dung mau: use-cases.md (Domain: Dang ky hoc) ==="
echo ""
cat << 'SAMPLE'
# Enrollment — Use Cases

## UC-ENR-001: Phu huynh dang ky lop cho con

**Actor:** Phu huynh (da dang nhap)
**Precondition:** Hoc sinh da co tai khoan, lop dang mo dang ky

**Steps:**
1. Vao trang "Danh sach lop dang mo"
2. Chon lop, bam "Dang ky"
3. He thong kiem tra: slot, lich, han dang ky
4. Hien trang thanh toan (neu require-payment = true)
5. Phu huynh thanh toan
6. He thong xac nhan, gui email

**Errors:**
- CLASS_FULL → "Lop da du si so. Vui long chon lop khac."
- SCHEDULE_CONFLICT → "Trung lich voi lop [ten]. Xem chi tiet."
- DEADLINE_PASSED → "Da qua han dang ky cho lop nay."

**FE Behavior:**
- Disable nut "Dang ky" khi dang xu ly
- Spinner khi kiem tra dieu kien
- Toast "Dang ky thanh cong!" + redirect ve danh sach lop cua hoc sinh
SAMPLE
````

````bash
# Xem noi dung mau cua api-contract.md
echo "=== Noi dung mau: api-contract.md (Domain: Dang ky hoc) ==="
echo ""
cat << 'SAMPLE'
# Enrollment — API Contract

## POST /api/v1/enrollments

**Description:** Dang ky hoc sinh vao lop
**Auth:** Bearer token (role: PARENT)

**Request:**
```json
{
  "studentId": 123,
  "classId": 456
}
```

**Response 201:**
```json
{
  "id": 789,
  "studentId": 123,
  "classId": 456,
  "status": "CONFIRMED",
  "enrolledAt": "2026-03-15T10:00:00Z"
}
```

**Errors:**
| Code | Error Key          | Message                    | Ref       |
|------|--------------------|----------------------------|-----------|
| 409  | CLASS_FULL         | Lop da du si so            | BR-ENR-001|
| 409  | SCHEDULE_CONFLICT  | Trung lich voi lop khac    | BR-ENR-003|
| 402  | PAYMENT_REQUIRED   | Chua thanh toan            | BR-ENR-002|
| 400  | DEADLINE_PASSED    | Qua han dang ky            | BR-ENR-004|
```
SAMPLE
````

## Quy tac vang: Docs va Code PHAI cung PR

Khi ban thay doi logic san pham:
- Doi gioi han hoc sinh tu 50 → 40
- Them buoc xac nhan email
- Them error code moi

Thi **docs va code phai cap nhat DONG THOI** trong cung 1 PR (Pull Request).

**Tai sao?**
- Neu chi doi code ma khong doi docs → lan sau AI doc docs cu → code sai
- Neu chi doi docs ma khong doi code → san pham van chay theo logic cu
- Cung PR = review 1 lan, khong bi lo

**Nho:** Business Docs la "nguon su that" (source of truth). Code phai phan anh dung docs. Docs phai phan anh dung thuc te.

## Loi thuong gap

| Loi | Hau qua | Cach tranh |

| Khong viet docs, nhay thang vao code | AI tu bia logic, phai sua nhieu lan | Luon viet rules.md truoc |

| Gop 3 files thanh 1 | Kho review, kho cap nhat, kho trace | Giu nguyen 3 files |

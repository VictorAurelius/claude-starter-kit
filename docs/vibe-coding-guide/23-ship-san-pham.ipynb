# 23. Ship Sản Phẩm — Đưa Sản Phẩm Đến Tay Người Dùng

**Phần 5: Thực hành từ đầu đến cuối**

---

Bạn đã xây xong sản phẩm, test kỹ, quality check đạt. Giờ đến bước cuối cùng: **Ship** — đưa sản phẩm lên internet để mọi người có thể dùng.

Hãy nghĩ về ngày **khai trương cửa hàng**: bạn đã trang trí xong, hàng hóa bày biện đẹp, nhân viên sẵn sàng. Việc còn lại là **mở cửa** và đón khách.

Ship sản phẩm cũng vậy — code đã viết xong, giờ bạn "mở cửa" cho thế giới vào xem.

## Checklist Trước Khi Ship

Trước khi khai trương, bạn sẽ kiểm tra lần cuối. Giống như checklist mở cửa hàng:

- Hàng hóa đủ chưa? (Tính năng hoàn thiện)
- Bảng hiệu có đẹp không? (Giao diện OK)
- Camera an ninh hoạt động? (Bảo mật)
- Nhân viên biết việc? (Test pass)

Đây là checklist cho sản phẩm phần mềm:

````bash
cat << 'EOF'
=== CHECKLIST TRUOC KHI SHIP ===

1. [  ] Quality Audit >= 85 diem (B+)?
2. [  ] Business Gap Check — 0 P1 (critical) gaps?
3. [  ] UI Review — cac trang chinh hien thi dung?
4. [  ] CI/CD green — tat ca test pass?
5. [  ] Documentation up-to-date?
6. [  ] Security basics OK?

=> Tat ca 6 muc deu tick? San sang ship!
EOF
````

### Chi Tiết Từng Mục

````bash
# Muc 1: Quality Audit
# Trong Claude Code, goi:
# /quality-audit

cat << 'EOF'
=== 1. QUALITY AUDIT ===

Goi "/quality-audit" trong Claude Code.

Ket qua can dat:
  >= 90 (A):  Tuyet voi, ship ngay
  >= 85 (B+): Tot, co the ship
  >= 75 (B):  Tam duoc, nen fix them truoc khi ship
  <  75:      Chua nen ship, can fix nhieu thu

Vi du:
  Quality Audit: 88/100 (B+)
  => DAT! Co the ship.
EOF
````

````bash
# Muc 2: Business Gap Check

cat << 'EOF'
=== 2. BUSINESS GAP CHECK ===

Kiem tra: co thieu tinh nang nghiep vu quan trong nao khong?

P1 (Critical): PHAI co truoc khi ship
  Vd: Trang thanh toan khong hoat dong -> KHONG SHIP

P2 (Major): Nen co, nhung co the ship truoc, fix sau
  Vd: Chua co sort/filter san pham -> Ship duoc, fix sau

P3 (Minor): Co thi tot, khong co cung duoc
  Vd: Chua co dark mode -> Ship binh thuong

Quy tac: P1 = 0 moi duoc ship!
EOF
````

````bash
# Muc 6: Security basics

cat << 'EOF'
=== 6. SECURITY BASICS ===

Kiem tra:
  [  ] Khong co API key/password trong code?
  [  ] .env file da gitignore?
  [  ] HTTPS (khoa xanh tren browser)?
  [  ] Input validation (nguoi dung khong nhap ma doc)?

Noi voi Claude:
"Kiem tra security basics truoc khi ship, 
 dac biet la API keys va .env files."

Claude se quet code va bao neu co van de.
EOF
````

## Deploy — Đưa Sản Phẩm Lên Internet

"Deploy" = đưa code từ máy bạn lên server để mọi người truy cập được.

Có nhiều cách deploy, từ đơn giản (kéo thả) đến phức tạp (tự setup server). Chúng ta sẽ xem qua 3 cấp độ:

### Cấp Độ 1: Đơn Giản Nhất — Vercel / Netlify

Phù hợp cho: Landing page, blog, portfolio, frontend apps

**Vercel** (vercel.com) hoặc **Netlify** (netlify.com) là dịch vụ "kéo thả" — bạn kết nối GitHub repo, nó tự build và deploy.

Cách làm:
1. Đăng ký tài khoản (miễn phí)
2. Kết nối GitHub repo
3. Chọn framework (Next.js, React, Vue...)
4. Nhấn Deploy
5. Chờ 1-2 phút → có link `your-app.vercel.app`

Giống như đăng bài lên Facebook — bạn chọn ảnh, viết caption, nhấn Post. Xong!

````bash
# Deploy voi Vercel (neu da cai Vercel CLI)

cat << 'EOF'
=== DEPLOY VOI VERCEL ===

# Cai Vercel CLI (1 lan)
$ npm install -g vercel

# Deploy
$ vercel

  Vercel CLI
  ? Set up and deploy? [Y/n] Y
  ? Which scope? your-username
  ? Link to existing project? N
  ? Project name: my-coffee-shop
  ? Framework: Next.js
  
  Building...
  Deploying...
  Ready! Deployed to:
  https://my-coffee-shop.vercel.app

=> Xong! Bat ky ai cung co the vao link nay de xem san pham cua ban.
EOF
````

### Cấp Độ 2: Full-Stack — Railway / Render

Phù hợp cho: App có backend (database, API, authentication)

**Railway** (railway.app) hoặc **Render** (render.com) cho phép deploy cả frontend lẫn backend.

Cách làm:
1. Đăng ký tài khoản
2. Kết nối GitHub repo
3. Cấu hình: database, environment variables
4. Deploy

Phức tạp hơn Vercel một chút, nhưng Claude có thể hướng dẫn bạn từng bước.

````bash
cat << 'EOF'
=== DEPLOY VOI RAILWAY ===

# Cai Railway CLI (1 lan)
$ npm install -g @railway/cli

# Dang nhap
$ railway login

# Tao project moi
$ railway init

# Them database (PostgreSQL)
$ railway add --plugin postgresql

# Deploy
$ railway up

  Deploying...
  Build Logs:
    Installing dependencies...
    Building...
  Deploy Logs:
    Server running on port 3000
  
  Deployment URL:
  https://my-coffee-shop.up.railway.app

=> App cua ban da online, co ca database!
EOF
````

### Cấp Độ 3: Production Serious — AWS / Oracle Cloud

Phù hợp cho: Sản phẩm "thật sự" với nhiều người dùng, cần ổn định cao

Starter Kit đã có sẵn **Terraform configs** (trong thư mục `infrastructure/`) để deploy lên AWS hoặc Oracle Cloud.

Cấp độ này phức tạp hơn nhiều — thường cần developer hoặc DevOps engineer hỗ trợ. Nhưng khi dự án lớn lên, đây là bước tất yếu.

Bạn chỉ cần biết: **công cụ đã có sẵn trong kit**, khi cần sẽ dùng.

## Post-Ship — Sau Khi Khai Trương

Ship xong không phải là hết. Giống như cửa hàng mở rồi, bạn vẫn cần:

### 1. Monitor — Theo dõi

App chạy có lỗi không? Có chậm không? Có ai truy cập không?

Các nền tảng (Vercel, Railway) đều có dashboard để bạn xem:
- Số người truy cập
- Có lỗi nào xảy ra không
- Tốc độ load trang

````bash
cat << 'EOF'
=== MONITORING CO BAN ===

Nhung thu can theo doi sau khi ship:

1. Error rate: Co bao nhieu loi xay ra?
   => 0 loi = tuyet voi
   => < 1% loi = binh thuong
   => > 5% loi = can sua gap

2. Response time: Trang load nhanh khong?
   => < 1 giay = tot
   => 1-3 giay = chap nhan
   => > 3 giay = can toi uu

3. Uptime: App co chay lien tuc khong?
   => 99.9% = tot (chi down 8.7 gio/nam)
   => < 99% = can xem lai infrastructure

Vercel/Railway co dashboard de xem nhung so nay.
EOF
````

### 2. Feedback — Thu Thập Phản Hồi

Sản phẩm hay nhất là sản phẩm được xây dựa trên phản hồi thực tế.

Cách thu thập:
- Nhờ bạn bè/đồng nghiệp dùng thử
- Đặt form feedback trên trang
- Xem analytics: người dùng click vào đâu nhiều nhất? Bỏ đi ở trang nào?

### 3. Iterate — Lặp Lại Quy Trình

Từ feedback, bạn lại quay về quy trình quen thuộc:

```
Feedback -> Brainstorm -> Task Breakdown -> TDD -> Implement -> Review -> Ship
```

Mỗi vòng lặp, sản phẩm tốt hơn một chút. Đó là cách mọi sản phẩm thành công được xây dựng.

## Tổng Kết Hành Trình

Hãy nhìn lại chặng đường bạn đã đi:

| Trước | Sau |
|-------|-----|
| "Tôi không biết gì về code" | "Tôi biết cách dùng AI để build sản phẩm" |
| "Error message = vô nghĩa" | "Error message = tin nhắn cần copy cho Claude" |
| "Deploy = phép thuật" | "Deploy = vài bước trên Vercel/Railway" |
| "Chất lượng = may rủi" | "Chất lượng = quy trình Superpowers" |

### Ba vai trò trong dự án của bạn:

```
BAN = Product Owner
  => Ra y tuong, quyet dinh, duyet ket qua

CLAUDE = Developer
  => Viet code, sua loi, implement tinh nang

STARTER KIT = QA Team
  => Kiem tra chat luong, dam bao tieu chuan
```

Bạn KHÔNG cần trở thành developer. Bạn cần trở thành **Product Owner giỏi** — người biết mình muốn gì và biết cách truyền đạt cho AI.

````bash
# Final quality check truoc khi ship

cat << 'EOF'
=== FINAL QUALITY CHECK WORKFLOW ===

# Buoc 1: Quality Audit
# Trong Claude Code: /quality-audit
# Ket qua: >= 85 diem? -> Tiep tuc

# Buoc 2: Business Gap Check
# Trong Claude Code: /business-gap-check  
# Ket qua: 0 P1 gaps? -> Tiep tuc

# Buoc 3: Kiem tra CI
$ git push
# Doi CI chay xong, tat ca test pass? -> Tiep tuc

# Buoc 4: Deploy
$ vercel --prod    # hoac railway up

# Buoc 5: Verify tren production
# Mo link production, kiem tra bang tay:
#   - Trang chu load dung?
#   - Cac nut click duoc?
#   - Form gui duoc?
#   - Mobile hien thi dung?

# Buoc 6: Xong! San pham da live!
echo "San pham da ship thanh cong!"
EOF
````

## Lời Khuyên Cuối Cùng

### 1. Bắt đầu nhỏ, làm dần
Đừng cố xây Facebook ngay từ đầu. Bắt đầu với 1 trang landing page. Thêm tính năng dần dần. MVP (Minimum Viable Product) = phiên bản nhỏ nhất mà vẫn hữu ích.

### 2. Đọc error = giải quyết 80% vấn đề
Phần lớn lỗi đều có thể sửa bằng cách copy error message cho Claude. Đừng sợ lỗi — hãy xem nó như phản hồi từ máy tính.

### 3. Quality > Speed
Chạy nhanh nhưng sản phẩm lỗi = mất thời gian sửa gấp đôi. Làm chậm nhưng có quality check = tiết kiệm thời gian về lâu dài.

### 4. Trust the process
Quy trình Superpowers (Brainstorm → Task Breakdown → TDD → Implement → Review) nghe có vẻ nhiều bước, nhưng mỗi bước đều có lý do. Tin vào quy trình, kết quả sẽ đến.

### 5. Bạn không đơn độc
Claude Code luôn ở đó. Starter Kit luôn ở đó. Cộng đồng developer luôn sẵn sàng giúp đỡ. Đừng ngại hỏi.

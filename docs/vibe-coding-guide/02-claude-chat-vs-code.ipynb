# 02 — Claude: Chat vs Code

Claude có nhiều "hình dạng" khác nhau. Trong notebook này, bạn sẽ hiểu:
- Claude trên browser (claude.ai) — dùng để chat
- Claude Code (CLI) — dùng để viết code, quản lý dự án

## 1. Claude trên Browser (claude.ai)

### Ví von: Nói chuyện với chuyên gia qua điện thoại

Khi bạn mở [claude.ai](https://claude.ai) và gõ câu hỏi, bạn đang:
- Nói chuyện 1-1 với AI
- Qua giao diện web (giống nhắn tin Messenger/Zalo)
- AI trả lời dựa trên kiến thức nó đã học

### Claude Chat làm được gì?

| Việc | Ví dụ cụ thể |
|------|-------------|
| Hỏi đáp | "Giải thích blockchain cho người mới" |
| Viết văn bản | "Viết email xin nghỉ phép" |
| Tóm tắt | Copy-paste tài liệu dài, nhờ tóm |
| Dịch thuật | "Dịch đoạn này sang tiếng Anh" |
| Brainstorm | "Cho tôi 10 ý tưởng tên sản phẩm" |
| Phân tích | Upload file CSV, nhờ phân tích |

### Giới hạn quan trọng

- **Không truy cập file trên máy bạn** (trừ khi bạn upload lên)
- **Không chạy được lệnh** (không thể tạo project, cài phần mềm)
- **Mỗi cuộc hội thoại là độc lập** — bắt đầu chat mới = AI quên hết
- **Không có memory dài hạn** giữa các phiên

````bash
# Ví dụ: Những gì bạn có thể hỏi Claude Chat
# (Đây là mô phỏng — thực tế bạn gõ trên claude.ai)

echo "=== Ví dụ prompt tốt cho Claude Chat ==="
echo ""
echo '1. "Tôi đang xây dựng app quản lý lớp học. Hãy giúp tôi liệt kê'
echo '    các tính năng cần có, ưu tiên theo mức độ quan trọng."'
echo ""
echo '2. "Giải thích REST API là gì, dùng ví von cho người không biết code."'
echo ""
echo '3. "Tôi có bảng dữ liệu học sinh (tên, điểm TB, lớp). Hãy phân tích'
echo '    và cho biết xu hướng." (rồi paste data vào)'
echo ""
echo '4. "Viết email tiếng Anh mời đối tác hợp tác, tone chuyên nghiệp nhưng'
echo '    thân thiện."'
````

## 2. Claude Code (CLI Tool)

### Ví von: Thuê chuyên gia đến TẬN NHÀ làm việc

Nếu Claude Chat là "gọi điện hỏi chuyên gia", thì Claude Code là **mời chuyên gia đến tận phòng làm việc của bạn**:

- Chuyên gia **nhìn thấy** tất cả tài liệu trên bàn bạn (đọc file code)
- Chuyên gia **tự tay làm** việc (chạy lệnh, tạo file, sửa code)
- Chuyên gia **nhớ** những gì đã làm lần trước (memory)
- Chuyên gia có **kỹ năng chuyên biệt** (skills)
- Chuyên gia **tự động làm một số việc** khi có sự kiện (hooks)

### Claude Code chạy ở đâu?

Claude Code chạy trong **Terminal** (cửa sổ dòng lệnh) — thường nằm bên trong VS Code.

```
┌──────────────────────────────────────────┐
│  VS Code                                  │
│  ┌────────┐  ┌────────────────────────┐  │
│  │ Files   │  │  Code Editor           │  │
│  │ folder/ │  │  (bạn xem code ở đây)  │  │
│  │  ├ src/ │  │                        │  │
│  │  ├ ...  │  │                        │  │
│  │         │  │                        │  │
│  └────────┘  └────────────────────────┘  │
│  ┌──────────────────────────────────────┐ │
│  │  Terminal (Claude Code chạy ở đây!)  │ │
│  │  $ claude                             │ │
│  │  > Xin chào! Tôi là Claude Code...   │ │
│  └──────────────────────────────────────┘ │
└──────────────────────────────────────────┘
```

````bash
# Cách cài Claude Code (chỉ cần chạy 1 lần)
# Bạn cần Node.js đã cài sẵn trên máy

echo "=== Cài đặt Claude Code ==="
echo ""
echo "Bước 1: Cài Node.js (nếu chưa có)"
echo "  Tải từ: https://nodejs.org/"
echo ""
echo "Bước 2: Cài Claude Code"
echo "  npm install -g @anthropic-ai/claude-code"
echo ""
echo "Bước 3: Chạy Claude Code"
echo "  cd duong-dan-den-project/"
echo "  claude"
echo ""
echo "(Lần đầu chạy sẽ yêu cầu đăng nhập Anthropic)"
````

### Claude Code làm được gì?

| Khả năng | Ví dụ | Claude Chat có? |
|---------|-------|----------------|
| Đọc file trong project | Đọc toàn bộ source code | Không |
| Chạy lệnh | `git commit`, `npm test`, `docker up` | Không |
| Tạo/sửa file | Viết code, tạo file mới | Không (chỉ show code) |
| Memory | Nhớ context giữa các phiên | Không |
| Skills | Kỹ năng chuyên biệt (audit, review, deploy) | Không |
| Hooks | Tự động chạy khi commit, push, etc. | Không |
| Multi-file edit | Sửa 10 file cùng lúc, đúng logic | Không |

### Ví dụ thực tế: Bạn nói, Claude Code làm

```
Bạn: "Tạo trang đăng nhập cho website, có form email và password,
       hiển thị lỗi khi nhập sai."

Claude Code sẽ:
1. Đọc cấu trúc project hiện tại
2. Tìm hiểu tech stack (React? Vue? Angular?)
3. Tạo file login.tsx
4. Viết code form đăng nhập
5. Thêm validation (kiểm tra email hợp lệ)
6. Kết nối với API backend
7. Chạy test để kiểm tra
```

Tất cả chỉ từ **1 câu mô tả** của bạn.

### Memory, Skills, Hooks — Ba siêu năng lực của Claude Code

#### Memory (Trí nhớ dài hạn)

Claude Code có file `CLAUDE.md` — giống như **sổ tay công việc**:
- Ghi lại quy tắc dự án ("code style dùng 2 space, không dùng tab")
- Ghi lại kiến trúc ("backend dùng Java Spring Boot, frontend dùng React")
- Ghi lại bài học kinh nghiệm ("Bug X xảy ra vì Y, đã fix bằng Z")

Mỗi lần bắt đầu phiên mới, Claude Code **đọc sổ tay này trước** — nên không cần bạn giải thích lại từ đầu.

#### Skills (Kỹ năng chuyên biệt)

Skills = bộ hướng dẫn cho Claude Code biết cách làm việc cụ thể:
- `/audit` — kiểm tra chất lượng code (chấm điểm /100)
- `/review` — review code trước khi merge
- `/deploy` — hướng dẫn deploy lên server

Giống như **bộ SOP (quy trình chuẩn)** trong nhà máy — đảm bảo AI làm đúng quy trình mỗi lần.

#### Hooks (Hành động tự động)

Hooks = "khi X xảy ra, tự động làm Y":
- Khi commit code → tự động chạy test
- Khi tạo PR → tự động kiểm tra chất lượng
- Khi sửa file → tự động format code

Giống như **dây chuyền kiểm tra tự động** trong nhà máy.

````bash
# Demo: Các lệnh Claude Code phổ biến

echo "=== Các lệnh Claude Code bạn sẽ dùng ==="
echo ""
echo "# Khởi động Claude Code (trong folder project)"
echo "claude"
echo ""
echo "# Hỏi Claude Code một câu nhanh (không cần vào interactive mode)"
echo 'claude "Giải thích file README.md trong project này"'
echo ""
echo "# Xem trạng thái memory"
echo "claude /memory"
echo ""
echo "# Dùng skill"
echo "claude /audit"
echo ""
echo "# Resume phiên làm việc trước"
echo "claude --resume"
````

## 3. So sánh: Claude Chat vs Claude Code

| Tiêu chí | Claude Chat (browser) | Claude Code (terminal) |
|---------|----------------------|----------------------|
| **Cách dùng** | Gõ trên claude.ai | Gõ trong terminal VS Code |
| **Ví von** | Gọi điện hỏi chuyên gia | Mời chuyên gia đến nhà |
| **Đọc file** | Chỉ file bạn upload | Toàn bộ project |
| **Chạy lệnh** | Không | Có (git, test, build...) |
| **Tạo/sửa file** | Không (chỉ show code) | Trực tiếp sửa file |
| **Memory** | Không (mỗi chat riêng) | Có (CLAUDE.md) |
| **Skills** | Không | Có (audit, review...) |
| **Hooks** | Không | Có (auto-test, auto-format) |
| **Phù hợp cho** | Hỏi đáp, viết văn, brainstorm | Viết code, quản lý dự án |
| **Cần cài đặt** | Không (dùng browser) | Cần cài Node.js + CLI |
| **Giá** | Free tier có | Cần Anthropic subscription |

### Khi nào dùng cái nào?

- **Hỏi nhanh, viết email, tóm tắt** → Claude Chat
- **Viết code, tạo feature, fix bug, deploy** → Claude Code
- **Vibe coding** → Claude Code (bắt buộc)

## 4. MCP — Model Context Protocol

### Ví von: Adapter (bộ chuyển đổi) cho AI

Bạn biết các adapter sạc điện thoại không? Ổ cắm ở Việt Nam khác Nhật Bản khác Mỹ — cần adapter để "nói chuyện" được.

**MCP giống vậy** — nó là bộ chuyển đổi để AI "nói chuyện" được với các công cụ khác:

```
┌───────────┐     MCP      ┌──────────────┐
│  Claude   │ ◄──────────► │  Gmail       │
│  Code     │ ◄──────────► │  Calendar    │
│           │ ◄──────────► │  Slack       │
│           │ ◄──────────► │  Database    │
│           │ ◄──────────► │  GitHub      │
└───────────┘              └──────────────┘
```

### Ví dụ thực tế

Với MCP, bạn có thể nói với Claude Code:
- *"Đọc email mới nhất từ khách hàng và tóm tắt"* (kết nối Gmail)
- *"Tạo cuộc họp lúc 3 giờ chiều mai"* (kết nối Google Calendar)
- *"Gửi thông báo deploy thành công lên Slack"* (kết nối Slack)

### Bạn cần biết gì?

- MCP tồn tại và ngày càng nhiều công cụ hỗ trợ
- Bạn **không cần** tự cài MCP ngay lúc này
- Khi cần kết nối AI với tool cụ thể, hỏi Claude Code cách cài MCP cho tool đó

````bash
# Ví dụ: Cấu hình MCP trong Claude Code
# (Bạn chưa cần làm ngay — chỉ để biết nó trông như thế nào)

echo "=== File cấu hình MCP (settings.json) ==="
echo ""
cat << 'EXAMPLE'
{
  "mcpServers": {
    "gmail": {
      "command": "npx",
      "args": ["@anthropic/mcp-gmail"]
    },
    "github": {
      "command": "npx",
      "args": ["@anthropic/mcp-github"]
    }
  }
}
EXAMPLE
echo ""
echo "=> Khi cần, Claude Code sẽ hướng dẫn bạn cài MCP phù hợp."
````

## 5. API — Dành cho Developer

### Ví von: Cửa sau dành cho robot

Khi bạn dùng claude.ai, bạn đang dùng **cửa chính** — giao diện web dành cho con người.

**API** (Application Programming Interface) là **cửa sau dành cho phần mềm** — để ứng dụng khác "nói chuyện" với Claude mà không cần giao diện.

Ví dụ:
- App chatbot trên website bạn → gọi Claude API để trả lời khách hàng
- Script tự động → gọi Claude API để phân tích 1000 đánh giá sản phẩm
- Ứng dụng mobile → gọi Claude API để dịch thuật real-time

### Bạn có cần dùng API không?

**Phần lớn là KHÔNG** khi vibe coding. Claude Code đã xử lý hết phần kỹ thuật cho bạn. Chỉ khi bạn muốn tích hợp AI vào sản phẩm của mình (ví dụ: chatbot hỗ trợ sinh viên), Claude Code sẽ giúp bạn viết phần gọi API.

````bash
# Ví dụ gọi API (chỉ minh họa — bạn không cần chạy lệnh này)

echo "=== Gọi Claude API trông như thế nào ==="
echo ""
echo "Đây là lệnh gọi API (dành cho developer):"
echo ""
cat << 'EXAMPLE'
curl https://api.anthropic.com/v1/messages \
  -H "x-api-key: YOUR_API_KEY" \
  -H "content-type: application/json" \
  -d '{
    "model": "claude-sonnet-4-20250514",
    "messages": [{"role": "user", "content": "Xin chào!"}]
  }'
EXAMPLE
echo ""
echo "=> Bạn không cần hiểu lệnh trên. Claude Code sẽ viết code này cho bạn"
echo "   khi cần tích hợp AI vào sản phẩm."
````

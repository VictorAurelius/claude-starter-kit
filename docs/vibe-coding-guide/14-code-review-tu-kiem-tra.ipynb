# 14. Code Review — Tự Kiểm Tra

**Part 3: Superpowers — Quy trình làm việc** | Bước 4-5 trong 5 bước Superpowers

---

> *"Không ai gửi hàng mà không kiểm tra trước."*

Sau khi AI viết xong code (bước 4: Implementation), bước cuối cùng là **tự kiểm tra** (bước 5: Code Review).  
Kit yêu cầu AI **tự review code của chính nó** — giống như nhân viên tự kiểm hàng trước khi giao cho khách.

## Code Review là gì? "Kiểm hàng" trước khi ship

Hãy tưởng tượng bạn bán hàng online:

| Bán hàng | Phần mềm |
|---------|----------|
| Đóng gói xong | Code xong |
| Kiểm tra: đúng sản phẩm? Đủ số lượng? Không hỏng? | Review: đúng yêu cầu? Xử lý lỗi? Không bug? |
| Phát hiện lỗi → sửa trước khi ship | Phát hiện lỗi → sửa trước khi deploy |
| Ship cho khách → khách hài lòng | Deploy → người dùng hài lòng |

**Nếu không kiểm hàng:**
- Gửi nhầm size → khách phàn nàn → tốn tiền đổi trả
- Ship sản phẩm hỏng → mất khách → mất uy tín

**Nếu không review code:**
- Tính năng không đúng yêu cầu → phải làm lại
- Bug lên production → người dùng gặp lỗi → mất niềm tin

## 2 giai đoạn kiểm tra

Kit yêu cầu AI chạy **2 giai đoạn** review, mỗi giai đoạn có checklist riêng:

```
Stage 1: Specification Compliance     Stage 2: Code Quality
"Có đúng yêu cầu không?"             "Code có tốt không?"
─────────────────────────             ────────────────────
BLOCKING — fail = dừng lại            GRADED — chấm điểm
15-20 phút                            20-30 phút
```

Stage 1 là **bắt buộc**. Nếu fail → dừng lại, sửa trước.  
Stage 2 là **chấm điểm**. Tùy mức độ nghiêm trọng mà sửa hay bỏ qua.

## Stage 1: Specification Compliance — "Có đúng yêu cầu không?"

Thời gian: 15-20 phút | Tính chất: **BLOCKING** (fail = không được tiếp tục)

AI tự kiểm tra 5 câu hỏi:

### 1. Code có đúng yêu cầu không?
- Bạn yêu cầu "trang đăng nhập" → có trang đăng nhập không?
- Bạn nói "cần xác nhận email" → có xác nhận email không?
- Tất cả yêu cầu đã được implement chưa?

### 2. Có xử lý trường hợp đặc biệt không?
- Nếu người dùng nhập rỗng → có thông báo lỗi không?
- Nếu server chết → có thông báo lỗi không?
- Nếu 2 người cùng thao tác → có xung đột không?

### 3. File đặt đúng chỗ không?
- Theo cấu trúc thư mục chuẩn
- Không đặt lung tung

### 4. API đúng contract không?
- Giao tiếp giữa frontend và backend đúng quy ước

### 5. Test chứng minh yêu cầu đúng không?
- Mỗi yêu cầu có ít nhất 1 test
- Tất cả test PASS

### Kết quả Stage 1:

**PASS:** Tất cả 5 câu hỏi đều OK → tiếp tục Stage 2.

**FAIL:** Ít nhất 1 câu hỏi không OK → **DỪNG LẠI.**

```
Stage 1 Result: FAIL
Reason: Yêu cầu "xác nhận email" chưa được implement.
Action: Phải thêm tính năng xác nhận email trước khi tiếp tục.
```

AI sẽ tự sửa, rồi chạy lại Stage 1 cho đến khi PASS.

## Stage 2: Code Quality — "Code có tốt không?"

Thời gian: 20-30 phút | Tính chất: **GRADED** (chấm điểm theo mức độ)

AI phân loại vấn đề thành 3 mức:

### Mức Critical (Nghiêm trọng) — BẮT BUỘC sửa

Vấn đề gây hậu quả nghiêm trọng:

| Loại | Ví dụ dễ hiểu |
|------|----------------|
| **Bảo mật** | Mật khẩu lưu dạng chữ thường (không mã hóa) — ai cũng đọc được |
| **Mất dữ liệu** | Xóa nhầm dữ liệu mà không thể khôi phục |
| **Breaking change** | Tính năng cũ bị hỏng sau khi thêm tính năng mới |

**Tương đương đời thực:** Gửi hàng nhưng hàng bị vỡ, bị mất, hoặc gửi nhầm địa chỉ.

### Mức Major (Quan trọng) — NÊN sửa

Vấn đề chưa gây hậu quả ngay, nhưng sẽ phiền phức về sau:

| Loại | Ví dụ dễ hiểu |
|------|----------------|
| **Hiệu năng** | Trang tải mất 10 giây (nên dưới 2 giây) |
| **Test coverage thấp** | Chỉ 30% code có test (nên trên 80%) |
| **Logic phức tạp** | Đoạn code quá khó hiểu, sau này khó sửa |

**Tương đương đời thực:** Đóng gói hơi lỏng — hàng không hỏng ngay, nhưng lâu dài có thể hỏng khi vận chuyển.

### Mức Minor (Nhẹ) — TÙY, có thể bỏ qua

Vấn đề nhỏ, không ảnh hưởng chức năng:

| Loại | Ví dụ dễ hiểu |
|------|----------------|
| **Đặt tên** | Biến tên `x` thay vì `userName` — khó đọc nhưng vẫn chạy |
| **Trùng lặp** | Cùng 1 đoạn code viết 2 lần — nên gộp lại |
| **Style** | Thụt đầu dòng không đều — nhìn xấu nhưng vẫn hoạt động |

**Tương đương đời thực:** Hộp hàng hơi méo — hàng bên trong vẫn OK, chỉ nhìn không đẹp.

## Cho người không biết code: Bạn cần biết gì?

Bạn **không cần đọc code** để hiểu kết quả review. AI sẽ báo cáo dạng:

```
=== CODE REVIEW REPORT ===

Stage 1: Specification Compliance
  Result: PASS
  All requirements implemented correctly.

Stage 2: Code Quality
  Critical issues: 0  (bắt buộc sửa)
  Major issues:    1  (nên sửa)
  Minor issues:    3  (tùy)

  Major #1: Trang tải chậm khi có >100 sản phẩm
    → Fix: Thêm phân trang, chỉ tải 20 sản phẩm/lần
    → Status: Fixed

  Minor #1-3: Naming, style issues
    → Skipped (không ảnh hưởng chức năng)
```

### Bạn chỉ cần:

| Kết quả | Bạn cần làm gì |
|---------|----------------|
| Critical = 0, Stage 1 PASS | Không cần làm gì. Yên tâm. |
| Critical > 0 | Nói AI: "Sửa các critical issues" |
| Major > 0 | Tùy bạn: sửa ngay hoặc sửa sau |
| Minor > 0 | Bỏ qua cũng được |

## Review = Phiếu kiểm tra chất lượng

Hãy nghĩ review như **phiếu kiểm tra chất lượng** khi mua xe:

```
PHIẾU KIỂM TRA CHẤT LƯỢNG
──────────────────────────
[x] Động cơ hoạt động         → Chức năng chính OK
[x] Thắng hoạt động           → Xử lý lỗi OK
[x] Đèn sáng                  → Giao diện OK
[x] Khóa cửa hoạt động        → Bảo mật OK
[ ] Vết xước nhỏ trên sơn     → Minor issue
──────────────────────────
KẾT LUẬN: ĐẠT (1 minor)
```

Bạn nhận phiếu này, nhìn qua, và quyết định: "OK, giao hàng đi" hoặc "Sửa vết xước trước đã".

## Cách chạy Code Review

Review **tự động chạy** khi bạn dùng `/start-pr`. Nhưng bạn cũng có thể yêu cầu thủ công:

````bash
# Cách 1: Tự động (khuyến nghị)
# /start-pr Tính năng quản lý sản phẩm
# → AI brainstorm → breakdown → TDD → implement → REVIEW (tự động)

# Cách 2: Yêu cầu thủ công
# "Hãy review code vừa viết. Chạy đầy đủ 2 stages."

# Cách 3: Dùng /check-pr (kiểm tra PR đã tạo)
# /check-pr 42
# → AI kiểm tra PR #42 và báo cáo kết quả
````

## Bước 4: Implementation — AI tự viết code

Giữa TDD (bước 3) và Review (bước 5) là **Implementation** — bước AI viết code.

Bước này bạn **không cần can thiệp nhiều**:

1. AI đã có kế hoạch từ brainstorm (bước 1)
2. AI đã có danh sách tasks (bước 2)
3. AI đã có test để kiểm tra (bước 3)
4. AI **viết code theo tasks**, chạy test sau mỗi task
5. Khi tất cả test PASS → chuyển sang review (bước 5)

**Bạn có thể quan sát** AI làm việc, hoặc đi pha cà phê — AI sẽ gọi bạn khi cần quyết định.

## Commit thường xuyên — Lưu tiến độ

Trong quá trình implement, AI sẽ **commit** (lưu) code sau mỗi task hoặc mỗi cụm tasks.

Commit giống như **bấm nút Save** trong game:
- Save thường xuyên → nếu chết, quay lại checkpoint gần nhất
- Không save → chết = mất hết tiến độ từ đầu

```
Commit 1: "feat: add login form UI"           ← Task 1 xong
Commit 2: "feat: add email validation"         ← Task 2 xong
Commit 3: "feat: connect login API"            ← Task 3 xong
Commit 4: "feat: add error handling"           ← Task 4 xong
Commit 5: "test: add login page tests"         ← Task 5 xong
```

Nếu task 4 sai → quay lại commit 3 → chỉ mất task 4, không mất cả 5 tasks.

## Quy trình đầy đủ: Từ ý tưởng → Code review xong

```
Bạn: "Tạo trang đăng nhập"
        |
        v
  1. BRAINSTORM (5-10 phút)
     AI hỏi: "Đăng nhập bằng gì? Quên mật khẩu?"
     Bạn trả lời → AI hiểu rõ
        |
        v
  2. TASK BREAKDOWN (5 phút)
     AI chia: 5 tasks, ~21 phút
     Bạn: "OK, bắt đầu đi"
        |
        v
  3. TDD — VIẾT TEST (5 phút)
     AI viết 12 test cases → chạy → 12 FAIL (chưa có code)
        |
        v
  4. IMPLEMENTATION (15 phút)
     AI viết code task 1 → test pass → commit
     AI viết code task 2 → test pass → commit
     ... tiếp tục đến task 5
     12/12 test PASS
        |
        v
  5. CODE REVIEW (20-30 phút)
     Stage 1: PASS ✓
     Stage 2: 0 Critical, 1 Major (đã sửa), 2 Minor (bỏ qua)
        |
        v
  DONE! Tạo PR → merge → deploy
```

## Tóm tắt

| Điều cần nhớ | Chi tiết |
|--------------|----------|
| **Code Review là gì** | AI tự kiểm tra code trước khi "giao hàng" |
| **2 giai đoạn** | Stage 1: Đúng yêu cầu? (BLOCKING) / Stage 2: Code tốt? (GRADED) |
| **3 mức issue** | Critical (bắt buộc sửa), Major (nên sửa), Minor (tùy) |
| **Bạn cần làm gì** | Đọc báo cáo, quyết định sửa hay bỏ qua Minor |
| **Implementation** | AI viết code theo tasks, commit sau mỗi task |
| **Tự động** | Dùng /start-pr → toàn bộ 5 bước chạy tự động |

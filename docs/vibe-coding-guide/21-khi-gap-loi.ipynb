# 21. Khi Gặp Lỗi — Đừng Hoảng, Có Claude Rồi

**Phần 5: Thực hành từ đầu đến cuối**

---

Gặp lỗi khi làm dự án là **100% chắc chắn xảy ra**. Không có ngoại lệ. Ngay cả những kỹ sư nhiều năm kinh nghiệm cũng gặp lỗi hàng ngày.

Giống như nấu ăn — lần đầu nấu phở, bạn có thể cho quá nhiều muối, nước dùng bị đục, hoặc bánh phở bị nát. Chuyện bình thường! Quan trọng là bạn biết **nếm thử** (phát hiện lỗi) và **biết cách sửa** (hoặc nhờ ai đó giúp).

Trong bài này, bạn sẽ học cách đối phó với lỗi mà KHÔNG cần hiểu code.

## Error Message Là Gì?

Error message = **tin nhắn từ máy tính nói cho bạn biết gì đang sai**.

Nó trông đáng sợ (chữ đỏ, nhiều dòng, toàn thuật ngữ), nhưng thực ra nó đang cố **giúp bạn**.

### Ví dụ một error message:

```
Error: Cannot find module 'react'
  at Function.Module._resolveFilename (node:internal/modules/cjs/loader:995:15)
  at /home/user/my-project/src/App.tsx:3:1
```

### Cách đọc (dành cho người không biết code):

| Phần | Nghĩa | Quan trọng? |
|------|--------|------------|
| `Error: Cannot find module 'react'` | Lỗi gì: không tìm thấy thư viện 'react' | **Rất quan trọng** |
| `at /home/user/.../App.tsx:3:1` | Lỗi ở đâu: file App.tsx, dòng 3 | Quan trọng |
| `at Function.Module._resolveFilename...` | Chi tiết kỹ thuật | Không cần đọc |

**Quy tắc:** Dòng đầu tiên thường nói **lỗi gì**. Dòng tiếp theo nói **ở đâu**. Phần còn lại bạn không cần hiểu.

````bash
# Ví dụ: tạo một lỗi thật để bạn thấy error message trông như nào

# Thử chạy một file không tồn tại
node khong-co-file-nay.js 2>&1 || true

echo ""
echo "=> Thay chua? Error message noi: khong tim thay file."
echo "   Ban KHONG can hieu chi tiet — chi can copy paste cho Claude."
````

````bash
# Một lỗi khác: port đang bị chiếm

cat << 'EOF'
=== VI DU LOI: PORT DA BI CHIEM ===

Error: listen EADDRINUSE: address already in use :::3000

Dich: Cong 3000 dang bi chuong trinh khac dung roi.
Giai phap: tat chuong trinh kia, hoac doi sang cong khac.

Ban noi voi Claude: "Toi gap loi port 3000 bi chiem, giup toi sua"
Claude se: tim process dang chiem port -> tat no -> chay lai
EOF
````

## Quy Trình 3 Bước Khi Gặp Lỗi

Bất kể lỗi gì, bạn luôn làm 3 bước này:

### Bước 1: COPY error message
Bôi đen toàn bộ dòng chữ đỏ (hoặc thông báo lỗi) → Ctrl+C

### Bước 2: PASTE cho Claude
Mở Claude Code, gõ:
```
Tôi gặp lỗi này, hãy sửa giúp:
[paste error message ở đây]
```

### Bước 3: Để Claude phân tích và sửa
Claude sẽ:
- Đọc error message
- Tìm nguyên nhân
- Sửa code/config
- Chạy lại để kiểm tra

Nếu vẫn lỗi → lặp lại từ Bước 1 với error mới.

````bash
# Demo: workflow gặp lỗi → nhờ Claude sửa

cat << 'EOF'
=== KICH BAN THUC TE ===

1. Ban chay: npm run dev
   => LOI: "Module not found: Can't resolve './Header'"

2. Ban copy loi, paste cho Claude:
   "Toi gap loi nay khi chay npm run dev:
    Module not found: Can't resolve './Header'"

3. Claude phan tich:
   "Loi nay nghia la file Header.tsx chua duoc tao.
    Toi se tao file nay cho ban."
   => Claude tao file Header.tsx

4. Ban chay lai: npm run dev
   => THANH CONG! Trang hien thi binh thuong.

=> Tong thoi gian: 30 giay. Ban khong can hieu loi gi.
EOF
````

## Systematic Debugging — Claude Làm, Bạn Theo Dõi

Với lỗi phức tạp hơn, Claude sẽ dùng quy trình **4 giai đoạn** để tìm và sửa lỗi. Bạn không cần hiểu chi tiết — chỉ cần biết Claude đang ở giai đoạn nào.

| Giai đoạn | Claude làm gì | Bạn thấy gì |
|-----------|--------------|------------|
| 1. Reproduce | Tái tạo lỗi — chạy lại để thấy lỗi lần nữa | Claude chạy command, lỗi xuất hiện lại |
| 2. Trace | Lần theo dấu vết — lỗi bắt đầu từ đâu? | Claude đọc nhiều file, tìm manh mối |
| 3. Root Cause | Xác định gốc rễ — TẠI SAO lỗi | Claude giải thích nguyên nhân |
| 4. Fix | Sửa + kiểm tra lại | Claude sửa code, chạy test, xác nhận OK |

Giống thợ sửa ống nước: nghe bạn mô tả (nước rò), đến xem (reproduce), tìm chỗ rò (trace), hiểu tại sao rò (root cause), vá lại (fix).

````bash
# Ví dụ Claude debug theo 4 giai đoạn:

cat << 'EOF'
=== SYSTEMATIC DEBUGGING (vi du) ===

LOI: "Trang san pham hien thi trang (blank page)"

[Giai doan 1: Reproduce]
Claude: "Toi se mo trang /products de xem..."
=> Dung that, trang trang.

[Giai doan 2: Trace]
Claude: "Toi kiem tra console log... thay loi:
  TypeError: products.map is not a function
  Loi o file ProductList.tsx, dong 15"

[Giai doan 3: Root Cause]
Claude: "Nguyen nhan: API tra ve object {data: [...]} 
  nhung code dang goi .map() tren object thay vi array.
  Can doi thanh products.data.map()"

[Giai doan 4: Fix]
Claude: "Da sua ProductList.tsx dong 15.
  Chay lai... trang hien thi 4 san pham. Da OK!"

=> Tong thoi gian: 2 phut.
EOF
````

## Các Tình Huống Hay Gặp (và Cách Nói Với Claude)

Đây là những lỗi phổ biến nhất mà người mới hay gặp. Bạn không cần nhớ cách sửa — chỉ cần biết **nói gì** với Claude.

````bash
cat << 'EOF'
=== TINH HUONG 1: TRANG TRANG (BLANK PAGE) ===

Trieu chung: Mo browser, thay trang trang tinh, khong co gi
Nguyen nhan thuong gap: Loi JavaScript trong frontend

Ban noi voi Claude:
"Mo trang http://localhost:3000 nhung chi thay trang trang.
 Khong co noi dung gi. Giup toi kiem tra."

Claude se: Kiem tra console errors -> tim loi -> sua -> reload
EOF
````

````bash
cat << 'EOF'
=== TINH HUONG 2: KHONG KET NOI DUOC ===

Trieu chung: Browser bao "This site can't be reached" 
             hoac "Connection refused"
Nguyen nhan thuong gap: Server chua chay

Ban noi voi Claude:
"Toi mo localhost:3000 nhung bao 'Connection refused'.
 Giup toi kiem tra server co dang chay khong."

Claude se: Kiem tra server -> khoi dong lai neu can -> xac nhan OK
EOF
````

````bash
cat << 'EOF'
=== TINH HUONG 3: PERMISSION DENIED ===

Trieu chung: Chay lenh bao "Permission denied" hoac "Access denied"
Nguyen nhan thuong gap: Khong co quyen truy cap file/folder

Ban noi voi Claude:
"Toi chay [lenh] nhung bao Permission denied. Giup toi sua."

Claude se: Kiem tra quyen -> cap quyen phu hop -> chay lai
EOF
````

````bash
cat << 'EOF'
=== TINH HUONG 4: PORT ALREADY IN USE ===

Trieu chung: Chay server bao "Port 3000 already in use"
Nguyen nhan thuong gap: Co chuong trinh khac dang dung port do

Ban noi voi Claude:
"Chay npm run dev bao port 3000 bi chiem. Giup toi giai quyet."

Claude se: Tim process dang chiem port -> tat no -> chay lai
EOF
````

````bash
# Lệnh hữu ích: kiểm tra port đang bị chiếm bởi ai

# Linux/Mac:
lsof -i :3000 2>/dev/null || echo "Port 3000 dang trong (khong ai dung)"

echo ""
echo "Neu co chuong trinh dang chiem, ban se thay PID (so ID) cua no."
echo "Claude se giup ban tat chuong trinh do."
````

## Khi Nào Cần DỪNG LẠI và Nhờ Developer

Claude rất giỏi, nhưng có những lúc bạn nên nhờ người thật:

### Dừng lại khi:

**1. Claude sửa 3 lần mà vẫn lỗi**
Nếu cùng một lỗi, Claude thử sửa 3 lần mà vẫn không giải quyết được, đây có thể là vấn đề phức tạp cần con người can thiệp.

**2. Liên quan đến dữ liệu quan trọng**
Nếu error message nhắc đến "database", "data loss", "corruption" — DỪNG LẠI. Dữ liệu bị mất không lấy lại được.

**3. Liên quan đến bảo mật**
Nếu thấy "security", "vulnerability", "exposed credentials", "API key leaked" — DỪNG LẠI và nhờ chuyên gia.

**4. Error message có từ "đáng sợ"**
Các từ như: `fatal`, `critical`, `data loss`, `unrecoverable`, `disk full` — đừng tự xử lý.

### Cách nhờ developer:
Copy toàn bộ error message + mô tả bạn đang làm gì + screenshot (nếu có) → gửi cho developer. Thông tin càng đầy đủ, developer sửa càng nhanh.

````bash
cat << 'EOF'
=== TEMPLATE NHO DEVELOPER ===

Chao anh/chi,

Toi dang lam [mo ta tinh nang] va gap loi sau:

1. Toi dang lam gi: [vd: chay npm run dev de test landing page]
2. Loi xuat hien: [paste error message]
3. Da thu: [vd: Claude da thu sua 3 lan nhung van loi]
4. Screenshot: [dinh kem neu co]

Cam on anh/chi!
EOF
````

## Mẹo Giảm Lỗi Từ Đầu

Phòng bệnh hơn chữa bệnh:

1. **Commit thường xuyên** — mỗi khi có gì hoạt động, commit ngay. Nếu lỗi, bạn quay lại bản commit trước.

2. **Thay đổi nhỏ** — đừng đổi 10 thứ cùng lúc. Đổi 1 thứ → test → OK → đổi thứ tiếp.

3. **Đọc cảnh báo** — Warning (cảnh báo vàng) không phải lỗi, nhưng nếu bỏ qua nhiều warning, chúng có thể biến thành lỗi.

4. **Dùng TDD** — viết test trước = bạn biết ngay khi có gì hỏng.

````bash
# Ví dụ: commit thường xuyên để "save game"

cat << 'EOF'
=== COMMIT THUONG XUYEN = SAVE GAME ===

Giong nhu choi game — ban save thuong xuyen de:
- Neu "chet" (gap loi), quay lai diem save gan nhat
- Khong mat tien do (khong mat code da viet)

$ git add .
$ git commit -m "feat: header component done"
  => SAVE!

$ git add .
$ git commit -m "feat: hero section done"
  => SAVE!

[gap loi o products section]
$ git checkout -- .   # Quay lai save gan nhat
  => An toan! Chi mat phan products, con header va hero van con.
EOF
````

## Tóm Tắt

| Tình huống | Bạn làm gì |
|-----------|------------|
| Gặp error message | Copy → paste cho Claude → để Claude sửa |
| Trang trắng | Nói Claude kiểm tra console errors |
| Không kết nối được | Nói Claude kiểm tra server |
| Claude sửa 3 lần vẫn lỗi | Dừng lại, nhờ developer |
| Error "fatal"/"data loss" | Dừng lại NGAY, nhờ chuyên gia |

**Quy tắc vàng:** Gặp lỗi = bình thường. Hoảng loạn = không cần thiết. Copy-paste cho Claude = giải pháp 80% trường hợp.

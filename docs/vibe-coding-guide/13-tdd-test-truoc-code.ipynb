# 13. TDD — Test Trước, Code Sau

**Part 3: Superpowers — Quy trình làm việc** | Bước 3 trong 5 bước Superpowers

---

> *"Đề thi trước, bài làm sau."*

TDD (Test-Driven Development) là cách làm phần mềm mà bạn **viết bài kiểm tra trước**, rồi mới viết code.  
Nghe ngược đời? Thực ra rất thông minh. Hãy đọc tiếp.

## TDD là gì? Hai ví von dễ hiểu

### Ví von 1: Đề thi trước, bài làm sau

Trong trường học:
- Giáo viên ra **đề thi** trước (yêu cầu chất lượng)
- Học sinh làm **bài** sau (sản phẩm)
- Chấm bài → Đúng = PASS, Sai = FAIL

Trong phần mềm:
- AI viết **test** trước (yêu cầu chất lượng)
- AI viết **code** sau (sản phẩm)
- Chạy test → PASS = code đúng, FAIL = code sai

### Ví von 2: Yêu cầu chất lượng trước khi sản xuất

Nhà máy sản xuất áo:
- **Trước khi may:** Ghi rõ tiêu chuẩn — "đường may thẳng, nút áo chắc, không xô vải"
- **Sau khi may:** Kiểm tra theo tiêu chuẩn
- Đạt → giao hàng. Không đạt → sửa lại.

**TDD y hệt vậy:** Ghi tiêu chuẩn (test) trước → Sản xuất (code) → Kiểm tra.

## 3 bước: RED → GREEN → REFACTOR

TDD có 3 bước, lặp đi lặp lại:

### Bước 1: RED — Viết test, chạy, FAIL

AI viết bài kiểm tra cho tính năng **chưa có**.

```
Test: "Khi nhập email đúng và mật khẩu đúng → đăng nhập thành công"
Chạy test → FAIL (màu đỏ)
```

FAIL là **đúng**! Vì chưa có code nào cả. Nếu PASS ngay → test viết sai.

### Bước 2: GREEN — Viết code đủ để test PASS

AI viết code để tính năng hoạt động.

```
Code: Tạo chức năng đăng nhập, kiểm tra email + mật khẩu
Chạy test → PASS (màu xanh)
```

PASS = code đúng! Tính năng hoạt động.

### Bước 3: REFACTOR — Dọn dẹp code

Code hoạt động rồi, giờ **dọn cho sạch** (đặt tên rõ hơn, bỏ trùng lặp).

```
Dọn dẹp code → Chạy test lại → Vẫn PASS
```

Test vẫn PASS = dọn dẹp không làm hỏng gì.

### Vòng lặp TDD — nhìn cho dễ hiểu

```
    Viết TEST
       |
       v
   [ RED ]  ← Test FAIL (đúng rồi, chưa có code)
       |
       v
   Viết CODE
       |
       v
   [ GREEN ]  ← Test PASS (code đúng!)
       |
       v
   Dọn dẹp CODE
       |
       v
   [ REFACTOR ]  ← Test vẫn PASS (dọn không hỏng gì)
       |
       v
   Quay lại viết TEST cho tính năng tiếp theo...
```

## Tại sao người không biết code CẦN quan tâm đến TDD?

Bạn không cần viết test. Bạn không cần đọc test. Nhưng bạn cần **yêu cầu** AI viết test.

### Lý do 1: Test = "Lưới an toàn"

Hãy tưởng tượng diễn viên xiếc đi trên dây:
- **Không có lưới:** Sai 1 bước → rơi → chấn thương nặng
- **Có lưới:** Sai 1 bước → rơi → lưới đỡ → leo lên đi tiếp

Test = lưới an toàn cho code:
- **Không có test:** Sửa code → hỏng tính năng khác → không biết → người dùng phát hiện
- **Có test:** Sửa code → test báo FAIL → biết ngay → sửa trước khi người dùng thấy

### Lý do 2: AI tự kiểm tra chính nó

Khi AI viết test rồi viết code:
- AI **tự đặt câu hỏi** cho chính mình
- AI **tự kiểm tra** câu trả lời
- Bạn **không cần kiểm tra từng dòng code** — test làm điều đó

### Lý do 3: Bấm 1 nút, biết ngay đúng/sai

Thay vì tự mở app, tự bấm từng nút, tự kiểm tra... bạn chỉ cần:

```
Chạy test → 47 tests PASS, 0 FAIL → Yên tâm, mọi thứ OK.
```

Hoặc:

```
Chạy test → 45 tests PASS, 2 FAIL → Có 2 chỗ sai, AI sẽ sửa.
```

## Không có test = Lái xe không thắng

Đây không phải nói quá:

| Lái xe | Phần mềm |
|--------|----------|
| Không thắng → không dám chạy nhanh | Không test → không dám sửa code |
| Không đèn → đi đêm không thấy đường | Không test → lỗi ẩn không ai biết |
| Va chạm → thiệt hại lớn | Bug lên production → mất khách hàng |

**Có test:**
- Sửa code thoải mái → test báo ngay nếu hỏng
- Thêm tính năng mới → chắc chắn tính năng cũ không bị ảnh hưởng
- Deploy (đưa lên internet) → tự tin, không lo

## Bạn KHÔNG cần viết test — AI viết cho bạn

Đây là điểm quan trọng nhất:

**Vai trò của bạn:**
1. Yêu cầu AI: *"Viết test trước"*
2. Xem kết quả: PASS hay FAIL?
3. Nếu FAIL → nói AI sửa
4. Nếu PASS → tiếp tục

**Bạn KHÔNG cần:**
- Hiểu code trong test
- Viết test bằng tay
- Biết ngôn ngữ lập trình

**Ví dụ tương tác:**

```
Bạn: "Tạo tính năng tìm kiếm sản phẩm. Viết test trước."

AI: "Tôi đã viết 5 test cases:
  1. Tìm 'áo' → hiện các sản phẩm có từ 'áo'
  2. Tìm không có kết quả → hiện 'Không tìm thấy'
  3. Tìm ký tự đặc biệt → không bị lỗi
  4. Tìm chuỗi rỗng → hiện tất cả sản phẩm
  5. Tìm khi không có mạng → hiện lỗi
  
  Chạy test → 5/5 FAIL (chưa có code). Tôi bắt đầu viết code..."

AI: "Code xong. Chạy test → 5/5 PASS. Tính năng hoạt động."

Bạn: "Tốt, tiếp tục task tiếp theo."
```

## Ví dụ đơn giản — không cần biết code

Giả sử bạn muốn tính năng **"tính tổng giỏ hàng"**.

### Test (viết trước):

```
Test 1: Giỏ hàng trống → tổng = 0 đồng
Test 2: 1 sản phẩm 50.000đ → tổng = 50.000đ
Test 3: 2 sản phẩm (50.000đ + 30.000đ) → tổng = 80.000đ
Test 4: Sản phẩm x2 số lượng (50.000đ x 2) → tổng = 100.000đ
Test 5: Giảm giá 10% → tổng = 90.000đ (từ 100.000đ)
```

Chạy test → 5/5 FAIL (chưa có code) → **RED**

### Code (viết sau):

AI viết code tính tổng giỏ hàng.

Chạy test → 5/5 PASS → **GREEN**

### Dọn dẹp:

AI tối ưu code. Chạy test → 5/5 vẫn PASS → **REFACTOR**

**Bạn thấy gì?** Chỉ cần nhìn: PASS hay FAIL. Không cần đọc code.

## Script hỗ trợ: Chạy test nhanh

Kit có sẵn script để chạy test trước khi gửi code lên:

````bash
# Chạy test nhanh trước khi push code
# ./scripts/test-local.sh --quick

# Kết quả bạn sẽ thấy:
# ============================================
# TEST RESULTS
# ============================================
# Backend tests:  47 passed, 0 failed
# Frontend tests: 23 passed, 0 failed
# ============================================
# ALL TESTS PASSED - Safe to push!
# ============================================

echo "Script test-local.sh chạy TẤT CẢ test trong project."
echo "PASS = yên tâm push code."
echo "FAIL = có lỗi, AI sẽ sửa trước khi push."
````

## Khi test FAIL thì sao?

Đừng lo! FAIL không phải thảm họa. FAIL = **phát hiện lỗi TRƯỚC khi người dùng thấy**.

### Quy trình khi FAIL:

```
Chạy test → 2 FAIL
  ↓
AI xem lỗi: "Test 3 fail vì giảm giá tính sai"
  ↓
AI sửa code
  ↓
Chạy test lại → 0 FAIL
  ↓
OK! Tiếp tục.
```

**Bạn không cần làm gì.** AI tự phát hiện, tự sửa, tự kiểm tra lại.

## Khi nào có thể bỏ qua test?

Gần như **KHÔNG BAO GIỜ** nên bỏ qua test. Nhưng có vài ngoại lệ:

| Trường hợp | Cần test? |
|-----------|----------|
| Thêm tính năng mới | BẮT BUỘC |
| Sửa bug | BẮT BUỘC (test chứng minh bug đã sửa) |
| Đổi logic nghiệp vụ | BẮT BUỘC |
| Đổi màu nút, font chữ | Không cần |
| Sửa lỗi chính tả | Không cần |
| Cập nhật tài liệu | Không cần |

**Quy tắc:** Nếu thay đổi ảnh hưởng đến **hành vi** (cái gì đó hoạt động khác đi) → cần test.

## Cách yêu cầu AI viết test

````bash
# Cách 1: Yêu cầu trực tiếp
# "Viết test trước, rồi mới viết code cho tính năng tìm kiếm."

# Cách 2: Dùng /start-pr (TDD tự động nằm trong quy trình)
# /start-pr Tính năng tìm kiếm sản phẩm

# Cách 3: Nhắc nhở nếu AI quên
# "Khoan, test đâu? Viết test trước rồi mới code."
````

## Lợi ích TDD — tóm gọn

| Không có TDD | Có TDD |
|-------------|--------|
| Sửa code → lo hỏng chỗ khác | Sửa code → test báo ngay nếu hỏng |
| Lỗi phát hiện khi người dùng dùng | Lỗi phát hiện ngay khi code xong |
| Không dám sửa code cũ | Thoải mái sửa, test bảo vệ |
| "Hình như nó chạy đúng..." | "47/47 test PASS. Chắc chắn đúng." |
| AI viết code → bạn phải tự kiểm tra | AI viết test + code → tự kiểm tra |

## Tóm tắt

| Điều cần nhớ | Chi tiết |
|--------------|----------|
| **TDD là gì** | Viết bài kiểm tra TRƯỚC, viết code SAU |
| **3 bước** | RED (test fail) → GREEN (code pass) → REFACTOR (dọn dẹp) |
| **Bạn cần làm gì** | Yêu cầu AI "viết test trước", xem PASS/FAIL |
| **Bạn KHÔNG cần** | Viết test, đọc code test, hiểu ngôn ngữ lập trình |
| **Test = lưới an toàn** | Sửa code thoải mái, test bảo vệ |
| **FAIL = tốt** | Phát hiện lỗi trước khi người dùng thấy |

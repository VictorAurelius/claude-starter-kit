## Starter Kit là gì?

Hãy tưởng tượng bạn thuê một **đội thợ xây nhà**.

Nếu bạn chỉ nói "xây cho tôi cái nhà" mà không đưa gì cả:
- Thợ tự mua búa, cưa, thước... mỗi người mua loại khác
- Không có bản vẽ, mỗi người xây theo ý mình
- Không có quy trình kiểm tra, sai cũng không biết

Nhưng nếu bạn đưa cho đội thợ một **bộ dụng cụ chuẩn**:
- Búa, cưa, thước đã chọn sẵn loại tốt
- Bản vẽ chi tiết từng phòng
- Checklist kiểm tra sau mỗi bước
- Sổ tay ghi lại bài học từ công trình trước

**Starter Kit chính là bộ dụng cụ chuẩn đó** — nhưng dành cho AI.

## Starter Kit KHÔNG phải code

Điều quan trọng cần hiểu: Starter Kit **không phải là phần mềm**. Nó là:

| Thành phần | Vai trò | Ví von đời thực |
|-----------|--------|------------------|
| **Quy trình** (Skills) | Dạy AI cách làm việc đúng | Sổ tay hướng dẫn nghề |
| **Quy tắc** (Rules) | Giới hạn AI không làm bừa | Nội quy công trường |
| **Công cụ** (Scripts) | Tự động hóa kiểm tra | Máy đo, máy kiểm tra |
| **Mẫu** (Templates) | File mẫu điền vào là xong | Form đơn có sẵn |
| **Trí nhớ** (Memory) | Bài học từ dự án trước | Sổ bàn giao ca trước |

Tất cả được gói gọn trong **1 folder** tên `.claude/` nằm trong dự án.

## Tại sao cần Kit?

Claude Code rất mạnh — nó có thể viết code, tạo file, chạy lệnh. Nhưng...

### Không có Kit:
- AI viết code kiểu gì tùy ý, mỗi lần mỗi khác
- Quên kiểm tra lỗi trước khi xong
- Không biết quy tắc riêng của dự án bạn
- Lặp lại sai lầm mà phiên trước đã sửa

### Có Kit:
- AI tuân theo **quy trình chuẩn**: suy nghĩ trước, viết test, review
- **Tự động kiểm tra** trước mỗi commit
- Biết đúng convention của dự án (đặt tên file, cấu trúc folder...)
- **Nhớ bài học** từ phiên trước, không lặp sai lầm

Nói ngắn gọn: **Kit biến AI từ "thợ tự do" thành "nhân viên được đào tạo"**.

## Giải thích từng folder

### 1. `rules/` — Nội quy công trường

Chứa các quy tắc mà AI **bắt buộc** phải tuân theo. Ví dụ:
- Cách viết skill đúng chuẩn (không dài quá, không dạy AI điều nó đã biết)
- Quy ước đặt tên file

Bạn không cần sửa trực tiếp — Kit đã có sẵn rules tốt.

### 2. `skills/` — Chứng chỉ nghề

Đây là **trái tim** của Kit. Mỗi skill dạy AI một kỹ năng cụ thể:

| Nhóm | Số lượng | Ví dụ |
|------|---------|-------|
| **core/** | 5 skills | Brainstorm, chia việc, viết test, review, debug |
| **workflow/** | 6+ skills | Quy trình tạo PR, theo dõi CI |
| **quality/** | 2 skills | Chấm điểm chất lượng /100, review giao diện |
| **reference/** | 10+ skills | Cấu trúc dự án, tài liệu nghiệp vụ |

AI **tự biết** khi nào cần dùng skill nào — nhờ phần mô tả trong mỗi skill.

### 3. `scripts/` — Nút bấm tự động

Thay vì gõ nhiều lệnh phức tạp, bạn chỉ cần chạy 1 script:

- `pre-commit-check.sh` — Kiểm tra tự động trước mỗi commit
- `commit-msg-check.sh` — Kiểm tra định dạng commit message

**Quy tắc quan trọng:** AI phải dùng scripts, KHÔNG được chạy lệnh trực tiếp. Vì scripts đã được kiểm tra và chuẩn hóa.

### 4. `memory/` — Sổ bàn giao

Claude Code mỗi lần mở = 1 phiên mới, **không nhớ** phiên trước. Memory files giải quyết vấn đề này:

- AI đọc memory lúc bắt đầu → biết ngay tình trạng dự án
- Ghi lại bài học sai lầm → không lặp lại
- Kit đi kèm 4 bài học từ dự án thực (seed memories)

```bash
# Xem phiên bản Starter Kit
# File .starter-kit-version chứa số version theo quy tắc semver (MAJOR.MINOR.PATCH)

echo "Ví dụ nội dung file .starter-kit-version:"
echo "2.1.0"
echo ""
echo "Ý nghĩa:"
echo "  2 = phiên bản lớn (thay đổi cấu trúc)"
echo "  1 = tính năng mới (thêm skill, script)"
echo "  0 = sửa nhỏ (cải thiện nội dung)"
```

## Versioning — Kit cũng có "phiên bản"

Giống như ứng dụng điện thoại có version 1.0, 2.0... Kit cũng vậy.

**Tại sao cần version?**
- Biết Kit trong dự án đang ở version nào
- Khi Kit được cập nhật (thêm skill mới, sửa lỗi), bạn có thể update
- Nhiều dự án dùng chung Kit → cần đồng bộ version

**Quy tắc version (Semantic Versioning):**

| Thay đổi | Loại bump | Ví dụ |
|----------|-----------|-------|
| Xóa/thay đổi cấu trúc lớn | MAJOR | 1.0.0 → **2**.0.0 |
| Thêm skill, script mới | MINOR | 2.0.0 → 2.**1**.0 |
| Sửa nội dung, cải thiện | PATCH | 2.1.0 → 2.1.**1** |

```bash
# Liệt kê các skill trong Kit
# Trong thực tế, bạn chạy lệnh này ở thư mục dự án

echo "=== Core Skills (dùng mỗi PR) ==="
echo "  brainstorming-methodology.md"
echo "  task-breakdown-guide.md"
echo "  tdd-enforcement.md"
echo "  two-stage-code-review.md"
echo "  systematic-debugging.md"
echo ""
echo "=== Workflow Skills ==="
echo "  development-workflow.md"
echo "  check-pr/"
echo "  fix-pr/"
echo "  start-pr/"
echo "  continue/"
echo ""
echo "=== Quality Skills ==="
echo "  quality-audit/  (chấm điểm /100)"
echo "  ui-review/      (đánh giá giao diện /128)"
echo ""
echo "=== Reference Skills ==="
echo "  business-docs-3-layer.md"
echo "  project-structure.md"
echo "  ide-setup.md"
echo "  ui-template-guide.md"
echo "  ... và nhiều hơn nữa"
```

## Cài đặt Kit vào dự án

Kit được cài vào dự án bằng cách **copy folder `.claude/`** vào gốc dự án.

Sau khi cài:
1. Claude Code tự động đọc `settings.local.json` (quyền hạn)
2. Tự động đọc `CLAUDE.md` ở gốc dự án (hướng dẫn tổng quát)
3. Khi cần, AI tự tìm đọc skills phù hợp
4. Scripts sẵn sàng chạy

Bạn không cần "cài đặt" hay "chạy setup" gì phức tạp — chỉ cần folder `.claude/` tồn tại.

```bash
# Kiểm tra Kit đã được cài chưa
# Chạy ở thư mục gốc dự án

echo "Kiểm tra Starter Kit:"
echo ""

# Kiểm tra folder .claude tồn tại
if [ -d ".claude" ]; then
    echo "[OK] Folder .claude/ tồn tại"
else
    echo "[X] Folder .claude/ KHÔNG tồn tại — cần cài Kit"
fi

# Kiểm tra version file
if [ -f ".claude/.starter-kit-version" ]; then
    echo "[OK] Version: $(cat .claude/.starter-kit-version)"
else
    echo "[X] Không tìm thấy file version"
fi

# Kiểm tra skills
if [ -d ".claude/skills" ]; then
    SKILL_COUNT=$(find .claude/skills -name '*.md' | wc -l)
    echo "[OK] Skills: $SKILL_COUNT files"
else
    echo "[X] Không có skills"
fi
```

## So sánh: Có Kit vs Không có Kit

| Tình huống | Không có Kit | Có Kit |
|-----------|-------------|--------|
| Bạn nhờ AI tạo tính năng mới | AI nhảy thẳng vào code | AI brainstorm trước, chia nhỏ, viết test |
| AI commit code | Có thể lỗi, không kiểm tra | Script tự động kiểm tra trước khi commit |
| Phiên làm việc mới | AI quên hết, bắt đầu lại | AI đọc memory, biết ngay tình trạng |
| AI cần tạo file mới | Tự nghĩ format, không nhất quán | Dùng template có sẵn, đúng chuẩn |
| Kiểm tra chất lượng | "Trông ổn rồi" | Chấm điểm /100 với 10+ tiêu chí cụ thể |

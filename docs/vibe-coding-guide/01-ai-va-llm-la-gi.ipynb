Trong notebook này, bạn sẽ hiểu được:

## 1. AI là gì?

### Ví von thực tế

Hãy tưởng tượng bạn thuê một **người học việc cực kỳ siêng năng**:
- Người này đã đọc hàng triệu cuốn sách, bài báo, tài liệu
- Có thể viết, tóm tắt, dịch thuật, thậm chí viết code
- Làm việc 24/7 không mệt mỏi
- **Nhưng** — đôi khi tự tin nói sai mà không biết mình sai

Đó chính là AI (Artificial Intelligence — Trí tuệ nhân tạo) ngày nay.

### Định nghĩa đơn giản

**AI = phần mềm được huấn luyện để thực hiện các tác vụ mà trước đây chỉ con người mới làm được** — như hiểu ngôn ngữ, nhận diện hình ảnh, ra quyết định.

AI **không phải** robot có ý thức. AI **không** "suy nghĩ" như con người. AI là một chương trình máy tính rất giỏi trong việc nhận ra pattern (mẫu) từ dữ liệu.

## 2. Machine Learning vs AI

Bạn sẽ nghe hai thuật ngữ này thường xuyên. Đây là cách phân biệt đơn giản:

| | AI (Trí tuệ nhân tạo) | Machine Learning (Học máy) |
|---|---|---|
| **Là gì** | Thuật ngữ chung cho máy "thông minh" | Một **cách** để tạo ra AI |
| **Ví von** | "Nấu ăn" | "Nấu ăn bằng công thức" |
| **Cách hoạt động** | Nhiều cách khác nhau | Cho máy học từ dữ liệu |

**Machine Learning** = thay vì lập trình từng bước cho máy tính ("nếu A thì làm B"), ta cho máy tính **học từ ví dụ** ("đây là 1 triệu email spam, hãy tự tìm ra quy luật").

> Bạn **không cần nhớ** sự khác biệt này để vibe code. Chỉ cần biết: AI ngày nay chủ yếu dùng Machine Learning.

## 3. LLM — Large Language Model (Mô hình ngôn ngữ lớn)

### Ví von: Người đọc hết thư viện

Hãy tưởng tượng một người:
- Đã **đọc hết** mọi sách, bài báo, website, tài liệu trên internet (hàng tỷ trang)
- Nhờ đọc nhiều nên **hiểu cách ngôn ngữ hoạt động** — biết sau "Xin" thường là "chào", sau "Hà Nội là thủ đô của" thường là "Việt Nam"
- Có thể **viết**, **tóm tắt**, **dịch**, **trả lời câu hỏi**, và thậm chí **viết code**

Đó là LLM. Claude, ChatGPT, Gemini — tất cả đều là LLM.

### Cách LLM hoạt động (cực kỳ đơn giản)

1. **Huấn luyện**: Đọc hàng tỷ trang văn bản từ internet
2. **Học pattern**: Tìm ra quy luật ngôn ngữ — từ nào hay đi cùng từ nào
3. **Dự đoán từ tiếp theo**: Khi bạn hỏi, AI "viết tiếp" câu trả lời, mỗi lần một từ

Nghe đơn giản, nhưng khi mô hình đủ lớn (hàng trăm tỷ tham số), nó bất ngờ có những khả năng **vượt xa** việc đoán từ — như suy luận, viết code, giải toán. Các nhà khoa học gọi đây là **emergent abilities** (khả năng nổi lên).

### LLM có thể làm gì?

| Việc | Ví dụ |
|------|-------|
| Viết văn bản | Email, báo cáo, bài thuyết trình |
| Tóm tắt | Đọc tài liệu 50 trang, tóm lại 1 trang |
| Dịch thuật | Anh-Việt, Việt-Nhật, bất kỳ ngôn ngữ nào |
| Trả lời câu hỏi | Giải thích khái niệm, tư vấn |
| Viết code | Tạo website, app, script tự động |
| Phân tích | Đọc data, tìm insight |

Và đây là lý do bạn — một người không biết code — có thể dùng AI để tạo ra phần mềm.

## 4. Những khái niệm bạn CẦN biết khi dùng AI

Phần này rất quan trọng. Bạn sẽ gặp các thuật ngữ này liên tục khi làm việc với AI.

### 4.1 Token — Đơn vị ngôn ngữ của AI

#### Ví von: Mảnh ghép chữ

Khi bạn nói chuyện với AI, AI không đọc từng **chữ cái** hay từng **từ** — nó đọc theo **token**.

Token giống như **mảnh ghép puzzle của ngôn ngữ**:
- Từ ngắn = 1 token: `cat`, `the`, `is`
- Từ dài = nhiều token: `programming` = `program` + `ming` (2 tokens)
- **Tiếng Việt tốn nhiều token hơn tiếng Anh** vì có dấu và ít phổ biến trong dữ liệu huấn luyện

#### Quy tắc ước lượng nhanh
- Tiếng Anh: ~4 ký tự = 1 token (100 từ gần bằng 75 tokens)
- Tiếng Việt: ~2-3 ký tự = 1 token (tốn gấp 1.5-2 lần tiếng Anh)

#### Tại sao cần biết?
- AI tính phí theo token (càng nhiều token = càng tốn tiền)
- AI có giới hạn token trong 1 lần hội thoại (xem phần Context Window)

```bash
# Demo: Đếm token bằng cách ước lượng
# (Đây chỉ là ước lượng — bộ đếm thật sự phụ thuộc vào từng model)

echo "=== Ước lượng Token ==="
echo ""

# Tiếng Anh: ~4 ký tự = 1 token
english="Hello, I want to build a website for my school"
en_chars=${#english}
en_tokens=$((en_chars / 4))
echo "Tiếng Anh: '$english'"
echo "  Ký tự: $en_chars | Ước lượng tokens: ~$en_tokens"
echo ""

# Tiếng Việt: ~2.5 ký tự = 1 token
vietnamese="Xin chào, tôi muốn xây dựng website cho trường học"
vi_chars=${#vietnamese}
vi_tokens=$((vi_chars * 10 / 25))  # chia 2.5
echo "Tiếng Việt: '$vietnamese'"
echo "  Ký tự: $vi_chars | Ước lượng tokens: ~$vi_tokens"
echo ""
echo "=> Tiếng Việt tốn nhiều token hơn cho cùng nội dung!"
```

### 4.2 Context Window — Bộ nhớ tạm của AI

#### Ví von: Bàn làm việc

Hãy tưởng tượng AI có một **bàn làm việc**:
- Bàn chỉ rộng có hạn (ví dụ: 200,000 tokens)
- Mọi thứ bạn nói + AI trả lời + file AI đọc = đều nằm trên bàn này
- Khi bàn đầy → những thứ cũ nhất bị "rơi xuống đất" (AI quên)
- **Bắt đầu cuộc hội thoại mới = dọn sạch bàn** (AI quên hết)

#### Context window của các model phổ biến

| Model | Context Window | Ví von |
|-------|---------------|--------|
| Claude (Anthropic) | 200,000 tokens | Bàn làm việc siêu rộng — chứa được ~500 trang sách |
| GPT-4o (OpenAI) | 128,000 tokens | Bàn rộng — ~320 trang |
| Gemini 1.5 Pro | 1,000,000 tokens | Cả phòng làm việc — ~2,500 trang |

#### Tại sao cần biết?

Khi làm dự án lớn (nhiều file code), context window quyết định **AI có thể "nhìn thấy" bao nhiêu code cùng lúc**. Claude Code thông minh trong việc chọn lọc file nào cần đọc để không "đầy bàn".

```bash
# Demo: Context window tương đương bao nhiêu nội dung?

echo "=== Context Window trong đời thực ==="
echo ""
echo "Claude 200K tokens tương đương:"
echo "  - ~150,000 từ tiếng Anh (khoảng 500 trang sách)"
echo "  - ~100,000 từ tiếng Việt (khoảng 330 trang sách)"
echo "  - ~5,000 dòng code Python"
echo "  - Cả một quyển tiểu thuyết ngắn!"
echo ""
echo "Thực tế khi vibe coding:"
echo "  - Prompt của bạn: ~500-2,000 tokens"
echo "  - File code AI đọc: ~1,000-10,000 tokens mỗi file"
echo "  - Câu trả lời AI: ~500-5,000 tokens"
echo "  - Lịch sử hội thoại tích lũy dần..."
echo ""
echo "=> Khi hội thoại quá dài, hãy bắt đầu cuộc mới!"
```

### 4.3 Temperature — Mức "sáng tạo" của AI

#### Ví von: Nút điều chỉnh trên bếp

Giống như bếp có nút chỉnh lửa:

| Temperature | Lửa | AI sẽ... | Dùng khi |
|------------|------|---------|----------|
| **0.0** | Tắt bếp | Luôn chọn câu trả lời "chắc chắn nhất" | Viết code, trả lời chính xác |
| **0.3-0.5** | Lửa nhỏ | Khá chính xác, thỉnh thoảng sáng tạo | Tóm tắt, phân tích |
| **0.7-1.0** | Lửa vừa | Đa dạng hơn, có thể bất ngờ | Viết văn, brainstorm |
| **1.5-2.0** | Lửa to | Rất "bay bổng", có thể vô nghĩa | Thơ, sáng tạo nghệ thuật |

#### Bạn cần nhớ gì?

- **Viết code** → temperature **thấp** (chính xác hơn)
- **Brainstorm ý tưởng** → temperature **cao hơn** (đa dạng hơn)
- Khi dùng Claude Code, temperature đã được chỉnh sẵn cho phù hợp — bạn **không cần điều chỉnh**

### 4.4 Hallucination — Khi AI "bịa"

#### Ví von: Học sinh giỏi nhưng hay bịa

Bạn hỏi AI: *"Tác giả cuốn sách 'Cánh Đồng Pha Lê' là ai?"*

AI có thể trả lời rất tự tin: *"Cuốn sách 'Cánh Đồng Pha Lê' được viết bởi Nguyễn Văn A, xuất bản năm 2015..."*

Vấn đề? **Cuốn sách đó có thể không tồn tại.** AI không tra cứu thông tin — nó **tạo ra** câu trả lời dựa trên pattern. Nếu pattern đủ "hợp lý", AI sẽ tự tin nói dù nó sai.

#### Tại sao AI hallucinate?

1. AI **không có cơ sở dữ liệu sự kiện** — nó chỉ biết pattern ngôn ngữ
2. AI được huấn luyện để **luôn đưa ra câu trả lời** (không được nói "tôi không biết" quá nhiều)
3. Thông tin cũ/hiếm → AI có ít dữ liệu → dễ bịa

#### Cách phòng tránh (rất quan trọng!)

| Cách | Ví dụ |
|------|-------|
| Yêu cầu AI nói "không biết" | "Nếu không chắc, hãy nói không biết" |
| Kiểm tra chéo | Google lại thông tin quan trọng |
| Cho AI context | Đưa tài liệu thật để AI dựa vào, thay vì hỏi "từ trí nhớ" |
| Dùng Claude Code | Claude Code đọc code thật → ít bịa hơn vì có file thật trước mắt |

```bash
# Mẹo thực tế: Cách giảm hallucination khi dùng AI

echo "=== 4 Mẹo giảm Hallucination ==="
echo ""
echo "1. Cung cấp context (tài liệu, file):"
echo '   BAD:  "Viết API cho quản lý học sinh"'
echo '   GOOD: "Đọc file rules.md, rồi viết API theo đúng business rules trong đó"'
echo ""
echo "2. Yêu cầu trích dẫn nguồn:"
echo '   "Trả lời dựa trên file nào? Trích dẫn dòng cụ thể."'
echo ""
echo "3. Hỏi lại khi không chắc:"
echo '   "Bạn có chắc không? Nếu không chắc, nói rõ mức độ tự tin."'
echo ""
echo "4. Dùng Claude Code thay vì Chat:"
echo "   Claude Code đọc file thật trong project => giảm bịa đáng kể"
```

## 5. Những thứ bạn KHÔNG CẦN biết

Nhiều tài liệu AI sẽ nói về các khái niệm sau. Bạn **hoàn toàn có thể bỏ qua** chúng khi vibe coding:

| Khái niệm | Tại sao không cần |
|-----------|-------------------|
| Neural Network | Kiến trúc bên trong AI — bạn dùng xe mà không cần hiểu động cơ |
| Training/Fine-tuning | Cách huấn luyện AI — bạn là người dùng, không phải nhà phát triển AI |
| Attention mechanism | Chi tiết kỹ thuật — không ảnh hưởng cách bạn dùng AI |
| Gradient descent | Thuật toán tối ưu — dành cho researcher |
| Embeddings | Cách AI biểu diễn ngôn ngữ — dành cho developer chuyên sâu |

Bạn **không cần biết cách động cơ hoạt động để lái xe**. Tương tự, bạn không cần hiểu neural network để dùng AI tạo phần mềm.

## 6. Mental Model — Cách nghĩ đúng về AI

Trước khi kết thúc, hãy xây dựng một "mô hình tư duy" đúng về AI:

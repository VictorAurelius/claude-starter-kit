## 1. "Vibe Coding" — Thuật ngữ mới

### Nguồn gốc

Tháng 2/2025, **Andrej Karpathy** — một trong những nhà khoa học AI hàng đầu thế giới (cựu director AI tại Tesla, co-founder OpenAI) — đã tweet:

> *"There's a new kind of coding I call 'vibe coding', where you fully give in to the vibes, embrace exponentials, and forget that the code even exists."*

Dịch nôm na:

> *"Có một kiểu code mới tôi gọi là 'vibe coding' — bạn hoàn toàn thả theo cảm xúc, tin tưởng AI, và quên đi sự tồn tại của code."*

### Nghĩa là gì?

**Vibe Coding = Bạn mô tả bằng ngôn ngữ tự nhiên, AI viết code cho bạn.**

Bạn không cần đọc code. Bạn không cần hiểu syntax. Bạn chỉ cần biết **bạn muốn gì**.

### Ví von: Bạn là đạo diễn phim

Hãy tưởng tượng bạn đang làm phim:

| Vai trò | Trong phim | Trong Vibe Coding |
|---------|-----------|-------------------|
| **Đạo diễn** (bạn) | Quyết định câu chuyện, góc quay, cảm xúc | Quyết định tính năng, giao diện, trải nghiệm |
| **Đoàn quay phim** (AI) | Quay phim, dựng hình, chỉnh màu | Viết code, tạo UI, kết nối database |

Đạo diễn **không cần biết** cách vận hành máy quay. Đạo diễn cần biết **câu chuyện hay trông như thế nào**.

Tương tự:
- **Bạn quyết định WHAT** (làm gì, cho ai, trải nghiệm ra sao)
- **AI quyết định HOW** (code bằng ngôn ngữ gì, kiến trúc thế nào, query database ra sao)

````bash
# Ví dụ: Vibe coding trong thực tế

echo "=== Ví dụ: Bạn nói gì, Claude Code làm gì ==="
echo ""
echo "--- Session 1: Khởi tạo project ---"
echo 'Ban: "Tao project website quan ly lop hoc, dung React + TypeScript."'
echo 'AI:  Tao 15 file, cai 8 packages, setup router, tao layout chinh.'
echo ""
echo "--- Session 2: Trang dang nhap ---"
echo 'Ban: "Tao trang dang nhap, co email va password, nut Dang nhap mau xanh."'
echo 'AI:  Tao LoginPage.tsx, validation, error messages, connect API.'
echo ""
echo "--- Session 3: Dashboard ---"
echo 'Ban: "Sau dang nhap, hien dashboard voi so luong hoc sinh, lop, khoa hoc."'
echo 'AI:  Tao DashboardPage.tsx, 3 stat cards, goi API lay du lieu.'
echo ""
echo "--- Session 4: Fix bug ---"
echo 'Ban: "Khi chua dang nhap, trang dashboard van truy cap duoc. Fix di."'
echo 'AI:  Them AuthGuard route, redirect ve login neu chua dang nhap.'
echo ""
echo "=> Ban khong viet 1 dong code nao. Ban chi mo ta." 
````

## 2. Tại sao non-dev CÓ THỂ vibe code?

Nhiều người nghĩ: *"Tôi không biết code, sao tôi tạo được phần mềm?"*

Câu trả lời: **Bạn biết thứ mà developer thiếu.**

### Bạn có lợi thế gì?

| Lợi thế của bạn | Tại sao quan trọng |
|-----------------|-------------------|
| **Hiểu sản phẩm** | Bạn biết người dùng cần gì — developer thường phải hỏi lại bạn |
| **Hiểu nghiệp vụ** | Bạn biết quy trình (nhập học, điểm danh, chấm điểm) — developer phải học từ đầu |
| **Hiểu user** | Bạn biết UX tốt trông thế nào — developer hay làm UI khó dùng |
| **Ra quyết định nhanh** | Bạn quyết ngay "tính năng này cần/không cần" — developer phải chờ PM quyết |

### AI bổ sung phần bạn thiếu

| Bạn thiếu | AI có |
|-----------|-------|
| Kiến thức lập trình | Biết hàng trăm ngôn ngữ, framework |
| Kinh nghiệm code | "Đọc" hàng triệu project, biết best practices |
| Tốc độ gõ code | Viết 100 dòng code trong vài giây |
| Kiên nhẫn debug | Thử đi thử lại không mệt |

## 3. Tại sao Vibe Coding DỄ THẤT BẠI

Vibe coding nghe tuyệt vời, nhưng **phần lớn người thử đều thất bại**. Tại sao?

### Vấn đề 1: AI bịa code, không ai kiểm tra

AI viết code trông hợp lý nhưng... sai. Và vì bạn không đọc code, bạn không biết nó sai.

```
Ban: "Tinh tong diem cua hoc sinh"
AI:  Viet code tinh tong... nhung tinh SAI khi co diem null
Ban: Khong biet, app hien so dep dep => "OK, ship thoi!"
User: "Sao diem cua con toi bi tinh sai?!"
```

### Vấn đề 2: Không có test, sửa chỗ này hỏng chỗ kia

```
Ban: "Fix bug trang diem danh"
AI:  Fix xong diem danh... nhung VO TINH lam hong trang xem diem
Ban: Khong biet vi chua kiem tra trang xem diem
1 tuan sau, giao vien phan nan: "Trang xem diem bi loi!"
```

### Vấn đề 3: Không có cấu trúc, project thành "mì Ý"

**Spaghetti code** — khi code chằng chéo như mì Ý, không ai hiểu, không sửa được:

```
Thang 1: Project chay ngon lanh (500 dong code)
Thang 3: Them tinh nang, bat dau cham (5,000 dong code)
Thang 6: Them 1 nut bam ma mat 3 ngay (20,000 dong code, chong cheo)
Thang 9: "Bo di lam lai tu dau con nhanh hon" (classic!)
```

## 4. GIẢI PHÁP: Starter Kit

### Ví von: Nhà máy sản xuất

Hãy tưởng tượng một nhà máy:

| Vai trò | Trong nhà máy | Trong Vibe Coding |
|---------|-------------|-------------------|
| **Giám đốc** (bạn) | Quyết định sản xuất gì, tiêu chuẩn nào | Quyết định tính năng, chất lượng |
| **Công nhân** (AI) | Sản xuất sản phẩm | Viết code |
| **QC** (Starter Kit) | Kiểm tra chất lượng mỗi sản phẩm | Kiểm tra code tự động |

Nếu nhà máy **không có QC** → sản phẩm lỗi ra thị trường → khách hàng phản ánh → mất uy tín.

Nếu vibe coding **không có Starter Kit** → code lỗi lên production → user gặp bug → dự án thất bại.

### Starter Kit là gì?

Starter Kit = **bộ quy trình + công cụ kiểm tra** được cài sẵn vào dự án:

```
.claude/
├── CLAUDE.md          ← "Sổ tay" cho AI (quy tắc, memory)
├── skills/            ← Kỹ năng chuyên biệt
│   ├── core/          ← Quy trình cốt lõi (brainstorm, TDD, review)
│   ├── quality/       ← Kiểm tra chất lượng (audit, UI review)
│   └── devops/        ← Triển khai (deploy, CI/CD)
├── hooks/             ← Hành động tự động
└── scripts/           ← Script tiện ích
```

### Starter Kit giải quyết từng vấn đề

| Vấn đề | Không có Kit | Có Kit |
|--------|-------------|--------|
| AI bịa code | Không ai kiểm tra | **TDD**: Viết test trước → code phải pass test |
| Sửa A hỏng B | Không biết B bị hỏng | **Auto-test**: Mỗi commit chạy toàn bộ test |
| Code spaghetti | Chằng chéo không sửa được | **Code Review**: AI tự review trước khi commit |
| Không hiểu code | Đúng, bạn không đọc code | **Quality Audit**: Chấm điểm /100, cho biết code tốt/xấu |
| AI quên context | Bắt đầu lại từ đầu | **Memory**: CLAUDE.md lưu mọi quy tắc |
| Không nhất quán | Mỗi lần AI làm khác | **Skills**: Quy trình chuẩn, AI làm giống nhau mỗi lần |

````bash
# Demo: Starter Kit kiểm soát chất lượng như thế nào

echo "=== Quy trinh vibe coding VOI Starter Kit ==="
echo ""
echo "Ban: 'Tao tinh nang diem danh hoc sinh'"
echo ""
echo "[Buoc 1: Brainstorm] - Skill: brainstorming-methodology"
echo "  AI phan tich: scope, risks, edge cases"
echo "  => 'Diem danh can xu ly: nghi phep, di muon, nghi khong phep'"
echo ""
echo "[Buoc 2: Task Breakdown] - Skill: task-breakdown-guide"
echo "  AI chia nho thanh 5 tasks:"
echo "  1. Tao database table (30 phut)"
echo "  2. Tao API endpoint (1 gio)"
echo "  3. Tao frontend page (1 gio)"
echo "  4. Ket noi FE + BE (30 phut)"
echo "  5. Test toan bo (30 phut)"
echo ""
echo "[Buoc 3: TDD] - Skill: tdd-enforcement"
echo "  AI viet test TRUOC: 'Khi giao vien diem danh, hoc sinh A = co mat'"
echo "  Chay test => DO (that bai vi chua co code)"
echo "  AI viet code => XANH (test pass)"
echo ""
echo "[Buoc 4: Implementation]"
echo "  AI code, commit thuong xuyen, moi commit chay test tu dong"
echo ""
echo "[Buoc 5: Code Review] - Skill: two-stage-code-review"
echo "  AI tu review code cua chinh minh"
echo "  => Tim duoc 2 loi: thieu validation, thieu error message"
echo "  => Fix truoc khi ban nhin thay"
echo ""
echo "[Buoc 6: Quality Audit] - Skill: quality-audit"
echo "  Diem: 92/100"
echo "  => Code sach, co test, co doc, san sang ship!"
echo ""
echo "=> TAT CA dieu nay xay ra TU DONG. Ban chi noi 1 cau dau tien."
````

## 6. Starter Kit chi tiết hơn

### Skills cốt lõi ("Superpowers")

Mỗi khi AI làm việc, nó sẽ **tự động** qua 5 bước:

```
1. BRAINSTORM  =>  2. TASK BREAKDOWN  =>  3. TDD  =>  4. CODE  =>  5. REVIEW
   (5 phut)           (5 phut)             (test     (viet       (tu kiem
                                            truoc)    code)       tra)
```

#### 1. Brainstorm
AI phân tích yêu cầu của bạn: cần làm gì, rủi ro gì, edge case nào?

#### 2. Task Breakdown
AI chia nhỏ công việc thành các bước cụ thể, ước lượng thời gian.

#### 3. TDD (Test-Driven Development)
AI viết **test trước** — định nghĩa "thành công" trông như thế nào. Rồi mới viết code để pass test.

#### 4. Implementation
AI viết code, commit thường xuyên (mỗi commit = 1 save point).

#### 5. Code Review
AI tự review code của chính nó — tìm lỗi, cải thiện trước khi bạn nhìn thấy.

### Quality Audit — Thước đo bạn HIỂU ĐƯỢC

Đây là phần hay nhất cho non-dev. Bạn không đọc code, nhưng bạn hiểu **điểm số**.

Khi chạy `/audit`, Starter Kit chấm điểm 10 hạng mục, tổng 100 điểm:

| Hạng mục | Điểm tối đa | Kiểm tra gì |
|----------|------------|-------------|
| Business Logic | 15 | Code có đúng nghiệp vụ không? |
| Test Coverage | 15 | Có đủ test không? |
| Code Quality | 10 | Code sạch, dễ đọc? |
| API Contract | 10 | API đúng chuẩn? |
| Error Handling | 10 | Xử lý lỗi tốt? |
| Security | 10 | An toàn? |
| Documentation | 10 | Có tài liệu? |
| Performance | 5 | Nhanh? |
| Accessibility | 5 | Ai cũng dùng được? |
| CI/CD | 10 | Deploy tự động? |

**Quy tắc**: Điểm >= 85 mới được merge vào sản phẩm.

Bạn không cần hiểu code. Bạn chỉ cần hỏi: **"Điểm bao nhiêu?"**

## 8. Hỏi đáp nhanh (FAQ)

**Q: Tôi có cần học lập trình trước khi vibe code không?**
A: Không. Nhưng hiểu cơ bản Terminal và Git (notebook 03) sẽ giúp bạn tự tin hơn.

**Q: Vibe coding có thay thế developer không?**
A: Không hoàn toàn. Với project phức tạp (triệu user, real-time, high security), vẫn cần developer kinh nghiệm. Nhưng cho MVP, prototype, internal tools — vibe coding rất hiệu quả.

**Q: Nếu AI viết code sai thì sao?**
A: Starter Kit có TDD (test bắt lỗi tự động), Code Review (AI tự kiểm tra), và Quality Audit (chấm điểm). 3 lớp bảo vệ này bắt được phần lớn lỗi.

**Q: Tốn bao nhiêu tiền?**
A: Claude Code cần subscription Anthropic (~$20/tháng cho Pro, hoặc trả theo usage). VS Code, Git, GitHub đều miễn phí.

**Q: Tôi có thể tạo app mobile được không?**
A: Được! Claude Code có thể tạo React Native (chạy trên cả iOS và Android), Flutter, hoặc web app responsive (hoạt động trên mobile).

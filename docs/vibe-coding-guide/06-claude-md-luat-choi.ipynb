> CLAUDE.md là file **quan trọng nhất** trong toàn bộ dự án. AI đọc nó MỖI LẦN bạn mở conversation.

## CLAUDE.md là gì?

Hãy nghĩ về việc thuê một **nhân viên mới**.

Ngày đầu đi làm, nhân viên cần:
- **Hợp đồng lao động** — nêu rõ công việc, quy tắc, kỳ vọng
- **Sổ tay nhân viên** — cách làm việc, quy trình, ai liên hệ
- **Nội quy công ty** — điều được làm, điều cấm

**CLAUDE.md chính là tất cả những thứ đó** gói lại trong 1 file.

Bạn viết gì trong CLAUDE.md → AI sẽ làm **chính xác** như vậy.
Bạn quên viết gì → AI sẽ tự quyết (và có thể sai).

## CLAUDE.md nằm ở đâu?

```
my-project/              <-- gốc dự án
├── CLAUDE.md            <-- FILE NÀY!
├── .claude/             <-- Starter Kit
├── src/                 <-- code
└── ...
```

- Nằm ở **gốc (root)** của dự án
- Claude Code **tự động đọc** file này khi bắt đầu mỗi conversation
- Không cần bảo AI đọc — nó luôn đọc
- Tên file phải **chính xác**: `CLAUDE.md` (viết hoa)

## Nội dung CLAUDE.md gồm những gì?

Một CLAUDE.md tốt thường có **8 phần** chính:

| # | Phần | Ví von | Mục đích |
|---|------|--------|----------|
| 1 | Communication Language | Ngôn ngữ giao tiếp | AI nói tiếng gì? |
| 2 | Project Overview | Mô tả dự án | AI hiểu dự án làm gì |
| 3 | Development Workflow | Quy trình làm việc | AI làm theo thứ tự nào |
| 4 | Scripts | Công cụ bắt buộc | AI dùng tool nào |
| 5 | Git Workflow | Quy trình quản lý code | AI commit, push thế nào |
| 6 | Business Logic Docs | Tài liệu nghiệp vụ | AI tìm quy tắc ở đâu |
| 7 | Living Documents | Docs phải cập nhật | AI update doc khi nào |
| 8 | Skills Reference | Danh sách skills | AI biết có những kỹ năng nào |

Hãy đi qua từng phần.

### Phần 1: Communication Language

Đây là điều **đầu tiên** trong CLAUDE.md — bảo AI nói tiếng gì.

```markdown
## CRITICAL: Communication Language

**ALWAYS communicate in Vietnamese (tiếng Việt)**
- All responses, explanations should be in Vietnamese
- Code comments can be in English
- Commit messages should be in English (git convention)
```

Nếu bạn không viết phần này, AI sẽ trả lời bằng tiếng Anh (mặc định).

### Phần 2: Project Overview

Mô tả ngắn gọn dự án để AI hiểu bối cảnh:

```markdown
## Project Overview

**My App** — Ứng dụng quản lý lớp học online
- Backend: Java Spring Boot
- Frontend: React + TypeScript
- Database: PostgreSQL
```

AI cần biết dự án làm gì để đưa ra giải pháp phù hợp. Ví dụ: dự án giáo dục khác hoàn toàn với dự án thương mại điện tử.

### Phần 3: Development Workflow

Phần **quan trọng nhất** — quyết định AI làm việc thế nào:

```markdown
## Quy trình bắt buộc cho mỗi PR:

1. Quick Brainstorm (suy nghĩ trước)
2. Task Breakdown (chia nhỏ công việc)
3. TDD - Test First (viết test trước code)
4. Implementation (viết code)
5. Code Review (tự kiểm tra)

### KHÔNG được bỏ qua:
- Nhảy thẳng vào code mà không brainstorm
- Viết code trước tests
- Commit mà không có tests đi kèm
```

Khi có phần này, AI sẽ **tuân thủ** — nếu bạn nhờ tạo tính năng mới, AI sẽ brainstorm trước, chia nhỏ, viết test, rồi mới code.

### Phần 4: Scripts

Bắt buộc AI dùng scripts thay vì gõ lệnh trực tiếp:

```markdown
## CRITICAL: Docker Scripts Required

KHÔNG BAO GIỜ chạy lệnh Docker trực tiếp. LUÔN LUÔN dùng scripts.

# SAI:
docker-compose up -d

# ĐÚNG:
./scripts/up.sh
```

Tại sao? Vì scripts đã được kiểm tra, chuẩn hóa, và xử lý edge cases.

### Phần 5: Git Workflow

Quy tắc quản lý code (Git):

```markdown
## Git Workflow

- ALWAYS create feature branch before changes
- NEVER commit directly to main
- Branch naming: feature/KC-XXX-description
```

Điều này đảm bảo AI không vô tình sửa code trên nhánh chính (main), gây hỏng dự án.

### Phần 6-8: Business Docs, Living Docs, Skills Reference

**Business Logic Docs:** Chỉ AI nơi tìm quy tắc nghiệp vụ
```markdown
Location: documents/01-business/ — SOURCE OF TRUTH cho business rules
```

**Living Documents:** Docs nào phải cập nhật liên tục
```markdown
| Doc | Update khi |
|-----|------------|
| README.md | Thêm/xóa folder, service |
| CLAUDE.md | Thay đổi quy trình, convention |
```

**Skills Reference:** Liệt kê tất cả skills có sẵn
```markdown
### Core Skills:
- core/brainstorming-methodology.md
- core/task-breakdown-guide.md
- core/tdd-enforcement.md
...
```

````bash
# Xem một CLAUDE.md mẫu đơn giản
# Đây là phiên bản rút gọn cho dự án nhỏ

cat << 'SAMPLE'
# Claude Code Instructions

## Communication Language
**ALWAYS communicate in Vietnamese (tieng Viet)**

## Project Overview
**My Blog** - Blog ca nhan bang Next.js + TypeScript
- Frontend: Next.js 14, Tailwind CSS
- Database: SQLite (via Prisma)
- Deploy: Vercel

## Development Workflow
1. Brainstorm truoc khi code
2. Viet test truoc
3. Code
4. Tu review truoc khi commit

## Git Workflow
- Tao feature branch truoc moi thay doi
- KHONG commit truc tiep vao main
- Branch naming: feature/description

## Folder Structure
my-blog/
├── src/
│   ├── app/        # Next.js pages
│   ├── components/ # React components
│   └── lib/        # Utilities
├── prisma/         # Database schema
└── public/         # Static files
SAMPLE

echo ""
echo "--- Day la CLAUDE.md don gian cho du an nho ---"
echo "Du an lon se co nhieu muc hon (business docs, scripts, skills...)"
````

## Tại sao CLAUDE.md là "Living Document"?

CLAUDE.md **không phải viết 1 lần rồi quên**. Nó phải sống cùng dự án:

| Khi nào update | Ví dụ |
|---------------|-------|
| Thêm folder mới | Thêm `api/` → update Folder Structure |
| Thêm công nghệ mới | Dùng Redis → update Project Overview |
| Thay đổi quy trình | Bỏ TDD → update Development Workflow |
| Thêm script mới | Tạo `deploy.sh` → update Scripts section |

**Hậu quả nếu không update:**
- AI đọc CLAUDE.md cũ → vẫn làm theo cách cũ
- Folder mới không có trong Folder Structure → AI không biết nó tồn tại
- Script mới không được liệt kê → AI không dùng

## Mẹo: Dùng từ khóa "CRITICAL" và "NEVER"

Claude Code đặc biệt chú ý đến những từ mạnh:

| Từ khóa | Ý nghĩa | Khi nào dùng |
|---------|---------|-------------|
| `CRITICAL` | Cực kỳ quan trọng | Quy tắc không được vi phạm |
| `NEVER` / `KHÔNG BAO GIỜ` | Cấm tuyệt đối | Hành động gây hại |
| `ALWAYS` / `LUÔN LUÔN` | Bắt buộc | Hành động phải làm |
| `IMPORTANT` | Quan trọng | Dễ quên nhưng cần nhớ |

```markdown
## CRITICAL: Docker Scripts Required
KHÔNG BAO GIỜ chạy lệnh Docker trực tiếp.
LUÔN LUÔN dùng scripts.
```

Khi bạn viết "CRITICAL", AI sẽ ưu tiên tuân thủ phần đó hơn các phần khác.

````bash
# Kiểm tra CLAUDE.md trong dự án

echo "Kiểm tra CLAUDE.md:"
echo ""

if [ -f "CLAUDE.md" ]; then
    LINES=$(wc -l < CLAUDE.md)
    echo "[OK] CLAUDE.md tồn tại ($LINES dòng)"
    echo ""
    echo "Các section chính:"
    grep '^## ' CLAUDE.md | head -15
else
    echo "[X] CLAUDE.md KHÔNG tồn tại!"
    echo "    Hãy tạo file này ở gốc dự án."
    echo "    Template có sẵn tại: .claude/templates/CLAUDE.md.template"
fi
````

## Template có sẵn

Starter Kit cung cấp template CLAUDE.md để bạn không phải viết từ đầu:

1. Copy template: `cp .claude/templates/CLAUDE.md.template ./CLAUDE.md`
2. Mở file, điền thông tin dự án vào các chỗ `[placeholder]`
3. Xóa các section không cần (dự án nhỏ có thể bỏ bớt)
4. Commit file vào git

Template đã có đầy đủ cấu trúc — bạn chỉ cần **điền nội dung**.

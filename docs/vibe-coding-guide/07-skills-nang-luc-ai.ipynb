## Skills là gì?

Hãy nghĩ về **chứng chỉ nghề**.

Một người tốt nghiệp đại học biết nhiều thứ chung chung. Nhưng khi vào công ty:
- Có **chứng chỉ kế toán** → biết cách làm sổ sách cho công ty đó
- Có **chứng chỉ an toàn lao động** → biết quy tắc an toàn cụ thể
- Có **chứng chỉ vận hành máy** → biết dùng máy móc đúng cách

**Skills trong Kit cũng vậy:**
- Claude Code đã biết cách lập trình (kiến thức chung)
- Nhưng skills dạy nó cách làm việc **đúng cho dự án cụ thể**
- Mỗi skill = 1 file `.md` mô tả **quy trình** + **bẫy cần tránh** (gotchas)

## Skills KHÔNG phải code

Điều quan trọng: skills là các **file văn bản** (Markdown), không phải code.

Mỗi skill chứa:
- **Mô tả** (description): AI đọc phần này để biết khi nào dùng skill
- **Quy trình** (process): Các bước cần làm
- **Gotchas** (bẫy): Những lỗi thường gặp + cách tránh
- **Tham chiếu** (reference): Link đến tài liệu chi tiết hơn

AI đọc skill → hiểu quy trình → làm theo. Đơn giản vậy thôi.

## Skill Activation — AI tự biết dùng khi nào

Bạn **không cần** bảo AI "hãy dùng skill X". AI tự biết nhờ **description** (mô tả) trong mỗi skill.

Ví dụ, skill quality-audit có description:
```
Use when user says 'audit', 'quality check', 'ready to merge?',
'kiểm tra chất lượng', or before merging to main.
```

Nên khi bạn nói:
- "Kiểm tra chất lượng" → AI tự kích hoạt quality-audit
- "Ready to merge?" → AI cũng kích hoạt quality-audit
- "Tạo tính năng mới" → AI kích hoạt brainstorm + breakdown + tdd

Description viết càng tốt → AI nhận diện càng chính xác.

## Gotchas — Phần GIÁ TRỊ NHẤT trong mỗi skill

"Gotcha" = **bẫy**, điều bất ngờ, lỗi dễ mắc phải.

Đây là phần quan trọng nhất vì:
- Claude Code đã biết lập trình (kiến thức chung)
- Nhưng nó **không biết** bẫy riêng của dự án bạn
- Gotchas = bài học từ kinh nghiệm thực tế

Ví dụ gotchas trong một dự án thực:
```markdown
## Gotchas
- KHÔNG gọi .instanceId() trên builder — BaseEntity tự set
- Test controller PHẢI có @ActiveProfiles("test") — thiếu = lỗi 403
- Đếm ngày trong test — tự đếm tay, KHÔNG ước lượng
```

Mỗi gotcha = **1 lần đã sai** → ghi lại → AI không sai lần 2.

````bash
# Xem nội dung mẫu một skill đơn giản

cat << 'SKILL'
---
name: brainstorming-methodology
description: >
  Use BEFORE starting any new feature, PR, or significant code change.
  Triggered by: 'new feature', 'tao tinh nang', 'start PR', 'bat dau'.
---

# Brainstorming Methodology

## Process (5-10 phut)
1. Phan tich scope: feature nay lam gi?
2. Xac dinh risks: co the sai o dau?
3. Liet ke dependencies: can gi tu ben ngoai?
4. Edge cases: tinh huong bat thuong?
5. Output: danh sach tasks cu the

## Gotchas
- KHONG brainstorm qua 10 phut — vua du, khong can hoan hao
- KHONG bat dau code khi chua co task list
- Edge cases thuong bi quen: empty input, concurrent access, timezone

## Reference
- Xem chi tiet: quick-reference/brainstorming-trade-off-matrix.md
SKILL

echo ""
echo "--- Nhan xet: Skill chi ~20 dong, ngan gon, du y ---"
````

## Progressive Disclosure — Load theo nhu cầu

Một nguyên tắc thiết kế quan trọng:

| Layer | Khi nào load | Chi phí |
|-------|-------------|--------|
| **Tên + mô tả** | Luôn luôn (khi khởi động) | Rất ít |
| **Nội dung SKILL.md** | Khi skill được kích hoạt | Trung bình |
| **Reference files** | Khi AI cần tra cứu chi tiết | Chỉ khi cần |

Giống như bạn đọc **mục lục sách** trước (nhanh), rồi mới mở đến **chương cần đọc** (chi tiết).

Điều này quan trọng vì Claude Code có giới hạn "bộ nhớ làm việc" (context window). Load ít = suy nghĩ nhanh hơn, chính xác hơn.

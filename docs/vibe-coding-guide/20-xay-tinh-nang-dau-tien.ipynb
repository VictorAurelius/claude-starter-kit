# 20. Xây Tính Năng Đầu Tiên — Từ Ý Tưởng Đến PR Merge

**Phần 5: Thực hành từ đầu đến cuối**

---

Bạn đã setup dự án xong (bài 19). Giờ đến phần thú vị nhất — **xây thứ gì đó thật sự**.

Chúng ta sẽ đi qua toàn bộ quy trình Superpowers, từ ý tưởng ban đầu đến tính năng hoàn chỉnh, sẵn sàng merge. Và bạn sẽ thấy: **bạn không cần viết một dòng code nào**.

Ví dụ thực tế: *"Tôi muốn trang Landing Page cho sản phẩm của mình."*

## Step 1: Brainstorm (5-10 phút)

Trước khi xây nhà, bạn cần bản vẽ. Trước khi code, bạn cần **brainstorm**.

### Bạn nói với Claude:

```
Tôi muốn tạo landing page giới thiệu sản phẩm cà phê handmade của tôi.
Hãy brainstorm giúp tôi.
```

### Claude sẽ hỏi lại (và bạn trả lời):

| Claude hỏi | Bạn trả lời |
|-----------|-------------|
| Target audience là ai? | Dân văn phòng 25-35 tuổi, thích cà phê specialty |
| Cần những section nào? | Hero, giới thiệu sản phẩm, đánh giá khách hàng, liên hệ |
| Call-to-action chính? | Nút "Đặt hàng ngay" dẫn về Shopee |
| Responsive (hiển thị trên điện thoại)? | Có, phải đẹp trên điện thoại |

Claude sẽ ghi lại mọi quyết định vào **Decision Log** — một bản tóm tắt để sau này ai cũng hiểu tại sao làm vậy.

````bash
# Trong thực tế, bạn nói chuyện với Claude trong VS Code.
# Đây là ví dụ prompt bạn sẽ gõ:

cat << 'EOF'
=== PROMPT BAN GOI CHO CLAUDE ===

Toi muon tao landing page gioi thieu san pham ca phe handmade.
- Target: dan van phong 25-35 tuoi
- Sections: Hero, san pham, danh gia khach hang, lien he
- CTA chinh: nut "Dat hang ngay" dan ve Shopee
- Phai dep tren dien thoai

Hay brainstorm giup toi, sau do tao task breakdown.
EOF
````

## Step 2: Task Breakdown

Sau brainstorm, Claude sẽ chia công việc thành các **task nhỏ**. Giống như xây nhà: không ai xây cả căn nhà cùng lúc — bạn đổ móng trước, rồi xây tường, rồi lợp mái.

### Claude đề xuất task breakdown:

| # | Task | Thời gian ước tính |
|---|------|-------------------|
| 1 | Setup Next.js project + Tailwind CSS | ~5 phút |
| 2 | Header (logo + navigation) | ~5 phút |
| 3 | Hero section (ảnh lớn + tagline + CTA) | ~10 phút |
| 4 | Product showcase (3-4 sản phẩm) | ~10 phút |
| 5 | Testimonials (đánh giá khách hàng) | ~5 phút |
| 6 | Footer (thông tin liên hệ) | ~5 phút |
| 7 | Responsive design | ~10 phút |
| 8 | Tests | ~5 phút |

**Tổng: ~55 phút** (thay vì 2-3 ngày nếu tự code)

Bạn xem qua, đồng ý hoặc điều chỉnh: *"Bỏ Testimonials đi, chưa có khách hàng nào"* — Claude sẽ cập nhật ngay.

## Step 3: TDD — Viết Test Trước

Nghe có vẻ ngược đời: viết test trước khi có code? Nhưng nghĩ thế này — bạn viết **danh sách kiểm tra** trước khi kiểm tra nhà mới:

- Tường có thẳng không?
- Điện có sáng không?
- Nước có chảy không?

Viết danh sách trước = bạn biết **tiêu chuẩn** cần đạt.

### Claude viết test:
```
- Landing page render được (không trang trắng)
- Hero section hiển thị tagline
- Nút CTA dẫn đúng link Shopee
- Responsive: trên mobile không bị vỡ layout
```

````bash
# Chạy test lần đầu — TẤT CẢ phải FAIL (RED)
# Đây là điều TỐT — nghĩa là test đang hoạt động

cat << 'EOF'
=== KET QUA CHAY TEST (vi du) ===

FAIL  src/__tests__/LandingPage.test.tsx
  x Landing page renders correctly
  x Hero section displays tagline
  x CTA button links to Shopee
  x Layout is responsive on mobile

Tests: 4 failed, 0 passed

=> Dung roi! Test FAIL vi chua co code.
   Buoc tiep theo: Claude se viet code de test PASS.
EOF
````

## Step 4: Implementation — Claude Code Bắt Tay Vào Làm

Đây là lúc phép màu xảy ra. Claude bắt đầu viết code cho từng task.

### Bạn làm gì trong lúc này?

Bạn là **đạo diễn**, Claude là **quay phim**.

- Claude code xong task 1 (Header) → bạn mở browser xem
- Bạn thấy logo nhỏ quá → nói: *"Logo to hơn, khoảng gấp đôi"*
- Claude sửa ngay
- Task 2 (Hero) → bạn xem → *"Màu nền đổi sang nâu đậm cho hợp cà phê"*
- Claude sửa
- Cứ thế cho đến khi hài lòng

````bash
# Trong lúc Claude code, bạn mở browser xem kết quả
# Claude sẽ chạy dev server cho bạn:

cat << 'EOF'
=== CLAUDE CHAY DEV SERVER ===

$ npm run dev

  > ready - started server on 0.0.0.0:3000
  > url   - http://localhost:3000

=> Mo browser, vao http://localhost:3000 de xem ket qua
=> Moi khi Claude sua code, trang tu dong cap nhat (hot reload)
EOF
````

````bash
# Ví dụ các yêu cầu điều chỉnh bạn nói với Claude:

cat << 'EOF'
=== CAC YEU CAU DIEU CHINH (vi du) ===

Ban: "Chu to hon, font dam hon"
Claude: [sua font-size va font-weight]

Ban: "Them hinh anh ca phe o Hero section"
Claude: [them img tag voi placeholder image]

Ban: "Doi mau nut CTA sang do cam"
Claude: [sua CSS button color]

Ban: "Tren dien thoai, cac san pham nen xep doc, khong xep ngang"
Claude: [sua responsive CSS cho mobile]

=> Ban KHONG can biet CSS, HTML, hay bat ky ngon ngu nao.
   Chi can MO TA nhung gi ban muon thay.
EOF
````

## Step 5: Review — Claude Tự Kiểm Tra

Trước khi "giao hàng", Claude tự review code 2 vòng:

**Vòng 1 — Quick scan:**
- Code có sạch không?
- Tên biến có rõ ràng không?
- Có gì bị trùng lặp không?

**Vòng 2 — Deep review:**
- Logic có đúng không?
- Responsive có hoạt động không?
- Performance có OK không?
- Security có vấn đề gì không?

````bash
# Kết quả review (ví dụ):

cat << 'EOF'
=== KET QUA REVIEW ===

Stage 1 - Quick Scan:
  [OK] Code sach, khong trung lap
  [OK] Ten bien ro rang

Stage 2 - Deep Review:
  [OK] Logic dung
  [OK] Responsive OK tren mobile va tablet
  [MINOR] Nen them alt text cho hinh anh (accessibility)

Ket qua: 0 Critical, 0 Major, 1 Minor
=> San sang de commit!
EOF
````

## Step 6: Git Workflow — Lưu và Chia Sẻ

Code xong, review xong — giờ lưu lại và tạo Pull Request (PR) để merge vào dự án chính.

Nhắc lại: PR giống như "gửi bài tập cho thầy duyệt" — bạn không sửa trực tiếp vào "bản chính", mà gửi riêng để kiểm tra trước.

````bash
# Buoc 1: Tao branch moi ("nhanh rieng" de lam viec)
git checkout -b feature/landing-page

echo "Da tao branch feature/landing-page"
````

````bash
# Buoc 2: Luu thay doi vao Git
git add .
git commit -m "feat: add landing page with hero and features sections"

echo "Da commit!"
````

````bash
# Buoc 3: Day len GitHub va tao PR
git push -u origin feature/landing-page

# Tao Pull Request
gh pr create --base main --title "feat: add landing page" --body "Landing page voi Hero, Products, va Footer sections. Responsive tren mobile."

echo "Da tao PR!"
````

## Step 7: Quality Check — Kiểm Tra Chất Lượng Cuối Cùng

Trước khi merge, chạy Quality Audit để chắc chắn mọi thứ đạt chuẩn.

````bash
# Trong Claude Code, ban goi:
# /quality-audit

cat << 'EOF'
=== KET QUA QUALITY AUDIT (vi du) ===

Category                  Score
-------------------------+------
Code Quality              9/10
Test Coverage             8/10
Documentation             7/10
Security                  10/10
Performance               9/10
Accessibility             7/10
Git Workflow              10/10
Business Logic            8/10
Error Handling            9/10
CI/CD                     8/10
-------------------------+------
TOTAL                     85/100 (B+)

=> Dat chuan de merge!
EOF
````

````bash
# Merge PR (sau khi quality check dat)
# gh pr merge <so-PR> --squash --delete-branch

cat << 'EOF'
=== MERGE PR ===

$ gh pr merge 1 --squash --delete-branch

Merging pull request #1 (feat: add landing page)
Squash and merge
Deleted branch feature/landing-page

=> Tinh nang da vao main! Xong!
EOF
````

## Bai Hoc Quan Trong Nhat

Trong toàn bộ quá trình trên, bạn đã:

| Bạn làm | Claude làm |
|---------|----------|
| Mô tả ý tưởng | Brainstorm chi tiết |
| Trả lời câu hỏi | Task breakdown |
| Xem preview, góp ý | Viết test + code |
| Duyệt kết quả | Review + quality check |
| Approve merge | Git workflow |

**Bạn KHÔNG viết một dòng code nào.** Bạn là **đạo diễn**, Claude là **quay phim**.

Vai trò của bạn quan trọng hơn bạn nghĩ — không ai biết sản phẩm cần gì bằng chính bạn. Claude chỉ là công cụ giúp hiện thực hóa tầm nhìn của bạn.

## Tóm Tắt Quy Trình

```
Y tuong
  |-- Brainstorm (5-10 phut)
  |-- Task Breakdown (5-10 phut)
  |-- TDD: Viet test truoc (RED)
  |-- Implementation: Claude code (GREEN)
  |-- Review: 2 vong kiem tra
  |-- Git: branch -> commit -> push -> PR
  |-- Quality Check: >= 85 diem
  |-- Merge!
```

## Pre-flight Check la gi?

Hay tuong tuong ban la **phi cong**.

Truoc moi chuyen bay, ban kiem tra:
- Nhien lieu du chua?
- Dong co hoat dong tot?
- Thoi tiet co an toan?
- Hanh khach da len het?

**Khong ai cam thay phien** khi phi cong lam viec nay. Vi mot lan bo qua = tham hoa.

Trong phan mem cung vay:
- **Truoc moi thay doi code** (PR) → kiem tra
- **Truoc moi module moi** → kiem tra ky hon
- **Dinh ky** → kiem tra toan bo du an

Pre-flight Check co **3 tang** (layer), tuy thoi diem ma chay tang nao.

## 3 Tang Kiem Tra

| Tang | Khi nao | Vi von | Mat bao lau |
|------|---------|--------|-------------|
| **Layer 1: PR** | Truoc MOI lan thay doi code | Kiem tra hang ngay truoc khi ra khoi nha | 5-10 phut |
| **Layer 2: Domain** | Truoc khi bat dau MODULE MOI | Kiem tra khi mo cua hang moi | 15-30 phut |
| **Layer 3: Project** | Moi 10-25 PRs | Kiem tra tong hang quy | 1-2 gio |

Cang len cao, cang toan dien nhung cung cang ton thoi gian. Vi vay:
- Layer 1: Chay **thuong xuyen** (moi PR)
- Layer 2: Chay **khi can** (module moi)
- Layer 3: Chay **dinh ky** (milestone, review tong)

## Layer 1: PR Pre-flight — Kiem tra truoc MOI thay doi

Nhu truoc khi ra khoi nha, ban kiem tra:
- Co chìa khoa chua? (business logic)
- Mac do phu hop chua? (code quality)
- Co vi tien khong? (testing)

### 1A. Business Logic Check

Kiem tra: code co **dung theo quy tac kinh doanh** khong?

| Hang muc | Y nghia | Vi du |
|---------|---------|-------|
| Quy tac tu config? | Khong hardcode con so trong code | `max-per-class = 50` tu config, khong viet "50" truc tiep |
| Email template ton tai? | Code gui email thi template phai co | Gui email xac nhan → template `enrollment-confirm.html` co chua? |
| Status transitions dung? | Trang thai chuyen dung thu tu | PENDING → CONFIRMED OK, nhung COMPLETED → PENDING thi SAI |
| Business docs da update? | Doi logic = doi docs | Them quy tac moi → rules.md da ghi chua? |

**Loi thuong gap:** Thay doi logic nhung quen update rules.md → lan sau AI doc docs cu → code sai.

### 1B. Code Quality Check

Kiem tra: code co **sach se, chuan muc** khong?

| Hang muc | Y nghia | Vi du |
|---------|---------|-------|
| Validation DTO day du? | Du lieu vao phai kiem tra | `studentId` khong duoc null, `classId` phai > 0 |
| Constants co configurable? | Hang so nen thay doi duoc | Thay vi `MAX_RETRY = 3` cung, dung config `retry.max-attempts` |
| Khong co TODO/FIXME? | Code xong phai xong, khong de "lam sau" | Tim tat ca `TODO:` va xu ly truoc khi PR |
| Dat ten ro rang? | Doc ten biet chuc nang | `checkSlotAvailability()` tot hon `check()` |

### 1C. Testing Check

Kiem tra: co **test kiem chung** khong?

| Hang muc | Y nghia | Vi du |
|---------|---------|-------|
| Business rules co test? | Moi quy tac phai co test | BR-ENR-001 (max 50) → test voi 50, 51 hoc sinh |
| Multi-tenant tests? | Tenant A khong thay du lieu tenant B | Tao 2 truong, kiem tra du lieu khong xen lan |
| Edge cases? | Truong hop dac biet | Dang ky vao phut cuoi truoc deadline, lop con dung 1 slot |

````bash
# Demo: Chay Layer 1 Pre-flight Check
echo "=== Layer 1: PR Pre-flight Check ==="
echo ""
echo "Cach chay (noi voi AI trong Claude Code):"
echo ""
echo '  /pre-flight-check pr'
echo ""
echo "AI se tu dong kiem tra:"
echo ""
echo "Business Logic:"
echo "  [PASS] Cac quy tac deu tu config key"
echo "  [PASS] Email templates ton tai"
echo "  [WARN] Status transition COMPLETED->PENDING chua co trong rules.md"
echo "  [PASS] Business docs da update"
echo ""
echo "Code Quality:"
echo "  [PASS] Validation DTO day du"
echo "  [PASS] Constants configurable"
echo "  [FAIL] Tim thay 2 TODO comments chua xu ly"
echo "  [PASS] Naming conventions"
echo ""
echo "Testing:"
echo "  [PASS] Business rules co test"
echo "  [PASS] Multi-tenant isolation"
echo "  [WARN] Edge case: deadline boundary chua co test"
echo ""
echo "Ket qua: 2 WARN, 1 FAIL → Can xu ly truoc khi merge"
````

## Layer 2: Domain Pre-flight — Kiem tra truoc MODULE MOI

Nhu khi ban **mo mot cua hang moi**, ban can kiem tra ky hon:
- Giay phep kinh doanh co chua?
- Nhan vien da dao tao chua?
- Quy trinh van hanh da viet chua?
- Menu da in chua?

Trong phan mem, truoc khi bat dau 1 domain/module moi, kiem tra:

### Checklist Domain Pre-flight

| Hang muc | Cau hoi | Vi du |
|---------|---------|-------|
| 3 files business docs | Da tao day du chua? | `enrollment/rules.md`, `use-cases.md`, `api-contract.md` |
| User journey | Hanh trinh nguoi dung ro rang? | Phu huynh: tim lop → xem chi tiet → dang ky → thanh toan |
| API contract | Day du endpoints? | GET (danh sach), POST (tao), PUT (sua), DELETE (huy) |
| Status lifecycle | Trang thai chuyen doi dung? | PENDING → CONFIRMED → COMPLETED; PENDING → CANCELLED |
| Error codes | Tat ca loi da dinh nghia? | CLASS_FULL, SCHEDULE_CONFLICT, PAYMENT_REQUIRED... |
| Dependencies | Phu thuoc module nao? | Enrollment phu thuoc Class (co lop moi dang ky duoc) |

### Tai sao phai check truoc khi code?

Vi **sua docs re hon sua code rat nhieu**:
- Phat hien thieu 1 error code trong docs → 5 phut sua
- Phat hien thieu sau khi code xong → co the mat nua ngay refactor

````bash
# Demo: Chay Layer 2 Domain Pre-flight
echo "=== Layer 2: Domain Pre-flight Check ==="
echo ""
echo "Cach chay:"
echo ""
echo '  /pre-flight-check domain enrollment'
echo ""
echo "AI se kiem tra:"
echo ""
echo "Business Docs:"
echo "  [PASS] rules.md ton tai (5 rules)"
echo "  [PASS] use-cases.md ton tai (3 use cases)"
echo "  [PASS] api-contract.md ton tai (4 endpoints)"
echo ""
echo "User Journey:"
echo "  [PASS] Happy path ro rang (UC-ENR-001)"
echo "  [WARN] Chua co use case cho ADMIN huy dang ky"
echo ""
echo "API Contract:"
echo "  [PASS] GET /api/v1/enrollments"
echo "  [PASS] POST /api/v1/enrollments"
echo "  [FAIL] Thieu DELETE /api/v1/enrollments/{id}"
echo "  [WARN] Thieu PATCH cho update status"
echo ""
echo "Status Lifecycle:"
echo "  [PASS] PENDING -> CONFIRMED -> COMPLETED"
echo "  [PASS] PENDING -> CANCELLED"
echo "  [WARN] Thieu transition CONFIRMED -> DROPPED"
echo ""
echo "Dependencies:"
echo "  [PASS] Class module da co (PR #15)"
echo "  [PASS] Student module da co (PR #8)"
echo "  [WARN] Payment module chua implement"
echo ""
echo "Ket qua: 1 FAIL, 4 WARN → Sua docs truoc khi bat dau code!"
````

## Layer 3: Project Pre-flight — Kiem tra tong dinh ky

Nhu **kiem tra tong hang quy** cua cong ty:
- Tat ca phong ban van hanh dung?
- Tai chinh on dinh?
- Nhan su du?
- Phap ly tuan thu?

Layer 3 chay **moi 10-25 PRs** hoac truoc cac milestone lon (release, demo, deploy).

### Checklist Project Pre-flight

| Hang muc | Mo ta | Vi du |
|---------|-------|-------|
| **Cross-domain consistency** | Cac module co nhat quan? | Enrollment va Grading deu dung chung Student model? |
| **Business gap check** | Co lo hong nghiep vu? | Module Payment da co nhung chua ket noi voi Enrollment? |
| **Quality audit** | Chat luong code tong the? | Cham diem /100, xem module nao yeu nhat |
| **Docs sync** | Tai lieu con dung? | README phan anh dung cau truc hien tai? |
| **Security review** | Bao mat on? | Token het han xu ly dung? Du lieu nhay duoc ma hoa? |
| **Deployment readiness** | San sang deploy? | Docker images build duoc? Database migration chuan bi? |

Day la tang **toan dien nhat** — kiem tra moi thu tu nghiep vu den ky thuat den van hanh.

````bash
# Demo: Chay Layer 3 Project Pre-flight
echo "=== Layer 3: Project Pre-flight Check ==="
echo ""
echo "Cach chay:"
echo ""
echo '  /pre-flight-check project'
echo ""
echo "AI se kiem tra TOAN BO du an:"
echo ""
echo "Cross-domain Consistency:"
echo "  [PASS] Student model nhat quan giua 4 modules"
echo "  [WARN] Class status enum khac nhau giua Enrollment va Attendance"
echo ""
echo "Business Gap Check:"
echo "  [PASS] 8/10 domains co day du 3 files"
echo "  [FAIL] Domain 'notification' thieu use-cases.md"
echo "  [FAIL] Domain 'reporting' thieu ca 3 files"
echo ""
echo "Quality Audit:"
echo "  Overall score: 78/100"
echo "  Weakest: Frontend testing (12/20)"
echo "  Strongest: API design (18/20)"
echo ""
echo "Docs Sync:"
echo "  [WARN] README.md thieu mention module Payment (moi them PR #32)"
echo "  [PASS] CLAUDE.md up-to-date"
echo ""
echo "Security:"
echo "  [PASS] JWT token validation"
echo "  [WARN] Rate limiting chua implement cho enrollment API"
echo ""
echo "Deployment:"
echo "  [PASS] Docker images build OK"
echo "  [PASS] Database migrations san sang"
echo ""
echo "Ket qua: 2 FAIL, 4 WARN — Tao action items de xu ly!"
````

### Truoc moi PR:
```
/pre-flight-check pr
```

### Truoc module moi:
```
/pre-flight-check domain enrollment
```

### Kiem tra tong:
```
/pre-flight-check project
```

AI se:
1. Tu dong chay tat ca cac muc kiem tra
2. Bao cao ket qua: PASS / WARN / FAIL
3. Goi y cach sua cho moi muc FAIL

**Viec cua ban:** Doc bao cao, quyet dinh co merge khong. Day la quyen cua **nguoi so huu san pham** — khong phai cua AI.

## Tong ket bang vi von

| Tang | Vi von doi thuc | Tan suat |
|------|----------------|----------|
| **Layer 1: PR** | Kiem tra truoc khi ra khoi nha moi ngay | Moi PR (hang ngay) |
| **Layer 2: Domain** | Kiem tra khi mo cua hang/chi nhanh moi | Moi module moi |
| **Layer 3: Project** | Kiem tra tong / kiem toan hang quy | Moi 10-25 PRs |

Cang len cao:
- Cang **toan dien** (kiem tra nhieu thu hon)
- Cang **ton thoi gian** (nhung dang de lam)
- Cang **it chay** (khong can hang ngay)

## Khi nao chay tang nao?

### Tinh huong thuc te:

**Ban vua hoan thanh 1 thay doi nho** (sua loi, them validation):
→ Chay `/pre-flight-check pr`

**Ban sap bat dau lam tinh nang hoan toan moi** (module thanh toan, module diem danh):
→ Chay `/pre-flight-check domain payment`

**Ban da lam 15 PRs va sap demo cho khach hang:**
→ Chay `/pre-flight-check project`

**Ban khong chac chac chay tang nao:**
→ Chay `/pre-flight-check pr` truoc. Neu co nhieu WARN/FAIL, xem xet chay tang cao hon.

### Quy tac don gian:
- Khong bao gio bo qua Layer 1 (PR check)
- Layer 2 bat buoc truoc module moi
- Layer 3 it nhat 1 lan truoc moi milestone

````bash
# Tom tat cac lenh Pre-flight Check
echo "=== Quick Reference: Pre-flight Check Commands ==="
echo ""
echo "Layer 1 — Truoc moi PR:"
echo '  /pre-flight-check pr'
echo ""
echo "Layer 2 — Truoc module moi:"
echo '  /pre-flight-check domain <ten-domain>'
echo '  /pre-flight-check domain enrollment'
echo '  /pre-flight-check domain attendance'
echo '  /pre-flight-check domain grading'
echo ""
echo "Layer 3 — Kiem tra tong:"
echo '  /pre-flight-check project'
echo ""
echo "Meo: Bat dau tu Layer 1 neu khong chac. AI se goi y neu can chay cao hon."
````

## Vi du thuc te: Phat hien van de nho Pre-flight

### Tinh huong:
Ban vua nho AI implement tinh nang "Dang ky hoc". AI bao "xong roi!".

### Khong co Pre-flight:
Ban merge code → deploy → phu huynh dang ky 2 lop cung gio → **he thong khong chan** → bug!

### Co Pre-flight:
```
Ban: /pre-flight-check pr

AI: [FAIL] BR-ENR-003 (khong trung lich) chua co test
    [FAIL] api-contract.md thieu error SCHEDULE_CONFLICT  
    [WARN] use-cases.md khong de cap truong hop trung lich
    
    Goi y: Bo sung validation trung lich trong EnrollmentService
```

Ban phat hien van de **truoc khi deploy**, tiet kiem hang gio fix bug va xin loi khach hang.

## Nhung van de thuong phat hien

Sau nhieu lan chay Pre-flight, day la nhung loi **hay gap nhat**:

| Van de | Tang | Tan suat | Muc do |
|--------|------|----------|--------|
| Business docs chua update | L1 | Rat thuong | Cao |
| Thieu test cho edge case | L1 | Thuong | Trung binh |
| Hardcode con so thay vi config | L1 | Thuong | Trung binh |
| Thieu error code trong api-contract | L2 | Kha thuong | Cao |
| Status lifecycle khong day du | L2 | Kha thuong | Cao |
| Module A va B dung data model khac | L3 | It | Rat cao |
| README loi thoi | L3 | Thuong | Thap |

Nhan ra pattern: van de **nghiep vu** (business) thuong nghiem trong hon van de **ky thuat** (technical).
Va nghiep vu la thu **ban** — nguoi hieu san pham — phat hien tot nhat.

## Business Gap Check la gi?

Hay tuong tuong ban la **thanh tra thuc pham**.

Ban den nha hang kiem tra:
- Bep co sach khong? (code quality — AI lam tot)
- Mon an ngon khong? (user experience — AI kha duoc)
- Nhung quan trong nhat: **quy trinh co day du khong?**
  - Nguyen lieu nhap ve co kiem tra khong?
  - Thuc an het han co quy trinh xu ly khong?
  - Khach bi di ung co cach xu ly khong?
  - Nhan vien co biet cach xu ly khi mat dien khong?

Nhung cau hoi nay **khong phai ve ky thuat nau an** (code). Ma la ve **quy trinh van hanh** (business logic).

**Business Gap Check** = tim nhung buoc, quy tac, truong hop ma **da bi quen** hoac **chua duoc implement**.

## Khac gi Quality Audit?

Hai cong cu nay bo sung cho nhau, nhung **kiem tra khac thu**:

| | Quality Audit | Business Gap Check |
|---|---|---|
| **Kiem tra gi?** | Ky thuat (code, test, CI) | Nghiep vu (logic, quy trinh, truong hop) |
| **Diem** | /100 diem | % phu (coverage %) |
| **Cau hoi chinh** | Code co **tot** khong? | Logic co **day du** khong? |
| **Ai gioi?** | Developer / AI | **Ban** (nguoi hieu san pham) |
| **Vi von** | Kiem tra chat luong xay dung | Kiem tra co xay **du phong** khong |

### Vi du:
- Quality Audit: "Code dang ky hoc viet sach, test day du, API chuan" → 90/100
- Business Gap: "Nhung khong co quy trinh khi hoc sinh muon DOI lop" → 80% coverage

**Ca 2 PHAI pass truoc khi ship san pham.**

## "Lo hong" nghiep vu trong nhu the nao?

Day la nhung vi du thuc te:

### Gap 1: Co tinh nang nhung thieu quy tac
- Da code trang dang ky hoc
- Nhung **khong co quy tac**: hoc sinh dang ky bao nhieu lop la toi da?
- Hau qua: 1 hoc sinh dang ky 20 lop → qua tai

### Gap 2: Co API nhung thieu validation
- API `POST /enrollments` hoat dong
- Nhung **khong kiem tra**: hoc sinh da dang ky lop nay roi chua?
- Hau qua: dang ky trung lap, du lieu sai

### Gap 3: Co happy path nhung thieu error handling
- Dang ky thanh cong → OK
- Nhung **thanh toan that bai thi sao?** Rollback? Thong bao? Thu lai?
- Hau qua: tien bi tru nhung dang ky khong xac nhan

### Gap 4: Co tao nhung thieu xoa/huy
- Tao duoc dang ky moi
- Nhung **huy dang ky thi sao?** Hoan tien? Xoa du lieu? Cap nhat slot?
- Hau qua: khong huy duoc, hoac huy xong nhung slot khong duoc tra lai

## Doc hieu bao cao Gap Check

### Coverage (%)
- **80-100%**: Tot. Hau het nghiep vu da duoc cover.
- **60-79%**: Kha. Co mot so lo hong can bo sung.
- **Duoi 60%**: Yeu. Nhieu tinh nang thieu, can review lai.

### Priority levels

| Priority | Y nghia | Khi nao lam | Vi du |
|----------|---------|------------|-------|
| **P1** | Urgent | Truoc khi ship/deploy | Thieu validation quan trong, mat du lieu |
| **P2** | Important | Sprint toi | Thieu tinh nang phu, UX chua tot |
| **P3** | Nice-to-have | Khi co thoi gian | Tinh nang nang cao, toi uu |

### PASS / FAIL / WARN
- **PASS**: Co va hoat dong dung
- **FAIL**: **Thieu hoan toan** hoac sai logic → phai sua
- **WARN**: Co nhung chua day du, hoac chua kiem tra duoc → xem xet

## Tai sao NON-DEV nen chay Gap Check?

Day la ly do quan trong nhat:

### 1. BAN hieu san pham → BAN biet thieu gi
AI code dung ky thuat nhung **khong biet** san pham cua ban can nhung gi. Ví dụ:
- AI code trang dang ky hoan hao
- Nhung ban biet: "O truong minh, phu huynh thuong dang ky nhieu lop cung luc" → can tinh nang "dang ky hang loat"
- AI khong bao gio biet dieu nay neu ban khong noi

### 2. AI code dung nhung thieu
- AI implement chinh xac nhung gi ban yeu cau
- Nhung ban co the **quen yeu cau** 1 truong hop
- Gap Check giup ban **he thong hoa** viec kiem tra, khong bo sot

### 3. Gap Check = tieng noi cua khach hang
- Developer nghi: "code chay dung la duoc"
- Ban nghi: "khach hang co dung duoc suon se khong?"
- Gap Check dung goc nhin cua **nguoi dung**, khong phai goc nhin ky thuat

## Meo de Gap Check hieu qua

### 1. Chay sau moi "wave" lon
Lam xong 5-10 PRs cho 1 module? Chay Gap Check truoc khi chuyen sang module khac.

### 2. Moi nguoi doi sang tao checklist
Developer thuong nghi ve ky thuat. Nguoi khong biet code thuong nghi ve **nguoi dung**. Goc nhin cua ban quy gia hon ban nghi.

### 3. Dung so thay FAIL nhieu
FAIL khong co nghia la du an te. FAIL co nghia la ban **tim ra van de truoc khi khach hang gap**. Do la dieu tot.

### 4. Uu tien P1 truoc
Khong can fix het moi thu. Fix P1 (urgent) truoc, P2 trong sprint toi, P3 khi co thoi gian.

### 5. Luu bao cao
Moi lan chay Gap Check, luu ket qua de so sanh:
- Lan 1: 64% coverage
- Lan 2 (sau khi fix P1): 78%
- Lan 3 (sau khi fix P2): 89%

Thay tien bo giup **giu dong luc**.

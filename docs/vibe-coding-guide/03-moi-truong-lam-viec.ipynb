## 1. VS Code — "Phòng làm việc" của bạn

### Ví von: Phòng làm việc đa năng

VS Code (Visual Studio Code) là phần mềm nơi bạn:
- **Xem file** (tủ hồ sơ bên trái)
- **Đọc/sửa code** (bàn làm việc ở giữa)
- **Gõ lệnh** (bàn ra lệnh bên dưới — Terminal)
- **Cài thêm công cụ** (Extensions — giống cài app trên điện thoại)

Tại sao dùng VS Code mà không dùng Notepad hay Word?
- VS Code **hiểu code** — tô màu, gợi ý, báo lỗi
- VS Code **có Terminal tích hợp** — không cần mở thêm cửa sổ
- VS Code **có Extensions** — cài Claude Code ngay trong đó

### Cài đặt VS Code

1. Truy cập: **https://code.visualstudio.com/**
2. Nhấn nút **Download** (tự động chọn đúng hệ điều hành)
3. Cài đặt như bình thường (Next → Next → Install)
4. Mở VS Code lên

### Giao diện VS Code (4 vùng chính)

```
┌─────────────────────────────────────────────────────┐
│  Menu Bar (File, Edit, View...)                      │
├──────────┬──────────────────────────────────────────┤
│          │                                          │
│  Side    │         Editor                           │
│  Bar     │         (Viết/đọc code ở đây)            │
│          │                                          │
│  (File   │                                          │
│  Explorer│                                          │
│  Search  │                                          │
│  Git     │                                          │
│  ...)    │                                          │
│          │                                          │
│          ├──────────────────────────────────────────┤
│          │  Terminal Panel                           │
│          │  (Gõ lệnh ở đây — Claude Code chạy đây) │
├──────────┴──────────────────────────────────────────┤
│  Status Bar (thông tin branch git, ngôn ngữ...)      │
└─────────────────────────────────────────────────────┘
```

**Mẹo quan trọng**: Nhấn `` Ctrl+` `` (phím Ctrl + phím backtick, ở góc trái dưới phím Esc) để mở/đóng Terminal.

````bash
# Kiểm tra VS Code đã cài chưa
# Mở Terminal trong VS Code (Ctrl+`) và chạy lệnh này:

echo "=== Kiểm tra VS Code ==="
if command -v code &> /dev/null; then
    echo "VS Code da duoc cai dat!"
    code --version
else
    echo "Chua cai VS Code. Tai tu: https://code.visualstudio.com/"
fi
````

Extensions giống như **cài app trên điện thoại** — thêm tính năng cho VS Code.

| Extension | Tại sao cần | Cách cài |

**Cách cài Extension:**

## 2. Terminal — "Cửa sổ ra lệnh"

### Ví von: Nhắn tin cho máy tính

Bình thường bạn dùng máy tính bằng cách **click chuột** — mở folder, kéo file, nhấn nút.

Terminal là cách **gõ lệnh bằng chữ** để ra lệnh cho máy tính. Giống như **nhắn tin cho máy tính**:

```
Bạn gõ:  ls
Máy trả lời:  Documents  Downloads  Desktop  Pictures

Bạn gõ:  cd Documents
Máy: (im lặng = đã vào folder Documents thành công)

Bạn gõ:  mkdir project-moi
Máy: (tạo folder "project-moi" xong)
```

### Tại sao cần Terminal?

- Claude Code **chạy trong Terminal**
- Git **dùng qua Terminal**
- Nhiều công cụ lập trình **chỉ có qua Terminal**
- Terminal **nhanh hơn** click chuột rất nhiều (khi quen)

### Các lệnh Terminal cơ bản (chỉ 6 lệnh — đủ dùng!)

Bạn chỉ cần nhớ **6 lệnh** này. Nghiêm túc, chỉ 6 thôi:

| Lệnh | Ý nghĩa | Ví von |
|-------|---------|--------|
| `pwd` | Đang ở đâu? (Print Working Directory) | "Tôi đang đứng ở phòng nào?" |
| `ls` | Liệt kê file/folder | "Phòng này có gì?" |
| `cd` | Di chuyển đến folder | "Đi sang phòng kia" |
| `mkdir` | Tạo folder mới | "Xây phòng mới" |
| `clear` | Xóa màn hình Terminal | "Lau bảng" |
| `cat` | Đọc nội dung file | "Đọc tờ giấy này" |

````bash
# Lệnh 1: pwd — Bạn đang ở đâu?

echo "=== pwd: Print Working Directory ==="
echo "Bạn đang ở thư mục:"
pwd
````

````bash
# Lệnh 2: ls — Liệt kê nội dung folder

echo "=== ls: Liệt kê file và folder ==="
echo ""
echo "Nội dung folder home (~):"
ls ~
echo ""
echo "Thêm -la để xem chi tiết (kích thước, ngày sửa):"
ls -la ~ | head -10
````

````bash
# Lệnh 3 & 4: cd và mkdir — Di chuyển và tạo folder

echo "=== cd & mkdir ==="
echo ""

# Tạo folder thử nghiệm
mkdir -p /tmp/vibe-coding-demo
echo "Da tao folder: /tmp/vibe-coding-demo"

# Di chuyển vào folder
cd /tmp/vibe-coding-demo
echo "Da vao folder:"
pwd

# Tạo folder con
mkdir -p src tests docs
echo ""
echo "Da tao 3 folder con:"
ls

echo ""
echo "Meo: 'cd ..' de quay lai folder cha"
echo "     'cd ~' de ve folder home"
````

````bash
# Lệnh 5: cat — Đọc nội dung file

echo "=== cat: Đọc file ==="
echo ""

# Tạo file mẫu
echo "Xin chao! Day la file dau tien cua toi." > /tmp/vibe-coding-demo/hello.txt
echo "Toi dang hoc vibe coding." >> /tmp/vibe-coding-demo/hello.txt

# Đọc file
echo "Noi dung file hello.txt:"
cat /tmp/vibe-coding-demo/hello.txt
````

### Mẹo Terminal hữu ích

| Phím tắt | Tác dụng |
|----------|----------|
| **Tab** | Tự động hoàn thành tên file/folder (gõ "Doc" rồi Tab → "Documents") |
| **Mũi tên lên/xuống** | Xem lại lệnh đã gõ trước đó |
| **Ctrl+C** | Hủy lệnh đang chạy ("dừng lại!") |
| **Ctrl+L** | Xóa màn hình (giống `clear`) |

**Đừng sợ Terminal!** Bạn sẽ quen rất nhanh. Và khi dùng Claude Code, bạn chủ yếu gõ tiếng Việt mô tả việc cần làm — Claude Code sẽ chạy lệnh giúp bạn.

## 3. Git — "Máy thời gian cho code"

### Ví von: Save game

Bạn chơi game RPG, đến boss khó, bạn **save game** trước khi đánh. Nếu thua, **load lại** save cũ.

**Git hoạt động y hệt vậy cho code:**

| Game | Git | Ý nghĩa |
|------|-----|----------|
| Save game | `git commit` | Lưu trạng thái hiện tại |
| Load game | `git checkout` | Quay lại trạng thái cũ |
| Slot save 1, 2, 3... | Branch (nhánh) | Thử nghiệm song song |
| Cloud save | `git push` | Đẩy lên cloud (GitHub) |
| Download save | `git pull` | Tải về từ cloud |

### Tại sao cần Git?

1. **An toàn**: Code hỏng? Quay lại version trước
2. **Theo dõi lịch sử**: Ai sửa gì, lúc nào, tại sao
3. **Làm nhóm**: Nhiều người sửa cùng project mà không đè lên nhau
4. **Claude Code dùng Git**: Mỗi thay đổi được commit tự động

````bash
# Kiểm tra Git đã cài chưa

echo "=== Kiểm tra Git ==="
if command -v git &> /dev/null; then
    echo "Git da duoc cai dat!"
    git --version
else
    echo "Chua cai Git."
    echo "  macOS: Git thuong co san. Neu chua, chay: xcode-select --install"
    echo "  Windows: Tai tu https://git-scm.com/download/win"
    echo "  Linux: sudo apt install git"
fi
````

### Các lệnh Git bạn cần biết (chỉ 7 lệnh)

| Lệnh | Ý nghĩa | Ví von |
|-------|---------|--------|
| `git clone` | Tải project về máy | Download game |
| `git status` | Xem có gì thay đổi | Kiểm tra tiến độ |
| `git add` | Chọn file để lưu | Chọn slot save |
| `git commit` | Lưu thay đổi | Nhấn nút Save |
| `git push` | Đẩy lên GitHub | Upload save lên cloud |
| `git pull` | Tải về từ GitHub | Download save mới nhất |
| `git log` | Xem lịch sử commit | Xem danh sách save cũ |

**Tin tốt**: Khi dùng Claude Code, nó sẽ chạy các lệnh Git giúp bạn. Bạn chỉ cần nói *"commit và push code"* — Claude Code lo phần còn lại.

````bash
# Demo: Tạo Git repository (kho code) đầu tiên

echo "=== Demo Git cơ bản ==="
echo ""

# Tạo folder project mới
cd /tmp && rm -rf my-first-project
mkdir my-first-project && cd my-first-project

# Khởi tạo Git
git init
echo "=> Da khoi tao Git repository!"
echo ""

# Tạo file đầu tiên
echo "# Project dau tien cua toi" > README.md
echo "Day la project hoc vibe coding." >> README.md
echo "=> Da tao file README.md"
````

````bash
# Demo: git status, add, commit

cd /tmp/my-first-project

echo "=== git status: Xem thay đổi ==="
git status
echo ""
echo "(README.md mau do = chua duoc 'save')" 
echo ""

echo "=== git add: Chọn file để save ==="
git add README.md
git status
echo ""
echo "(README.md mau xanh = san sang de save)"
echo ""

echo "=== git commit: Lưu lại ==="
git commit -m "feat: tao project dau tien"
echo ""
echo "=> Da save thanh cong!"
````

````bash
# Demo: git log — Xem lịch sử

cd /tmp/my-first-project

# Thêm thay đổi và commit lần 2
echo "## Tinh nang" >> README.md
echo "- Quan ly hoc sinh" >> README.md
echo "- Diem danh" >> README.md
git add README.md
git commit -m "docs: them danh sach tinh nang"

echo ""
echo "=== git log: Lịch sử commit ==="
git log --oneline
echo ""
echo "=> Moi dong la 1 'save point' — ban co the quay lai bat ky luc nao!"
````

### GitHub — "Google Drive cho code"

**Git** = công cụ quản lý version trên máy bạn (offline)

**GitHub** = website lưu trữ Git project trên cloud (online)

| Google Drive | GitHub |
|-------------|--------|
| Lưu file Word, Excel | Lưu code |
| Chia sẻ link cho người khác | Chia sẻ project cho team |
| Version history | Commit history |
| Cộng tác real-time | Pull Request (đề xuất thay đổi) |

**Cách tạo tài khoản GitHub:**
1. Truy cập: **https://github.com**
2. Nhấn **Sign up**
3. Điền email, tạo password, chọn username
4. Xong! Bạn có nơi lưu code trên cloud

````bash
# Demo: git clone — Tải project từ GitHub

echo "=== git clone: Tải project từ GitHub ==="
echo ""
echo "Cú pháp:"
echo "  git clone https://github.com/ten-nguoi-dung/ten-project.git"
echo ""
echo "Ví dụ thực tế:"
echo "  git clone https://github.com/VictorAurelius/claude-starter-kit.git"
echo ""
echo "Sau khi clone:"
echo "  cd claude-starter-kit    # Vào folder project"
echo "  ls                        # Xem nội dung"
echo "  claude                    # Khởi động Claude Code!"
````

### Branch — "Thế giới song song"

Branch (nhánh) là tính năng mạnh nhất của Git. Ví von:

Bạn đang viết luận văn (bản chính = **main**). Bạn muốn thử viết lại chương 3 theo hướng khác:
- Tạo **bản copy** (branch mới) → thử nghiệm thoải mái
- Nếu hay → **merge** (gộp) vào bản chính
- Nếu tệ → **xóa** bản copy, bản chính không bị ảnh hưởng

```
main:          A --- B --- C --- D --- E (bản chính)
                         \
feature/login:            X --- Y --- Z (thử nghiệm)
                                       \
                                        merge → E bao gồm X, Y, Z
```

Khi vibe coding với Claude Code:
- **main** = bản sản phẩm ổn định
- **feature/xxx** = đang làm tính năng mới
- Claude Code tự tạo branch và merge cho bạn

## 4. Lưu ý cho Windows: WSL

### Vấn đề

Phần lớn công cụ lập trình được thiết kế cho **Linux/macOS**. Nếu bạn dùng **Windows**, một số thứ sẽ không hoạt động trơn tru.

### Giải pháp: WSL (Windows Subsystem for Linux)

WSL = chạy **Linux bên trong Windows**. Ví von: giống như chạy app Android trên máy tính Windows.

**Bạn có cần WSL không?**
- Dùng **macOS** → KHÔNG cần (Terminal macOS đã tương thích)
- Dùng **Linux** → KHÔNG cần
- Dùng **Windows** → NÊN cài WSL

````bash
# Cài WSL trên Windows
# (Chạy lệnh này trong PowerShell với quyền Admin)

echo "=== Cài WSL trên Windows ==="
echo ""
echo "Bước 1: Mở PowerShell với quyền Admin"
echo "  (Click phải Start Menu → 'Windows Terminal (Admin)')"
echo ""
echo "Bước 2: Chạy lệnh:"
echo "  wsl --install"
echo ""
echo "Bước 3: Khởi động lại máy tính"
echo ""
echo "Bước 4: Mở Ubuntu từ Start Menu, tạo username & password"
echo ""
echo "Bước 5: Trong VS Code, cài extension 'WSL'"
echo "  (Tìm 'WSL' trong Extensions → Install)"
echo ""
echo "Bước 6: Mở VS Code, nhấn F1 → gõ 'WSL: Connect to WSL'"
echo "  => Bây giờ VS Code chạy bên trong Linux!"
````

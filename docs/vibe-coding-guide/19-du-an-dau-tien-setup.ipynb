# 19. Dự Án Đầu Tiên — Setup Từ Con Số 0

**Phần 5: Thực hành từ đầu đến cuối**

---

Hãy tưởng tượng bạn sắp bắt đầu một ngày làm việc mới. Việc đầu tiên bạn làm là gì? **Dọn bàn làm việc**, pha cà phê, mở máy tính, chuẩn bị mọi thứ sẵn sàng.

Setup dự án cũng y hệt vậy — bạn đang "dọn phòng làm việc" cho AI trước khi bắt đầu.

Trong bài này, chúng ta sẽ đi từ **con số 0 tuyệt đối** đến một dự án có đầy đủ công cụ, sẵn sàng để Claude Code giúp bạn xây dựng sản phẩm.

## Bước 1: Cài Đặt Công Cụ

Giống như thợ mộc cần búa, cưa, thước — bạn cần vài công cụ cơ bản trước khi bắt đầu.

### 1.1 VS Code — Phòng làm việc của bạn

VS Code là nơi bạn nhìn thấy code, tương tác với Claude, và xem kết quả.

- Tải tại: https://code.visualstudio.com/
- Cài đặt bình thường (Next → Next → Finish)

### 1.2 Claude Code Extension

Đây là "người trợ lý AI" ngồi ngay trong VS Code.

- Mở VS Code → Extensions (biểu tượng ô vuông bên trái) → Tìm "Claude Code" → Install

### 1.3 Git — Sổ nhật ký dự án

Git giúp lưu lại mọi thay đổi. Thường máy đã cài sẵn. Hãy kiểm tra:

````bash
# Kiểm tra Git đã cài chưa
git --version
````

Nếu thấy dòng `git version 2.x.x` là OK. Nếu báo lỗi "command not found", bạn cần cài Git:
- **Windows:** https://git-scm.com/download/win
- **Mac:** Mở Terminal, gõ `xcode-select --install`
- **Linux:** `sudo apt install git`

### 1.4 Node.js — "Động cơ" cho ứng dụng web

Nếu bạn làm web (phần lớn dự án), cần Node.js:

````bash
# Kiểm tra Node.js
node --version
npm --version
````

Nếu chưa có, tải tại https://nodejs.org/ (chọn phiên bản LTS).

## Bước 2: Tạo Project Trống

Bây giờ bạn đã có đủ công cụ. Hãy tạo một "phòng làm việc" trống — nơi dự án sẽ sống.

````bash
# Tạo thư mục dự án (thay "my-first-project" thành tên bạn muốn)
mkdir my-first-project

# Di chuyển vào thư mục
cd my-first-project

# Khởi tạo Git — bắt đầu ghi nhật ký
git init

echo "Da tao xong project!"
````

Lúc này bạn có một thư mục trống với Git đã sẵn sàng. Giống như phòng đã dọn sạch, bàn ghế kê đúng chỗ — chỉ chờ bạn bắt tay vào làm.

## Bước 3: Cài Starter Kit

Starter Kit giống như **bộ đồ nghề có sẵn** — thay vì mua từng món (búa, cưa, đinh...), bạn mua nguyên bộ.

Starter Kit mang lại:
- File CLAUDE.md ("hồ sơ dự án" cho AI đọc)
- Các skill sẵn (brainstorm, review code, quality check...)
- Scripts tự động (kiểm tra lỗi, format code...)
- Templates (mẫu sẵn cho nhiều thứ)

````bash
# Tải Starter Kit về máy (chỉ cần 1 lần)
git clone https://github.com/VictorAurelius/claude-starter-kit.git /tmp/kit

echo "Da tai Starter Kit!"
````

````bash
# Cài Kit vào project của bạn
# Dấu "." nghĩa là "ngay thư mục hiện tại"
cd my-first-project
bash /tmp/kit/bin/init-project.sh .

echo "Da cai Starter Kit vao project!"
````

````bash
# Xem những gì Kit đã tạo ra
cd my-first-project
ls -la
echo "---"
echo "Cac file va thu muc da duoc tao:"
find . -maxdepth 2 -not -path './.git/*' | head -30
````

## Bước 4: Customize CLAUDE.md

Đây là bước **quan trọng nhất**. CLAUDE.md giống như "sơ yếu lý lịch" của dự án — Claude sẽ đọc file này đầu tiên mỗi khi bắt đầu làm việc.

Template mặc định có các placeholder (chỗ trống) cần bạn điền vào:

| Placeholder | Thay bằng | Ví dụ |
|-------------|----------|-------|
| `{LANGUAGE}` | Ngôn ngữ giao tiếp | `Vietnamese` |
| `{PROJECT_NAME}` | Tên dự án | `My Coffee Shop App` |
| Architecture section | Mô tả cấu trúc | `Next.js frontend + Express backend` |
| Scripts section | Liệt kê scripts | `npm run dev`, `npm test` |

````bash
# Xem nội dung CLAUDE.md hiện tại
cd my-first-project
cat CLAUDE.md | head -40
echo ""
echo "... (con tiep)"
````

````bash
# Ví dụ: thay thế placeholder cho dự án "My Coffee Shop"
cd my-first-project

# Thay ngôn ngữ
sed -i 's/{LANGUAGE}/Vietnamese/g' CLAUDE.md

# Thay tên dự án
sed -i 's/{PROJECT_NAME}/My Coffee Shop App/g' CLAUDE.md

echo "Da customize CLAUDE.md!"
echo "---"
head -10 CLAUDE.md
````

**Mẹo:** Bạn cũng có thể mở CLAUDE.md trong VS Code và sửa trực tiếp bằng tay — không nhất thiết phải dùng lệnh. Quan trọng là điền đủ thông tin để Claude hiểu dự án của bạn.

## Bước 5: Verify — Kiểm tra Claude đã hiểu chưa

Giống như khi bạn giao việc cho nhân viên mới, bạn cần hỏi lại: "Em hiểu chưa? Kể lại cho anh/chị nghe."

### Cách kiểm tra:

1. Mở VS Code trong thư mục project
2. Mở Claude Code (Ctrl+Shift+P → "Claude Code")
3. Gõ:

```
Đọc CLAUDE.md và cho tôi biết bạn hiểu gì về dự án này
```

### Claude nên trả lời được:
- Tên dự án là gì
- Dùng công nghệ gì
- Ngôn ngữ giao tiếp là tiếng Việt
- Quy trình làm việc (brainstorm → task breakdown → TDD → implement → review)

Nếu Claude trả lời mơ hồ hoặc sai, bạn cần quay lại Bước 4 và bổ sung thông tin.

````bash
# Mở VS Code trong thư mục project (nếu dùng terminal)
cd my-first-project
code .
````

## Bước 6: First Commit — Lưu lại "trạng thái ban đầu"

Như chụp ảnh "trước khi sửa nhà" — bạn lưu lại trạng thái ban đầu để luôn có thể quay về nếu cần.

````bash
cd my-first-project

# Thêm tất cả file vào Git
git add -A

# Lưu lại (commit) với ghi chú
git commit -m "chore: init project with claude-starter-kit v2.1.0"

echo "---"
echo "Commit dau tien thanh cong!"
git log --oneline
````

## Gotchas — Những Lỗi Hay Gặp Khi Mới Bắt Đầu

### 1. Quên `git init` trước khi cài Kit
Kit cần Git để hoạt động. Nếu bạn cài Kit trước khi `git init`, một số tính năng (hooks, commit checks) sẽ không chạy.

**Cách sửa:** Chạy `git init` rồi cài lại Kit.

### 2. Không customize CLAUDE.md
Template mặc định toàn placeholder `{PROJECT_NAME}`, `{LANGUAGE}`. Nếu không thay, Claude sẽ không hiểu gì về dự án của bạn — giống như giao việc mà không nói rõ yêu cầu.

**Cách sửa:** Mở CLAUDE.md, thay tất cả placeholder bằng thông tin thực.

### 3. Commit file `settings.local.json`
File này chứa cài đặt riêng của bạn (API key, preferences cá nhân). Starter Kit đã thêm nó vào `.gitignore` nên bình thường sẽ không bị commit. Nhưng nếu bạn force add, nó sẽ lộ thông tin nhạy cảm.

**Quy tắc:** Không bao giờ commit file có chứa key, password, hoặc thông tin cá nhân.

````bash
# Kiểm tra .gitignore đã chặn file nhạy cảm chưa
cd my-first-project
echo "Cac file bi gitignore:"
cat .gitignore 2>/dev/null || echo "(chua co .gitignore)"
````
